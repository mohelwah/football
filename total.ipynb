## Adds the number of games played by the team before specific match

```python
def get_last_matches_number(playing_stat):
    
    # Diffrent lists for HomeTeam and AwayTeam because sometimes number of games played before 
    # can be diffrent becasue of postponing the match
    HT_Match_number = []
    AT_Match_number = []
    
    # dataset iteration over each row
    for i, row in playing_stat.iterrows():
        df_nad = playing_stat.iloc[0:i]
        df_specific_match = playing_stat.iloc[i]

        ht = df_specific_match.HomeTeam
        at = df_specific_match.AwayTeam

        x = len(df_nad[(df_nad.HomeTeam == ht) | (df_nad.AwayTeam == ht)])
        y = len(df_nad[(df_nad.HomeTeam == at) | (df_nad.AwayTeam == at)])

        HT_Match_number.append(x)
        AT_Match_number.append(y)
        
    playing_stat['HT_last_matches_number'] = HT_Match_number
    playing_stat['AT_last_matches_number'] = AT_Match_number
    
    return playing_stat
```

```python
# apply function on every season data
for season in seasons:
    season = get_last_matches_number(season)
```

## Goals scored and conceded before specific match

```python
# Gets the goals scored aggregation arranged by teams and matchweek
def get_goals_scored(playing_stat):
    # Create a dictionary with team names as keys
    teams = {}
    for i in playing_stat.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
    
    # the value corresponding to keys is a list containing the match location.
    for i in range(len(playing_stat)):
        HTGS = playing_stat.iloc[i]['FTHG']
        ATGS = playing_stat.iloc[i]['FTAG']
        teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
        teams[playing_stat.iloc[i].AwayTeam].append(ATGS)
        
    # Create a dataframe for goals scored where rows are teams and cols are matchweek.
    GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
    GoalsScored[0] = 0
    # Aggregate to get uptil that point
    for i in range(2,39): # we have 38 matchweeks (number of teams*2 - 2 becacue you cant play with yourself)
        GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
    return GoalsScored

# Gets the goals conceded aggregation arranged by teams and matchweek
def get_goals_conceded(playing_stat):
    teams = {}
    for i in playing_stat.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
    
    # the value corresponding to keys is a list containing the match location.
    for i in range(len(playing_stat)):
        ATGC = playing_stat.iloc[i]['FTHG']
        HTGC = playing_stat.iloc[i]['FTAG']
        teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
        teams[playing_stat.iloc[i].AwayTeam].append(ATGC)
    
    # Create a dataframe for goals conceded where rows are teams and cols are matchweek.
    GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
    GoalsConceded[0] = 0
    # Aggregate to get uptil that point
    for i in range(2,len(teams["Arsenal"])+1):
        GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
    return GoalsConceded

# function summarizing and adding variables
def get_gss(playing_stat):
    GC = get_goals_conceded(playing_stat)
    GS = get_goals_scored(playing_stat)
   
    HTGS = []
    ATGS = []
    HTGC = []
    ATGC = []
    
    for i in range(len(playing_stat)):
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam

        ht_match = playing_stat.iloc[i].HT_last_matches_number
        at_match = playing_stat.iloc[i].AT_last_matches_number

        HTGS.append(GS.loc[ht][ht_match])
        ATGS.append(GS.loc[at][at_match])
        
        HTGC.append(GC.loc[ht][ht_match])
        ATGC.append(GC.loc[at][at_match])
        
        
    playing_stat['HTGS'] = HTGS
    playing_stat['ATGS'] = ATGS
    playing_stat['HTGC'] = HTGC
    playing_stat['ATGC'] = ATGC
    
    return playing_stat
```

```python
for season in seasons:
    season = get_gss(season)
```

## Shots on target took and allowed before specific match

```python
# Gets the shots on target took aggregation arranged by teams and matchweek
def get_shots_on_target_took(playing_stat):
    # Create a dictionary with team names as keys
    teams = {}
    for i in playing_stat.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
    
    # the value corresponding to keys is a list containing the match location.
    for i in range(len(playing_stat)):
        HTSOT = playing_stat.iloc[i]['HST']
        ATSOT = playing_stat.iloc[i]['AST']
        teams[playing_stat.iloc[i].HomeTeam].append(HTSOT)
        teams[playing_stat.iloc[i].AwayTeam].append(ATSOT)
        
    # Create a dataframe for shots on target took where rows are teams and cols are matchweek.
    ShotsOT = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
    ShotsOT[0] = 0
    # Aggregate to get uptil that point
    for i in range(2,39):
        ShotsOT[i] = ShotsOT[i] + ShotsOT[i-1]
    return ShotsOT

# Gets the shots on target allowed aggregation arranged by teams and matchweek
def get_shots_on_target_allowed(playing_stat):
    # Create a dictionary with team names as keys
    teams = {}
    for i in playing_stat.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
    
    # the value corresponding to keys is a list containing the match location.
    for i in range(len(playing_stat)):
        ATSOT_A = playing_stat.iloc[i]['HST']
        HTSOT_A = playing_stat.iloc[i]['AST']
        teams[playing_stat.iloc[i].HomeTeam].append(HTSOT_A)
        teams[playing_stat.iloc[i].AwayTeam].append(ATSOT_A)
    
    # Create a dataframe for shots on target allowed where rows are teams and cols are matchweek.
    ShotsOT_Allowed = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
    ShotsOT_Allowed[0] = 0
    # Aggregate to get uptil that point
    for i in range(2,len(teams["Arsenal"])+1):
        ShotsOT_Allowed[i] = ShotsOT_Allowed[i] + ShotsOT_Allowed[i-1]
    return ShotsOT_Allowed

# function summarizing and adding variables
def get_sot(playing_stat):
    SOT_T = get_shots_on_target_took(playing_stat)
    SOT_A = get_shots_on_target_allowed(playing_stat)
   
    HTSOT_T = []
    ATSOT_T = []
    HTSOT_A = []
    ATSOT_A = []
    
    for i in range(len(playing_stat)):
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam

        ht_match = playing_stat.iloc[i].HT_last_matches_number
        at_match = playing_stat.iloc[i].AT_last_matches_number

        HTSOT_T.append(SOT_T.loc[ht][ht_match])
        ATSOT_T.append(SOT_T.loc[at][at_match])
        
        HTSOT_A.append(SOT_A.loc[ht][ht_match])
        ATSOT_A.append(SOT_A.loc[at][at_match])
        
        
    playing_stat['HTSOT_T'] = HTSOT_T
    playing_stat['ATSOT_T'] = ATSOT_T
    playing_stat['HTSOT_A'] = HTSOT_A
    playing_stat['ATSOT_A'] = ATSOT_A
    
    return playing_stat
```

```python
for season in seasons:
    season = get_sot(season)
```

```python
# Get shots accuracy
def get_shots_acc(playing_stat):
    
    HT_Shot_Acc = playing_stat.HTGS / playing_stat.HTSOT_T # HomeTeam shots took accuracy (Goals scoring accuracy)
    AT_Shot_Acc = playing_stat.ATGS / playing_stat.ATSOT_T # AwayTeam shots took accuracy (Goals scoring accuracy)
    
    HT_Op_Shot_Acc = playing_stat.HTGC / playing_stat.HTSOT_A # HomeTeam shots allowed accuracy (Goals conceded accuracy)
    AT_Op_Shot_Acc = playing_stat.ATGC / playing_stat.ATSOT_A # AwayTeam shots allowed accuracy (Goals conceded accuracy)
    
    playing_stat['HT_Shot_Acc'] = HT_Shot_Acc
    playing_stat['AT_Shot_Acc'] = AT_Shot_Acc
    playing_stat['HT_Op_Shot_Acc'] = HT_Op_Shot_Acc
    playing_stat['AT_Op_Shot_Acc'] = AT_Op_Shot_Acc
    
    return playing_stat
```

```python
for season in seasons:
    season = get_shots_acc(season)
```

## Cumulative points before specific match

```python
# Function for getting specific number of points depends on the match result
def get_points(result):
    if result == 'W':
        return 3
    elif result == 'D':
        return 1
    else:
        return 0
    
# Get match results for every team in every match week
def get_matchres(season):
    # Create a dictionary with team names as keys
    teams = {}
    for i in season.groupby('HomeTeam').mean().T.columns:
        teams[i] = []

    # the value corresponding to keys is a list containing the match result
    for i in range(len(season)):
        if season.iloc[i].FTR == 'H':
            teams[season.iloc[i].HomeTeam].append('W') 
            teams[season.iloc[i].AwayTeam].append('L') # If HomeTeam wins the AwayTeam lose so wee need to add L
        elif season.iloc[i].FTR == 'A':
            teams[season.iloc[i].AwayTeam].append('W') 
            teams[season.iloc[i].HomeTeam].append('L') # If AwayTeam wins the HomeTeam lose so wee need to add L
        else:
            teams[season.iloc[i].AwayTeam].append('D')
            teams[season.iloc[i].HomeTeam].append('D')

    return pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T

# Get cumulative points connected with match results
def get_cuml_points(matchres):
    matchres_points = matchres.applymap(get_points)
    for i in range(2,38):
        matchres_points[i] = matchres_points[i] + matchres_points[i-1]
        
    matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(20)])
    return matchres_points

# function summarizing and adding variables
def get_agg_points(season):
    matchres = get_matchres(season)
    cum_pts = get_cuml_points(matchres)
    HTP = []
    ATP = []

    for i in range(len(season)):
        ht = season.iloc[i].HomeTeam
        at = season.iloc[i].AwayTeam
        
        ht_match = season.iloc[i].HT_last_matches_number
        at_match = season.iloc[i].AT_last_matches_number
        
        HTP.append(cum_pts.loc[ht][ht_match])
        ATP.append(cum_pts.loc[at][at_match])
            
    season['HTP'] = HTP
    season['ATP'] = ATP
    return season
```

```python
for season in seasons:
    season = get_agg_points(season)
```

## Current teams form with match results for 5 last games

```python
# get and add to dataset last 5 match results
def add_form(playing_stat,num):
    form = get_matchres(playing_stat)
    
    for number in range(5+1): # I am interesing to 5 last games
        form[-number] = 'ND' # No data
    
    HT_form = []
    AT_form = []
    
    for i in range(len(playing_stat)):
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam

        ht_match = playing_stat.iloc[i].HT_last_matches_number
        at_match = playing_stat.iloc[i].AT_last_matches_number

        HT_form.append(form.loc[ht][ht_match - (num - 1)])
        AT_form.append(form.loc[at][at_match - (num - 1)])
    
    # every last match will have different variable  
    playing_stat['HM' + str(num)] = HT_form                
    playing_stat['AM' + str(num)] = AT_form
    
    return playing_stat

# function summarizing and adding variables
def add_form_df(playing_statistics):
    playing_statistics = add_form(playing_statistics,1)
    playing_statistics = add_form(playing_statistics,2)
    playing_statistics = add_form(playing_statistics,3)
    playing_statistics = add_form(playing_statistics,4)
    playing_statistics = add_form(playing_statistics,5)
    return playing_statistics
```

## Cumulative number of wins, losses nad draws before specific match

```python
# Binary variables connected with match results
def get_wins(result):
    if result == 'W':
        return 1
    elif result == 'L' or  result == 'D':
        return 0
    else:
        return np.nan 

def get_losses(result):
    if result == 'L':
        return 1
    elif result == 'D' or  result == 'W':
        return 0
    else:
        return np.nan     

def get_draws(result):
    if result == 'D':
        return 1
    elif result == 'L' or  result == 'W':
        return 0
    else:
        return np.nan 
    
# Get cumulative results during specific season
def get_agg_results(season):
    matchres = get_matchres(season)
    cum_wins = matchres.applymap(get_wins).cumsum(axis=1)
    cum_losses = matchres.applymap(get_losses).cumsum(axis=1)
    cum_draws = matchres.applymap(get_draws).cumsum(axis=1)
    
    cum_wins[0] = 0
    cum_losses[0] = 0
    cum_draws[0] = 0
    
    HT_cum_wins = []
    AT_cum_wins = []
    
    HT_cum_losses = []
    AT_cum_losses = []
    
    HT_cum_draws = []
    AT_cum_draws = []
    
    for i in range(len(season)):
        ht = season.iloc[i].HomeTeam
        at = season.iloc[i].AwayTeam
        
        ht_match = season.iloc[i].HT_last_matches_number
        at_match = season.iloc[i].AT_last_matches_number
        
        HT_cum_wins.append(cum_wins.loc[ht][ht_match])
        AT_cum_wins.append(cum_wins.loc[at][at_match])
        
        HT_cum_losses.append(cum_losses.loc[ht][ht_match])
        AT_cum_losses.append(cum_losses.loc[at][at_match])
        
        HT_cum_draws.append(cum_draws.loc[ht][ht_match])
        AT_cum_draws.append(cum_draws.loc[at][at_match])
            
    season['HT_cum_wins'] = HT_cum_wins
    season['AT_cum_wins'] = AT_cum_wins
    
    season['HT_cum_losses'] = HT_cum_losses
    season['AT_cum_losses'] = AT_cum_losses
    
    season['HT_cum_draws'] = HT_cum_draws
    season['AT_cum_draws'] = AT_cum_draws
    
    return season
```

```python
for season in seasons:
    season = get_agg_results(season)
```

## Offensive and Defensive strength

The entire calculation process is presented in the link below:
* [Calcuation process](https://help.smarkets.com/hc/en-gb/articles/115001457989-How-to-calculate-Poisson-distribution-for-football-betting)

```python
def get_off_def_strength(playing_stat):
    
    List_off_strength_HT = []
    List_def_strength_HT = []
    List_off_strength_AT = []
    List_def_strength_AT = []

    for i in range(len(playing_stat)):
        df_specific_match = playing_stat.iloc[i]
        df_day_before_SpecMat = playing_stat[playing_stat.Date < df_specific_match.Date]

        ht = df_specific_match.HomeTeam
        at = df_specific_match.AwayTeam

        Avg_score_Home = np.mean(df_day_before_SpecMat.FTHG)
        Avg_score_Away = np.mean(df_day_before_SpecMat.FTAG)

        Avg_conced_Home = Avg_score_Away
        Avg_conced_Away = Avg_score_Home

        # Home Team offensive strength
        Avg_home_goals_scored_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTHG.mean()
        Off_strength_HT = Avg_home_goals_scored_by_HT / Avg_score_Home

        # Away Team defensive strength
        Avg_away_goals_conceded_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTHG.mean()
        Def_strength_AT = Avg_away_goals_conceded_by_AT / Avg_conced_Away

        List_off_strength_HT.append(Off_strength_HT)
        List_def_strength_AT.append(Def_strength_AT)

        # Away Team offensive strength
        Avg_away_goals_scored_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTAG.mean()
        Off_strength_AT = Avg_away_goals_scored_by_AT / Avg_score_Away

        # Home Team defensive strength
        Avg_home_goals_conceded_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTAG.mean()
        Def_strength_HT = Avg_home_goals_conceded_by_HT / Avg_conced_Home

        List_off_strength_AT.append(Off_strength_AT)
        List_def_strength_HT.append(Def_strength_HT)

    playing_stat['Off_strength_HT'] = List_off_strength_HT
    playing_stat['Def_strength_HT'] = List_def_strength_HT
    playing_stat['Off_strength_AT'] = List_off_strength_AT
    playing_stat['Def_strength_AT'] = List_def_strength_AT
  
```

```python
for season in seasons:
    season = get_off_def_strength(season)
```

## League position from last season

In European football leauges every season teams (usually 3 but it depends) can get promotion to the highest league and other teams can fall out. This way we get different teams every seasons. So to get league postion from last season I decided to assign to new teams a specific postion:
* number of teams in the highest league + new team's league postion from the second highest league
* for example in 2020/21 season Leeds was new team and they have 1st postion in 2019/20 season in the second highest league so I assign number 21

```python
Standings = pd.read_csv('league_tables/BPL_table_rank.csv', sep=';') # This table was created manually 
Standings.set_index(['Team'], inplace=True)
```

```python
# Add variables with last league postion
def get_last(playing_stat, Standings, year):
    HomeTeamLP = []
    AwayTeamLP = []
    for i in range(380):
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam
        HomeTeamLP.append(Standings.loc[ht][year])
        AwayTeamLP.append(Standings.loc[at][year])
    playing_stat['HomeTeamLP'] = HomeTeamLP
    playing_stat['AwayTeamLP'] = AwayTeamLP
    return playing_stat
```

## Calcuating betting odds
* 1x (Home win or Draw) = 100 / (100/Home_win_odds + 100/Draw_odds)
* x2 (Away win or Draw) = 100 / (100/Away_win_odds + 100/Draw_odds)

```python
def get_more_odds(playing_stat):
    x1 = []
    x2 = []
    
    for i, row in playing_stat.iterrows():
        b365_H = playing_stat.iloc[i]['B365H']
        b365_D = playing_stat.iloc[i]['B365D']
        b365_A = playing_stat.iloc[i]['B365A']
        
        odd_H = 100/b365_H
        odd_D = 100/b365_D
        odd_A = 100/b365_A
        
        b365_H_D = round((100/(odd_H + odd_D)), 2)
        b365_A_D = round((100/(odd_A + odd_D)), 2)
        
        x1.append(b365_H_D)
        x2.append(b365_A_D)
        
    
    playing_stat['B365_HD'] = x1
    playing_stat['B365_AD'] = x2
    
    return playing_stat
```

## Head to Head (H2H) variables
* Getting last 3 head to head match results and change it for points

```python
def h2h(playing_stat):
    
    playing_stat["H2H_Home"] = ""
    playing_stat["H2H_Away"] = ""

    for i in range(1,len(playing_stat)):
        # get teams from specifc match and all data before this match
        df = playing_stat.iloc[0:i-1,]
        ht = playing_stat.loc[i,"HomeTeam"]
        at = playing_stat.loc[i,"AwayTeam"]
        
        # getting last 3 head to head matches
        dff = df[(((df.HomeTeam == ht) &  (df.AwayTeam == at)) | ((df.HomeTeam == at) &  (df.AwayTeam == ht)))].tail(3)
        
        # using data with minimum 1 head to head match
        if (len(dff) >= 1):
            for j in range(0,len(dff)):
                # Consider all possible scenarios
                if ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "H")):
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "H")):
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "A")):
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "A")):
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "D")):
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "D")):
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                    playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                    
        # if there was no match beetwen the teams get 'ND'
        else: 
            playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] = "ND"
            playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] = "ND"
    return playing_stat


# Calculating points for Home and Away teams in head to head matches and adding variables
def get_points_h2h(playing_stat, col):
    
    playing_stat[col+"_pts"] = 0
    
    for i in range(1,len(playing_stat)):
        if (len(playing_stat.iloc[i][col]) == 1):
            if playing_stat.iloc[i][col][0] =='L':
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
            elif playing_stat.iloc[i][col][0] =='W':
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
            else:
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
        elif (len(playing_stat.iloc[i][col]) == 2):
            for j in range(0,2):
                if playing_stat.iloc[i][col][j] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][j] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                elif playing_stat.iloc[i][col][j] =='D':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
                else:
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
        elif (len(playing_stat.iloc[i][col]) == 3):
            for j in range(0,3):
                if playing_stat.iloc[i][col][j] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][j] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                elif playing_stat.iloc[i][col][j] =='D':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
        else:
            playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] = 0
    
    return playing_stat
```

```python
def BPL_dataset_creator():

    # -*- coding: utf-8 -*-
    """BPL_dataset_creator.ipynb

    Automatically generated by Colaboratory.

    Original file is located at
        https://colab.research.google.com/drive/1iRP3H_01n-CLzzfVhUJyqdOdHN-nPpq9

    ## Uploading data and basic operations (add season variable and change the date format)
    """

    import numpy as np
    import pandas as pd

    league_code = 'E0'
    seasons = ["00-01", "01-02", "02-03", "03-04", "04-05", 
              "05-06", "06-07", "07-08", "08-09", "09-10", 
              "10-11", "11-12", "12-13", "13-14", "14-15", 
              "15-16", "16-17", "17-18", "18-19", "19-20", "20-21"]

    urls = []
    for season in seasons:
        x = 'https://www.football-data.co.uk/mmz4281/' + season.replace("-", "") + '/'+league_code + '.csv'
        urls.append(x)

    columns = ['Div', 'Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'B365H', 'B365D', 'B365A']

    """I want to focus on the last 10 completed seasons (from 2011/12 to 2020/21) but i need more historical data to calculate H2H (Head To Head) variables"""

    BPL_00_01_data = pd.read_csv(urls[0])[columns[0:-3]]
    BPL_01_02_data = pd.read_csv(urls[1])[columns[0:-3]]
    BPL_02_03_data = pd.read_csv(urls[2])[columns[0:-3]]
    BPL_03_04_data =  pd.read_csv(urls[1])[columns[0:-3]]
    BPL_04_05_data =  pd.read_csv(urls[1])[columns[0:-3]]
    #BPL_03_04_data = pd.read_csv("match_data/E0_03_04.csv", sep=';')[columns[0:-3]] # Error Tokenizing Data with urls upload
    #BPL_04_05_data = pd.read_csv("match_data/E0_04_05.csv", sep=';', encoding= 'unicode_escape')[columns[0:-3]] # Error Tokenizing Data with urls upload
    BPL_05_06_data = pd.read_csv(urls[5])[columns[0:-3]]
    BPL_06_07_data = pd.read_csv(urls[6])[columns[0:-3]]
    BPL_07_08_data = pd.read_csv(urls[7])[columns[0:-3]]
    BPL_08_09_data = pd.read_csv(urls[8])[columns[0:-3]]
    BPL_09_10_data = pd.read_csv(urls[9])[columns[0:-3]]
    BPL_10_11_data = pd.read_csv(urls[10])[columns[0:-3]]
    BPL_11_12_data = pd.read_csv(urls[11])[columns]
    BPL_12_13_data = pd.read_csv(urls[12])[columns]
    BPL_13_14_data = pd.read_csv(urls[13])[columns]
    BPL_14_15_data = pd.read_csv(urls[14])[columns]
    BPL_15_16_data = pd.read_csv(urls[15])[columns]
    BPL_16_17_data = pd.read_csv(urls[16])[columns]
    BPL_17_18_data = pd.read_csv(urls[17])[columns]
    BPL_18_19_data = pd.read_csv(urls[18])[columns]
    BPL_19_20_data = pd.read_csv(urls[19])[columns]
    BPL_20_21_data = pd.read_csv(urls[20])[columns]

    BPL_20_21_data['Season'] = '2020/21'
    BPL_19_20_data['Season'] = '2019/20'
    BPL_18_19_data['Season'] = '2018/19'
    BPL_17_18_data['Season'] = '2017/18'
    BPL_16_17_data['Season'] = '2016/17'
    BPL_15_16_data['Season'] = '2015/16'
    BPL_14_15_data['Season'] = '2014/15'
    BPL_13_14_data['Season'] = '2013/14'
    BPL_12_13_data['Season'] = '2012/13'
    BPL_11_12_data['Season'] = '2011/12'
    BPL_10_11_data['Season'] = '2010/11'
    BPL_09_10_data['Season'] = '2009/10'
    BPL_08_09_data['Season'] = '2008/09'
    BPL_07_08_data['Season'] = '2007/08'
    BPL_06_07_data['Season'] = '2006/07'
    BPL_05_06_data['Season'] = '2005/06'
    BPL_04_05_data['Season'] = '2004/05'
    BPL_03_04_data['Season'] = '2003/04'
    BPL_02_03_data['Season'] = '2002/03'
    BPL_01_02_data['Season'] = '2001/02'
    BPL_00_01_data['Season'] = '2000/01'

    seasons = [BPL_00_01_data, BPL_01_02_data, BPL_02_03_data, BPL_03_04_data, BPL_04_05_data,
              BPL_05_06_data, BPL_06_07_data, BPL_07_08_data, BPL_08_09_data, BPL_09_10_data, 
              BPL_10_11_data, BPL_11_12_data, BPL_12_13_data, BPL_13_14_data, BPL_14_15_data, 
              BPL_15_16_data, BPL_16_17_data, BPL_17_18_data, BPL_18_19_data, BPL_19_20_data, 
              BPL_20_21_data]

    # checking if every season was uploaded correctly without NA
    for season in seasons: 
        print(season.Season.unique())
        print(season.Date[0])
        season.dropna(inplace=True)
        print(len(season)) # every season should have 380 rows
        print('|----------------------------|')

    # changing Date format
    for season in seasons: 
        season.Date = pd.to_datetime(season.Date, dayfirst=True)

    """## Adds the number of games played by the team before specific match"""

    def get_last_matches_number(playing_stat):
        
        # Diffrent lists for HomeTeam and AwayTeam because sometimes number of games played before 
        # can be diffrent becasue of postponing the match
        HT_Match_number = []
        AT_Match_number = []
        
        # dataset iteration over each row
        for i, row in playing_stat.iterrows():
            df_nad = playing_stat.iloc[0:i]
            df_specific_match = playing_stat.iloc[i]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            x = len(df_nad[(df_nad.HomeTeam == ht) | (df_nad.AwayTeam == ht)])
            y = len(df_nad[(df_nad.HomeTeam == at) | (df_nad.AwayTeam == at)])

            HT_Match_number.append(x)
            AT_Match_number.append(y)
            
        playing_stat['HT_last_matches_number'] = HT_Match_number
        playing_stat['AT_last_matches_number'] = AT_Match_number
        
        return playing_stat

    # apply function on every season data
    for season in seasons:
        season = get_last_matches_number(season)

    """## Goals scored and conceded before specific match"""

    # Gets the goals scored aggregation arranged by teams and matchweek
    def get_goals_scored(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTGS = playing_stat.iloc[i]['FTHG']
            ATGS = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGS)
            
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        GoalsScored[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,39): # we have 38 matchweeks (number of teams*2 - 2 becacue you cant play with yourself)
            GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
        return GoalsScored

    # Gets the goals conceded aggregation arranged by teams and matchweek
    def get_goals_conceded(playing_stat):
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATGC = playing_stat.iloc[i]['FTHG']
            HTGC = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGC)
        
        # Create a dataframe for goals conceded where rows are teams and cols are matchweek.
        GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        GoalsConceded[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams["Arsenal"])+1):
            GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
        return GoalsConceded

    # function summarizing and adding variables
    def get_gss(playing_stat):
        GC = get_goals_conceded(playing_stat)
        GS = get_goals_scored(playing_stat)
       
        HTGS = []
        ATGS = []
        HTGC = []
        ATGC = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTGS.append(GS.loc[ht][ht_match])
            ATGS.append(GS.loc[at][at_match])
            
            HTGC.append(GC.loc[ht][ht_match])
            ATGC.append(GC.loc[at][at_match])
            
            
        playing_stat['HTGS'] = HTGS
        playing_stat['ATGS'] = ATGS
        playing_stat['HTGC'] = HTGC
        playing_stat['ATGC'] = ATGC
        
        return playing_stat

    for season in seasons:
        season = get_gss(season)

    """## Shots on target took and allowed before specific match"""

    # Gets the shots on target took aggregation arranged by teams and matchweek
    def get_shots_on_target_took(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTSOT = playing_stat.iloc[i]['HST']
            ATSOT = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT)
            
        # Create a dataframe for shots on target took where rows are teams and cols are matchweek.
        ShotsOT = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        ShotsOT[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,39):
            ShotsOT[i] = ShotsOT[i] + ShotsOT[i-1]
        return ShotsOT

    # Gets the shots on target allowed aggregation arranged by teams and matchweek
    def get_shots_on_target_allowed(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATSOT_A = playing_stat.iloc[i]['HST']
            HTSOT_A = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT_A)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT_A)
        
        # Create a dataframe for shots on target allowed where rows are teams and cols are matchweek.
        ShotsOT_Allowed = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        ShotsOT_Allowed[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams["Arsenal"])+1):
            ShotsOT_Allowed[i] = ShotsOT_Allowed[i] + ShotsOT_Allowed[i-1]
        return ShotsOT_Allowed

    # function summarizing and adding variables
    def get_sot(playing_stat):
        SOT_T = get_shots_on_target_took(playing_stat)
        SOT_A = get_shots_on_target_allowed(playing_stat)
       
        HTSOT_T = []
        ATSOT_T = []
        HTSOT_A = []
        ATSOT_A = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTSOT_T.append(SOT_T.loc[ht][ht_match])
            ATSOT_T.append(SOT_T.loc[at][at_match])
            
            HTSOT_A.append(SOT_A.loc[ht][ht_match])
            ATSOT_A.append(SOT_A.loc[at][at_match])
            
            
        playing_stat['HTSOT_T'] = HTSOT_T
        playing_stat['ATSOT_T'] = ATSOT_T
        playing_stat['HTSOT_A'] = HTSOT_A
        playing_stat['ATSOT_A'] = ATSOT_A
        
        return playing_stat

    for season in seasons:
        season = get_sot(season)

    # Get shots accuracy
    def get_shots_acc(playing_stat):
        
        HT_Shot_Acc = playing_stat.HTGS / playing_stat.HTSOT_T # HomeTeam shots took accuracy (Goals scoring accuracy)
        AT_Shot_Acc = playing_stat.ATGS / playing_stat.ATSOT_T # AwayTeam shots took accuracy (Goals scoring accuracy)
        
        HT_Op_Shot_Acc = playing_stat.HTGC / playing_stat.HTSOT_A # HomeTeam shots allowed accuracy (Goals conceded accuracy)
        AT_Op_Shot_Acc = playing_stat.ATGC / playing_stat.ATSOT_A # AwayTeam shots allowed accuracy (Goals conceded accuracy)
        
        playing_stat['HT_Shot_Acc'] = HT_Shot_Acc
        playing_stat['AT_Shot_Acc'] = AT_Shot_Acc
        playing_stat['HT_Op_Shot_Acc'] = HT_Op_Shot_Acc
        playing_stat['AT_Op_Shot_Acc'] = AT_Op_Shot_Acc
        
        return playing_stat

    for season in seasons:
        season = get_shots_acc(season)

    """## Cumulative points before specific match"""

    # Function for getting specific number of points depends on the match result
    def get_points(result):
        if result == 'W':
            return 3
        elif result == 'D':
            return 1
        else:
            return 0
        
    # Get match results for every team in every match week
    def get_matchres(season):
        # Create a dictionary with team names as keys
        teams = {}
        for i in season.groupby('HomeTeam').mean().T.columns:
            teams[i] = []

        # the value corresponding to keys is a list containing the match result
        for i in range(len(season)):
            if season.iloc[i].FTR == 'H':
                teams[season.iloc[i].HomeTeam].append('W') 
                teams[season.iloc[i].AwayTeam].append('L') # If HomeTeam wins the AwayTeam lose so wee need to add L
            elif season.iloc[i].FTR == 'A':
                teams[season.iloc[i].AwayTeam].append('W') 
                teams[season.iloc[i].HomeTeam].append('L') # If AwayTeam wins the HomeTeam lose so wee need to add L
            else:
                teams[season.iloc[i].AwayTeam].append('D')
                teams[season.iloc[i].HomeTeam].append('D')

        return pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T

    # Get cumulative points connected with match results
    def get_cuml_points(matchres):
        matchres_points = matchres.applymap(get_points)
        for i in range(2,38):
            matchres_points[i] = matchres_points[i] + matchres_points[i-1]
            
        matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(20)])
        return matchres_points

    # function summarizing and adding variables
    def get_agg_points(season):
        matchres = get_matchres(season)
        cum_pts = get_cuml_points(matchres)
        HTP = []
        ATP = []

        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HTP.append(cum_pts.loc[ht][ht_match])
            ATP.append(cum_pts.loc[at][at_match])
                
        season['HTP'] = HTP
        season['ATP'] = ATP
        return season

    for season in seasons:
        season = get_agg_points(season)

    """## Current teams form with match results for 5 last games"""

    # get and add to dataset last 5 match results
    def add_form(playing_stat,num):
        form = get_matchres(playing_stat)
        
        for number in range(5+1): # I am interesing to 5 last games
            form[-number] = 'ND' # No data
        
        HT_form = []
        AT_form = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HT_form.append(form.loc[ht][ht_match - (num - 1)])
            AT_form.append(form.loc[at][at_match - (num - 1)])
        
        # every last match will have different variable  
        playing_stat['HM' + str(num)] = HT_form                
        playing_stat['AM' + str(num)] = AT_form
        
        return playing_stat

    # function summarizing and adding variables
    def add_form_df(playing_statistics):
        playing_statistics = add_form(playing_statistics,1)
        playing_statistics = add_form(playing_statistics,2)
        playing_statistics = add_form(playing_statistics,3)
        playing_statistics = add_form(playing_statistics,4)
        playing_statistics = add_form(playing_statistics,5)
        return playing_statistics

    for season in seasons:
        season = add_form_df(season)

    """## Cumulative number of wins, losses nad draws before specific match"""

    # Binary variables connected with match results
    def get_wins(result):
        if result == 'W':
            return 1
        elif result == 'L' or  result == 'D':
            return 0
        else:
            return np.nan 

    def get_losses(result):
        if result == 'L':
            return 1
        elif result == 'D' or  result == 'W':
            return 0
        else:
            return np.nan     

    def get_draws(result):
        if result == 'D':
            return 1
        elif result == 'L' or  result == 'W':
            return 0
        else:
            return np.nan 
        
    # Get cumulative results during specific season
    def get_agg_results(season):
        matchres = get_matchres(season)
        cum_wins = matchres.applymap(get_wins).cumsum(axis=1)
        cum_losses = matchres.applymap(get_losses).cumsum(axis=1)
        cum_draws = matchres.applymap(get_draws).cumsum(axis=1)
        
        cum_wins[0] = 0
        cum_losses[0] = 0
        cum_draws[0] = 0
        
        HT_cum_wins = []
        AT_cum_wins = []
        
        HT_cum_losses = []
        AT_cum_losses = []
        
        HT_cum_draws = []
        AT_cum_draws = []
        
        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HT_cum_wins.append(cum_wins.loc[ht][ht_match])
            AT_cum_wins.append(cum_wins.loc[at][at_match])
            
            HT_cum_losses.append(cum_losses.loc[ht][ht_match])
            AT_cum_losses.append(cum_losses.loc[at][at_match])
            
            HT_cum_draws.append(cum_draws.loc[ht][ht_match])
            AT_cum_draws.append(cum_draws.loc[at][at_match])
                
        season['HT_cum_wins'] = HT_cum_wins
        season['AT_cum_wins'] = AT_cum_wins
        
        season['HT_cum_losses'] = HT_cum_losses
        season['AT_cum_losses'] = AT_cum_losses
        
        season['HT_cum_draws'] = HT_cum_draws
        season['AT_cum_draws'] = AT_cum_draws
        
        return season

    for season in seasons:
        season = get_agg_results(season)

    """## Offensive and Defensive strength

    The entire calculation process is presented in the link below:
    * [Calcuation process](https://help.smarkets.com/hc/en-gb/articles/115001457989-How-to-calculate-Poisson-distribution-for-football-betting)
    """

    def get_off_def_strength(playing_stat):
        
        List_off_strength_HT = []
        List_def_strength_HT = []
        List_off_strength_AT = []
        List_def_strength_AT = []

        for i in range(len(playing_stat)):
            df_specific_match = playing_stat.iloc[i]
            df_day_before_SpecMat = playing_stat[playing_stat.Date < df_specific_match.Date]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            Avg_score_Home = np.mean(df_day_before_SpecMat.FTHG)
            Avg_score_Away = np.mean(df_day_before_SpecMat.FTAG)

            Avg_conced_Home = Avg_score_Away
            Avg_conced_Away = Avg_score_Home

            # Home Team offensive strength
            Avg_home_goals_scored_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTHG.mean()
            Off_strength_HT = Avg_home_goals_scored_by_HT / Avg_score_Home

            # Away Team defensive strength
            Avg_away_goals_conceded_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTHG.mean()
            Def_strength_AT = Avg_away_goals_conceded_by_AT / Avg_conced_Away

            List_off_strength_HT.append(Off_strength_HT)
            List_def_strength_AT.append(Def_strength_AT)

            # Away Team offensive strength
            Avg_away_goals_scored_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTAG.mean()
            Off_strength_AT = Avg_away_goals_scored_by_AT / Avg_score_Away

            # Home Team defensive strength
            Avg_home_goals_conceded_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTAG.mean()
            Def_strength_HT = Avg_home_goals_conceded_by_HT / Avg_conced_Home

            List_off_strength_AT.append(Off_strength_AT)
            List_def_strength_HT.append(Def_strength_HT)

        playing_stat['Off_strength_HT'] = List_off_strength_HT
        playing_stat['Def_strength_HT'] = List_def_strength_HT
        playing_stat['Off_strength_AT'] = List_off_strength_AT
        playing_stat['Def_strength_AT'] = List_def_strength_AT

    for season in seasons:
        season = get_off_def_strength(season)

    """## League position from last season

    In European football leauges every season teams (usually 3 but it depends) can get promotion to the highest league and other teams can fall out. This way we get different teams every seasons. So to get league postion from last season I decided to assign to new teams a specific postion:
    * number of teams in the highest league + new team's league postion from the second highest league
    * for example in 2020/21 season Leeds was new team and they have 1st postion in 2019/20 season in the second highest league so I assign number 21
    """

    Standings = pd.read_csv('league_tables/BPL_table_rank.csv', sep=';') # This table was created manually 
    Standings.set_index(['Team'], inplace=True)

    # Add variables with last league postion
    def get_last(playing_stat, Standings, year):
        HomeTeamLP = []
        AwayTeamLP = []
        for i in range(380):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam
            HomeTeamLP.append(Standings.loc[ht][year])
            AwayTeamLP.append(Standings.loc[at][year])
        playing_stat['HomeTeamLP'] = HomeTeamLP
        playing_stat['AwayTeamLP'] = AwayTeamLP
        return playing_stat

    BPL_11_12_data = get_last(BPL_11_12_data, Standings, '2010/11')
    BPL_12_13_data = get_last(BPL_12_13_data, Standings, '2011/12')
    BPL_13_14_data = get_last(BPL_13_14_data, Standings, '2012/13')
    BPL_14_15_data = get_last(BPL_14_15_data, Standings, '2013/14')
    BPL_15_16_data = get_last(BPL_15_16_data, Standings, '2014/15')
    BPL_16_17_data = get_last(BPL_16_17_data, Standings, '2015/16')
    BPL_17_18_data = get_last(BPL_17_18_data, Standings, '2016/17')
    BPL_18_19_data = get_last(BPL_18_19_data, Standings, '2017/18')
    BPL_19_20_data = get_last(BPL_19_20_data, Standings, '2018/19')
    BPL_20_21_data = get_last(BPL_20_21_data, Standings, '2019/20')

    # adding binary variable if specific team is a new team in specific season
    def get_newTeam(playing_stat):
        
        teams_number = len(playing_stat.HomeTeam.unique())
        
        playing_stat['Is_HT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.HomeTeamLP]
        playing_stat['Is_AT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.AwayTeamLP]
        
        return playing_stat

    # Data was only for the last 10 completed season so I cant use every historical dataset
    for season in [ BPL_11_12_data,
                BPL_12_13_data, 
                BPL_13_14_data,
                BPL_14_15_data, 
                BPL_15_16_data, 
                BPL_16_17_data, 
                BPL_17_18_data, 
                BPL_18_19_data, 
                BPL_19_20_data, 
                BPL_20_21_data]:
        season = get_newTeam(season)

    # Checking if everything is correct
    for season in [ BPL_11_12_data,
                BPL_12_13_data, 
                BPL_13_14_data,
                BPL_14_15_data, 
                BPL_15_16_data, 
                BPL_16_17_data, 
                BPL_17_18_data, 
                BPL_18_19_data, 
                BPL_19_20_data, 
                BPL_20_21_data]:
        print(season[season.HomeTeamLP > 18].HomeTeam.unique())

    """## Team data for every season from transfermarkt
    * Average age
    * Number of foreigners
    * Average market value in milions of Euro
    * Total market value in milions of Euro

    Example data from 2016/17 season:
    https://www.transfermarkt.com/premier-league/startseite/wettbewerb/GB1/plus/?saison_id=2016
    """

    BPL_budget = pd.read_csv("budget_data/BPL_transfermaket_data.csv", sep=';')
    BPL_budget.set_index(['Club'], inplace=True)
    BPL_budget

    """Changing string variables for float"""

    BPL_budget['Avg_age'].replace(',', '.', regex=True, inplace=True)
    BPL_budget['Avg_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)
    BPL_budget['Total_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)

    BPL_budget.info()

    BPL_budget.Avg_age = BPL_budget.Avg_age.astype(float)
    BPL_budget.Avg_market_value_MLN_EUR = BPL_budget.Avg_market_value_MLN_EUR.astype(float)
    BPL_budget.Total_market_value_MLN_EUR = BPL_budget.Total_market_value_MLN_EUR.astype(float)

    # function summarizing and adding variables
    def get_transfermarkt_data(season_data, Budget):
        
        HomeTeam_TM_avg_age = []
        HomeTeam_TM_foreigners_number = []
        HomeTeam_TM_avg_market_value_MLN_EUR = []
        HomeTeam_TM_total_market_value_MLN_EUR = []
        
        AwayTeam_TM_avg_age = []
        AwayTeam_TM_foreigners_number = []
        AwayTeam_TM_avg_market_value_MLN_EUR = []
        AwayTeam_TM_total_market_value_MLN_EUR = []
        
        for season in list(season_data.Season.unique()):
            
            BPL_budget_season = BPL_budget[BPL_budget.Season == season]
            
            for i in range(380):
                
                home_team = season_data[season_data.Season == season].iloc[i].HomeTeam
                away_team = season_data[season_data.Season == season].iloc[i].AwayTeam
                
                HomeTeam_TM_avg_age.append(BPL_budget_season.loc[home_team]['Avg_age'])
                HomeTeam_TM_foreigners_number.append(BPL_budget_season.loc[home_team]['Foreigners'])
                HomeTeam_TM_avg_market_value_MLN_EUR.append(BPL_budget_season.loc[home_team]['Avg_market_value_MLN_EUR'])
                HomeTeam_TM_total_market_value_MLN_EUR.append(BPL_budget_season.loc[home_team]['Total_market_value_MLN_EUR'])
                
                AwayTeam_TM_avg_age.append(BPL_budget_season.loc[away_team]['Avg_age'])
                AwayTeam_TM_foreigners_number.append(BPL_budget_season.loc[away_team]['Foreigners'])
                AwayTeam_TM_avg_market_value_MLN_EUR.append(BPL_budget_season.loc[away_team]['Avg_market_value_MLN_EUR'])
                AwayTeam_TM_total_market_value_MLN_EUR.append(BPL_budget_season.loc[away_team]['Total_market_value_MLN_EUR'])
                
                
        season_data['HomeTeam_TM_avg_age'] = HomeTeam_TM_avg_age
        season_data['HomeTeam_TM_foreigners_number'] = HomeTeam_TM_foreigners_number
        season_data['HomeTeam_TM_avg_market_value_MLN_EUR'] = HomeTeam_TM_avg_market_value_MLN_EUR
        season_data['HomeTeam_TM_total_market_value_MLN_EUR'] = HomeTeam_TM_total_market_value_MLN_EUR
        
        season_data['AwayTeam_TM_avg_age'] = AwayTeam_TM_avg_age
        season_data['AwayTeam_TM_foreigners_number'] = AwayTeam_TM_foreigners_number
        season_data['AwayTeam_TM_avg_market_value_MLN_EUR'] = AwayTeam_TM_avg_market_value_MLN_EUR
        season_data['AwayTeam_TM_total_market_value_MLN_EUR'] = AwayTeam_TM_total_market_value_MLN_EUR
        
        return season_data

    for season in [BPL_11_12_data, BPL_12_13_data, BPL_13_14_data, BPL_14_15_data, BPL_15_16_data, 
                  BPL_16_17_data, BPL_17_18_data, BPL_18_19_data, BPL_19_20_data, BPL_20_21_data]:
        
        get_transfermarkt_data(season, BPL_budget)

    """## Data concatenating"""

    BPL_data = pd.concat([BPL_00_01_data, 
                          BPL_01_02_data, 
                          BPL_02_03_data, 
                          BPL_03_04_data, 
                          BPL_04_05_data, 
                          BPL_05_06_data, 
                          BPL_06_07_data, 
                          BPL_07_08_data, 
                          BPL_08_09_data, 
                          BPL_09_10_data, 
                          BPL_10_11_data, 
                          BPL_11_12_data,
                          BPL_12_13_data, 
                          BPL_13_14_data,
                          BPL_14_15_data, 
                          BPL_15_16_data, 
                          BPL_16_17_data, 
                          BPL_17_18_data, 
                          BPL_18_19_data, 
                          BPL_19_20_data, 
                          BPL_20_21_data], ignore_index=True)

    # deleting unnecessary data
    del BPL_00_01_data, BPL_01_02_data, BPL_02_03_data, BPL_03_04_data, BPL_04_05_data, BPL_05_06_data
    del BPL_06_07_data, BPL_07_08_data, BPL_08_09_data, BPL_09_10_data
    del season, seasons

    """## Calcuating betting odds
    * 1x (Home win or Draw) = 100 / (100/Home_win_odds + 100/Draw_odds)
    * x2 (Away win or Draw) = 100 / (100/Away_win_odds + 100/Draw_odds)
    """

    def get_more_odds(playing_stat):
        x1 = []
        x2 = []
        
        for i, row in playing_stat.iterrows():
            b365_H = playing_stat.iloc[i]['B365H']
            b365_D = playing_stat.iloc[i]['B365D']
            b365_A = playing_stat.iloc[i]['B365A']
            
            odd_H = 100/b365_H
            odd_D = 100/b365_D
            odd_A = 100/b365_A
            
            b365_H_D = round((100/(odd_H + odd_D)), 2)
            b365_A_D = round((100/(odd_A + odd_D)), 2)
            
            x1.append(b365_H_D)
            x2.append(b365_A_D)
            
        
        playing_stat['B365_HD'] = x1
        playing_stat['B365_AD'] = x2
        
        return playing_stat

    get_more_odds(BPL_data)

    """## Add points collect by teams from last 3 and 5 games"""

    # Gets the form points.
    def get_form_points(string):
        sum = 0
        for letter in string:
            sum += get_points(letter) # using other function
        return sum

    # Concatenate variables with last team's results
    BPL_data['HT_3match_FormPtsStr'] = BPL_data['HM1'] + BPL_data['HM2'] + BPL_data['HM3']
    BPL_data['AT_3match_FormPtsStr'] = BPL_data['AM1'] + BPL_data['AM2'] + BPL_data['AM3']

    BPL_data['HT_5match_FormPtsStr'] = BPL_data['HM1'] + BPL_data['HM2'] + BPL_data['HM3'] + BPL_data['HM4'] + BPL_data['HM5']
    BPL_data['AT_5match_FormPtsStr'] = BPL_data['AM1'] + BPL_data['AM2'] + BPL_data['AM3'] + BPL_data['AM4'] + BPL_data['AM5']

    # Applying function to get points
    BPL_data['HT_3match_FormPts'] = BPL_data['HT_3match_FormPtsStr'].apply(get_form_points)
    BPL_data['AT_3match_FormPts'] = BPL_data['AT_3match_FormPtsStr'].apply(get_form_points)

    BPL_data['HT_5match_FormPts'] = BPL_data['HT_5match_FormPtsStr'].apply(get_form_points)
    BPL_data['AT_5match_FormPts'] = BPL_data['AT_5match_FormPtsStr'].apply(get_form_points)

    """## Win or loss streaks"""

    # Identify Win/Loss Streaks if any.
    def get_3game_ws(string):
        if string == 'WWW':
            return 1
        else:
            return 0
        
    def get_5game_ws(string):
        if string == 'WWWWW':
            return 1
        else:
            return 0
        
    def get_3game_ls(string):
        if string == 'LLL':
            return 1
        else:
            return 0
        
    def get_5game_ls(string):
        if string == 'LLLLL':
            return 1
        else:
            return 0

    BPL_data['HTWinStreak3'] = BPL_data['HT_3match_FormPtsStr'].apply(get_3game_ws)
    BPL_data['HTWinStreak5'] = BPL_data['HT_5match_FormPtsStr'].apply(get_5game_ws)
    BPL_data['HTLossStreak3'] = BPL_data['HT_3match_FormPtsStr'].apply(get_3game_ls)
    BPL_data['HTLossStreak5'] = BPL_data['HT_5match_FormPtsStr'].apply(get_5game_ls)

    BPL_data['ATWinStreak3'] = BPL_data['AT_3match_FormPtsStr'].apply(get_3game_ws)
    BPL_data['ATWinStreak5'] = BPL_data['AT_5match_FormPtsStr'].apply(get_5game_ws)
    BPL_data['ATLossStreak3'] = BPL_data['AT_3match_FormPtsStr'].apply(get_3game_ls)
    BPL_data['ATLossStreak5'] = BPL_data['AT_5match_FormPtsStr'].apply(get_5game_ls)

    """## Head to Head (H2H) variables
    * Getting last 3 head to head match results and change it for points
    """

    def h2h(playing_stat):
        
        playing_stat["H2H_Home"] = ""
        playing_stat["H2H_Away"] = ""

        for i in range(1,len(playing_stat)):
            # get teams from specifc match and all data before this match
            df = playing_stat.iloc[0:i-1,]
            ht = playing_stat.loc[i,"HomeTeam"]
            at = playing_stat.loc[i,"AwayTeam"]
            
            # getting last 3 head to head matches
            dff = df[(((df.HomeTeam == ht) &  (df.AwayTeam == at)) | ((df.HomeTeam == at) &  (df.AwayTeam == ht)))].tail(3)
            
            # using data with minimum 1 head to head match
            if (len(dff) >= 1):
                for j in range(0,len(dff)):
                    # Consider all possible scenarios
                    if ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                        
            # if there was no match beetwen the teams get 'ND'
            else: 
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] = "ND"
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] = "ND"
        return playing_stat


    # Calculating points for Home and Away teams in head to head matches and adding variables
    def get_points_h2h(playing_stat, col):
        
        playing_stat[col+"_pts"] = 0
        
        for i in range(1,len(playing_stat)):
            if (len(playing_stat.iloc[i][col]) == 1):
                if playing_stat.iloc[i][col][0] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][0] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                else:
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            elif (len(playing_stat.iloc[i][col]) == 2):
                for j in range(0,2):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
                    else:
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
            elif (len(playing_stat.iloc[i][col]) == 3):
                for j in range(0,3):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            else:
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] = 0
        
        return playing_stat

    h2h(BPL_data)

    get_points_h2h(BPL_data, "H2H_Home") 
    get_points_h2h(BPL_data, "H2H_Away")

    """## Choose last 10 completed seasons"""

    BPL_data = BPL_data[BPL_data.Season.isin(['2011/12', 
                                             '2012/13', 
                                             '2013/14', 
                                             '2014/15', 
                                             '2015/16', 
                                             '2016/17', 
                                             '2017/18', 
                                             '2018/19', 
                                             '2019/20', 
                                             '2020/21'])]

    BPL_data.reset_index(drop=True, inplace=True)
    BPL_data.shape

    """## Correct positioning of the columns and changing col names"""

    BPL_data_final = BPL_data[[
        'Div', 'Season', 'Date', 'HomeTeam', 'AwayTeam','FTHG', 'FTAG', 'FTR', 'HS','AS', 'HST', 'AST',
        'HT_last_matches_number', 'AT_last_matches_number', 'B365H', 'B365D', 'B365A', 'B365_HD', 'B365_AD',
        'HTGS', 'ATGS', 'HTGC', 'ATGC',
        'HTSOT_T','ATSOT_T', 'HTSOT_A', 'ATSOT_A', 
        'HT_Shot_Acc', 'AT_Shot_Acc','HT_Op_Shot_Acc', 'AT_Op_Shot_Acc',
        'HTP', 'ATP',
        'HT_cum_wins', 'AT_cum_wins', 'HT_cum_losses', 'AT_cum_losses', 'HT_cum_draws', 'AT_cum_draws',
        'Off_strength_HT', 'Def_strength_HT', 'Off_strength_AT', 'Def_strength_AT',
        'HM1', 'AM1', 'HM2', 'AM2', 'HM3', 'AM3', 'HM4', 'AM4', 'HM5', 'AM5',
        'HT_3match_FormPtsStr', 'AT_3match_FormPtsStr', 'HT_5match_FormPtsStr','AT_5match_FormPtsStr', 
        'HT_3match_FormPts', 'AT_3match_FormPts', 'HT_5match_FormPts', 'AT_5match_FormPts',
        'HTWinStreak3','HTWinStreak5', 'HTLossStreak3', 'HTLossStreak5', 
        'ATWinStreak3','ATWinStreak5', 'ATLossStreak3', 'ATLossStreak5',
        'HomeTeamLP', 'AwayTeamLP',
        'Is_HT_newTeam', 'Is_AT_newTeam',
        'H2H_Home', 'H2H_Away', 'H2H_Home_pts','H2H_Away_pts',
        'HomeTeam_TM_avg_age','HomeTeam_TM_foreigners_number', 
        'HomeTeam_TM_avg_market_value_MLN_EUR','HomeTeam_TM_total_market_value_MLN_EUR',
        'AwayTeam_TM_avg_age', 'AwayTeam_TM_foreigners_number', 
        'AwayTeam_TM_avg_market_value_MLN_EUR', 'AwayTeam_TM_total_market_value_MLN_EUR'
                              ]]

    BPL_data_final.rename(columns={'HM1': 'HT_last_match1', 'AM1': 'AT_last_match1', 
                                   'HM2': 'HT_last_match2', 'AM2': 'AT_last_match2', 
                                   'HM3': 'HT_last_match3', 'AM3': 'AT_last_match3', 
                                   'HM4': 'HT_last_match4', 'AM4': 'AT_last_match4', 
                                   'HM5': 'HT_last_match5', 'AM5': 'AT_last_match5', 
                                   'HomeTeamLP': 'HT_last_LP', 'AwayTeamLP': 'AT_last_LP', 
                                   #'HTGD': 'HT_goal_diff', 'ATGD': 'AT_goal_diff', 
                                   #'DiffLP': 'Diff_last_LP',
                                   'HTP': 'HT_Pts', 'ATP': 'AT_Pts',
                                   'HTSOT_T': 'HT_Shots_OT_took','ATSOT_T': 'AT_Shots_OT_took', 
                                   'HTSOT_A': 'HT_Shots_OT_allowed','ATSOT_A': 'AT_Shots_OT_allowed', 
                                   'HT_Shot_Acc': 'HT_Shots_took_Acc', 'AT_Shot_Acc': 'AT_Shots_took_Acc',
                                   'HT_Op_Shot_Acc': 'HT_Shots_allowed_Acc', 'AT_Op_Shot_Acc': 'AT_Shots_allowed_Acc'
                                  }, inplace=True)

    """## Save created dataset"""

    BPL_data_final.to_csv(path_or_buf= 'Final_data/BPL_data_final.csv', index=False)

```

```python
def Bundesliga_dataset_creator():   
    # -*- coding: utf-8 -*-
    """Bundesliga_dataset_creator.ipynb

    Automatically generated by Colaboratory.

    Original file is located at
        https://colab.research.google.com/drive/1bYpHHYg72wsRfQz7R9jagURjIqe9TsuS

    ## Wczytanie danych i podstawowe operacje (dodanie seasonu oraz zmiana formatu daty)
    """

    import numpy as np
    import pandas as pd

    league_code = 'D1'
    seasons = ["00-01", "01-02", "02-03", "03-04", "04-05", 
              "05-06", "06-07", "07-08", "08-09", "09-10", 
              "10-11", "11-12", "12-13", "13-14", "14-15", 
              "15-16", "16-17", "17-18", "18-19", "19-20", "20-21"]

    urls = []
    for season in seasons:
        x = 'https://www.football-data.co.uk/mmz4281/' + season.replace("-", "") + '/'+league_code + '.csv'
        urls.append(x)

    columns = ['Div', 'Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'B365H', 'B365D', 'B365A']

    """I want to focus on the last 10 completed seasons (from 2011/12 to 2020/21) but i need more historical data to calculate H2H (Head To Head) variables

    In some data from historical season there was no columns with shots data so I decided to also cut shots data from 00/01 to 10/11 season
    """

    Bundesliga_00_01_data = pd.read_csv(urls[0],encoding='unicode_escape')[columns[0:-7]]
    Bundesliga_01_02_data = pd.read_csv(urls[1],encoding='unicode_escape')[columns[0:-7]]
    Bundesliga_02_03_data = pd.read_csv("match_data/D1_02_03.csv", sep=';',encoding='unicode_escape')[columns[0:-7]] # Error Tokenizing Data with urls upload
    Bundesliga_03_04_data = pd.read_csv("match_data/D1_03_04.csv", sep=';')[columns[0:-7]] # Error Tokenizing Data with urls upload
    Bundesliga_04_05_data = pd.read_csv(urls[4])[columns[0:-7]]
    Bundesliga_05_06_data = pd.read_csv(urls[5])[columns[0:-7]]
    Bundesliga_06_07_data = pd.read_csv(urls[6])[columns[0:-7]]
    Bundesliga_07_08_data = pd.read_csv(urls[7])[columns[0:-7]]
    Bundesliga_08_09_data = pd.read_csv(urls[8])[columns[0:-7]]
    Bundesliga_09_10_data = pd.read_csv(urls[9])[columns[0:-7]]
    Bundesliga_10_11_data = pd.read_csv(urls[10])[columns[0:-7]]
    Bundesliga_11_12_data = pd.read_csv(urls[11])[columns]
    Bundesliga_12_13_data = pd.read_csv(urls[12])[columns]
    Bundesliga_13_14_data = pd.read_csv(urls[13])[columns]
    Bundesliga_14_15_data = pd.read_csv(urls[14])[columns]
    Bundesliga_15_16_data = pd.read_csv(urls[15])[columns]
    Bundesliga_16_17_data = pd.read_csv(urls[16])[columns]
    Bundesliga_17_18_data = pd.read_csv(urls[17])[columns]
    Bundesliga_18_19_data = pd.read_csv(urls[18])[columns]
    Bundesliga_19_20_data = pd.read_csv(urls[19])[columns]
    Bundesliga_20_21_data = pd.read_csv(urls[20])[columns]

    Bundesliga_20_21_data['Season'] = '2020/21'
    Bundesliga_19_20_data['Season'] = '2019/20'
    Bundesliga_18_19_data['Season'] = '2018/19'
    Bundesliga_17_18_data['Season'] = '2017/18'
    Bundesliga_16_17_data['Season'] = '2016/17'
    Bundesliga_15_16_data['Season'] = '2015/16'
    Bundesliga_14_15_data['Season'] = '2014/15'
    Bundesliga_13_14_data['Season'] = '2013/14'
    Bundesliga_12_13_data['Season'] = '2012/13'
    Bundesliga_11_12_data['Season'] = '2011/12'
    Bundesliga_10_11_data['Season'] = '2010/11'
    Bundesliga_09_10_data['Season'] = '2009/10'
    Bundesliga_08_09_data['Season'] = '2008/09'
    Bundesliga_07_08_data['Season'] = '2007/08'
    Bundesliga_06_07_data['Season'] = '2006/07'
    Bundesliga_05_06_data['Season'] = '2005/06'
    Bundesliga_04_05_data['Season'] = '2004/05'
    Bundesliga_03_04_data['Season'] = '2003/04'
    Bundesliga_02_03_data['Season'] = '2002/03'
    Bundesliga_01_02_data['Season'] = '2001/02'
    Bundesliga_00_01_data['Season'] = '2000/01'

    seasons = [Bundesliga_00_01_data, Bundesliga_01_02_data, Bundesliga_02_03_data, Bundesliga_03_04_data, Bundesliga_04_05_data,
              Bundesliga_05_06_data, Bundesliga_06_07_data, Bundesliga_07_08_data, Bundesliga_08_09_data, Bundesliga_09_10_data, 
              Bundesliga_10_11_data, Bundesliga_11_12_data, Bundesliga_12_13_data, Bundesliga_13_14_data, Bundesliga_14_15_data, 
              Bundesliga_15_16_data, Bundesliga_16_17_data, Bundesliga_17_18_data, Bundesliga_18_19_data, Bundesliga_19_20_data, 
              Bundesliga_20_21_data]

    # checking if every season was uploaded correctly without NA
    for season in seasons:
        print(season.Season.unique())
        print(season.Date[0])
        season.dropna(inplace=True)
        print(len(season)) # every season should have 306 rows (18 teams in Bundesliga)
        print('|----------------------------|')

    # changing Date format
    for season in seasons:
        season.Date = pd.to_datetime(season.Date, dayfirst=True)

    """## Adds the number of games played by the team before specific match"""

    def get_last_matches_number(playing_stat):
        
        # Diffrent lists for HomeTeam and AwayTeam because sometimes number of games played before 
        # can be diffrent becasue of postponing the match
        HT_Match_number = []
        AT_Match_number = []
        
        # dataset iteration over each row
        for i, row in playing_stat.iterrows():
            df_nad = playing_stat.iloc[0:i]
            df_specific_match= playing_stat.iloc[i]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            x = len(df_nad[(df_nad.HomeTeam == ht) | (df_nad.AwayTeam == ht)])
            y = len(df_nad[(df_nad.HomeTeam == at) | (df_nad.AwayTeam == at)])

            HT_Match_number.append(x)
            AT_Match_number.append(y)
            
        playing_stat['HT_last_matches_number'] = HT_Match_number
        playing_stat['AT_last_matches_number'] = AT_Match_number
        
        return playing_stat

    # apply function on every season data
    for season in seasons:
        season = get_last_matches_number(season)

    """## Goals scored and conceded before specific match"""

    # Gets the goals scored aggregation arranged by teams and matchweek
    def get_goals_scored(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTGS = playing_stat.iloc[i]['FTHG']
            ATGS = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGS)
            
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        GoalsScored[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1): # we 34 matchweeks (number of teams*2 - 2 becacue you cant play with yourself)
            GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
        return GoalsScored


    # Gets the goals conceded aggregation arranged by teams and matchweek
    def get_goals_conceded(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATGC = playing_stat.iloc[i]['FTHG']
            HTGC = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGC)
        
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        GoalsConceded[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
        return GoalsConceded

    # function summarizing and adding variables
    def get_gss(playing_stat):
        GC = get_goals_conceded(playing_stat)
        GS = get_goals_scored(playing_stat)
       
        HTGS = []
        ATGS = []
        HTGC = []
        ATGC = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTGS.append(GS.loc[ht][ht_match])
            ATGS.append(GS.loc[at][at_match])
            
            HTGC.append(GC.loc[ht][ht_match])
            ATGC.append(GC.loc[at][at_match])
            
            
        playing_stat['HTGS'] = HTGS
        playing_stat['ATGS'] = ATGS
        playing_stat['HTGC'] = HTGC
        playing_stat['ATGC'] = ATGC
        
        return playing_stat

    for season in seasons:
        season = get_gss(season)

    """## Shots on target took and allowed before specific match"""

    # Gets the shots on target took aggregation arranged by teams and matchweek
    def get_shots_on_target_took(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTSOT = playing_stat.iloc[i]['HST']
            ATSOT = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT)
            
        # Create a dataframe for shots on target took where rows are teams and cols are matchweek.
        ShotsOT = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        ShotsOT[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            ShotsOT[i] = ShotsOT[i] + ShotsOT[i-1]
        return ShotsOT


    # Gets the shots on target allowed aggregation arranged by teams and matchweek
    def get_shots_on_target_allowed(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
    #         print("check {} \n".format(i))
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATSOT_A = playing_stat.iloc[i]['HST']
            HTSOT_A = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT_A)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT_A)
        
        # Create a dataframe for shots on target allowed where rows are teams and cols are matchweek.
        ShotsOT_Allowed = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        ShotsOT_Allowed[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            ShotsOT_Allowed[i] = ShotsOT_Allowed[i] + ShotsOT_Allowed[i-1]
        return ShotsOT_Allowed

    # function summarizing and adding variables
    def get_sot(playing_stat):
        SOT_T = get_shots_on_target_took(playing_stat)
        SOT_A = get_shots_on_target_allowed(playing_stat)
       
        HTSOT_T = []
        ATSOT_T = []
        HTSOT_A = []
        ATSOT_A = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTSOT_T.append(SOT_T.loc[ht][ht_match])
            ATSOT_T.append(SOT_T.loc[at][at_match])
            
            HTSOT_A.append(SOT_A.loc[ht][ht_match])
            ATSOT_A.append(SOT_A.loc[at][at_match])
            
            
        playing_stat['HTSOT_T'] = HTSOT_T
        playing_stat['ATSOT_T'] = ATSOT_T
        playing_stat['HTSOT_A'] = HTSOT_A
        playing_stat['ATSOT_A'] = ATSOT_A
        
        return playing_stat

    # Datasets with shots data
    for season in [ Bundesliga_11_12_data,
                Bundesliga_12_13_data, 
                Bundesliga_13_14_data,
                Bundesliga_14_15_data, 
                Bundesliga_15_16_data, 
                Bundesliga_16_17_data, 
                Bundesliga_17_18_data, 
                Bundesliga_18_19_data, 
                Bundesliga_19_20_data, 
                Bundesliga_20_21_data]:
        season = get_sot(season)

    # Get shots accuracy
    def get_shots_acc(playing_stat):
        
        HT_Shot_Acc = playing_stat.HTGS / playing_stat.HTSOT_T # HomeTeam shots took accuracy (Goals scoring accuracy)
        AT_Shot_Acc = playing_stat.ATGS / playing_stat.ATSOT_T # AwayTeam shots took accuracy (Goals scoring accuracy)
        
        HT_Op_Shot_Acc = playing_stat.HTGC / playing_stat.HTSOT_A # HomeTeam shots allowed accuracy (Goals conceded accuracy)
        AT_Op_Shot_Acc = playing_stat.ATGC / playing_stat.ATSOT_A # AwayTeam shots allowed accuracy (Goals conceded accuracy)
        
        playing_stat['HT_Shot_Acc'] = HT_Shot_Acc
        playing_stat['AT_Shot_Acc'] = AT_Shot_Acc
        playing_stat['HT_Op_Shot_Acc'] = HT_Op_Shot_Acc
        playing_stat['AT_Op_Shot_Acc'] = AT_Op_Shot_Acc
        
        return playing_stat

    # Datasets with shots data
    for season in [ Bundesliga_11_12_data,
                Bundesliga_12_13_data, 
                Bundesliga_13_14_data,
                Bundesliga_14_15_data, 
                Bundesliga_15_16_data, 
                Bundesliga_16_17_data, 
                Bundesliga_17_18_data, 
                Bundesliga_18_19_data, 
                Bundesliga_19_20_data, 
                Bundesliga_20_21_data]:
        season = get_shots_acc(season)

    """## Cumulative points before specific match"""

    # Function for getting specific number of points depends on the match result
    def get_points(result):
        if result == 'W':
            return 3
        elif result == 'D':
            return 1
        else:
            return 0

    # Get match results for every team in every match week
    def get_matchres(season):
        # Create a dictionary with team names as keys
        teams = {}
        for i in season.groupby('HomeTeam').mean().T.columns:
            teams[i] = []

        # the value corresponding to keys is a list containing the match result
        for i in range(len(season)):
            if season.iloc[i].FTR == 'H':
                teams[season.iloc[i].HomeTeam].append('W')
                teams[season.iloc[i].AwayTeam].append('L') # If HomeTeam wins the AwayTeam lose so wee need to add L
            elif season.iloc[i].FTR == 'A':
                teams[season.iloc[i].AwayTeam].append('W')
                teams[season.iloc[i].HomeTeam].append('L') # If AwayTeam wins the HomeTeam lose so wee need to add L
            else:
                teams[season.iloc[i].AwayTeam].append('D')
                teams[season.iloc[i].HomeTeam].append('D')

                
        return pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T

    # Get cumulative points connected with match results
    def get_cuml_points(matchres):
        
        matchres_points = matchres.applymap(get_points)
        for i in range(2,len(matchres)*2-2):
            matchres_points[i] = matchres_points[i] + matchres_points[i-1]
            
        matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(len(matchres))])
        return matchres_points

    # function summarizing and adding variables
    def get_agg_points(season):
        matchres = get_matchres(season)
        cum_pts = get_cuml_points(matchres)
        HTP = []
        ATP = []

        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HTP.append(cum_pts.loc[ht][ht_match])
            ATP.append(cum_pts.loc[at][at_match])
                
        season['HTP'] = HTP
        season['ATP'] = ATP
        return season

    for season in seasons:
        season = get_agg_points(season)

    """## Current teams form with match results for 5 last games"""

    # get and add to dataset last 5 match results
    def add_form(playing_stat,num):
        form = get_matchres(playing_stat)
        
        for number in range(5+1): # I am interesing to 5 last games
            form[-number] = 'ND'  # No data
        
        HT_form = []
        AT_form = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HT_form.append(form.loc[ht][ht_match - (num - 1)])
            AT_form.append(form.loc[at][at_match - (num - 1)])
            
        # every last match will have different variable    
        playing_stat['HM' + str(num)] = HT_form                 
        playing_stat['AM' + str(num)] = AT_form
        
        return playing_stat

    # function summarizing and adding variables
    def add_form_df(playing_statistics):
        playing_statistics = add_form(playing_statistics,1)
        playing_statistics = add_form(playing_statistics,2)
        playing_statistics = add_form(playing_statistics,3)
        playing_statistics = add_form(playing_statistics,4)
        playing_statistics = add_form(playing_statistics,5)
        return playing_statistics

    for season in seasons:
        season = add_form_df(season)

    """## Cumulative number of wins, losses nad draws before specific match"""

    # Binary variables connected with match results
    def get_losses(result):
        if result == 'L':
            return 1
        elif result == 'D' or  result == 'W':
            return 0
        else:
            return np.nan     

    def get_wins(result):
        if result == 'W':
            return 1
        elif result == 'L' or  result == 'D':
            return 0
        else:
            return np.nan 

    def get_draws(result):
        if result == 'D':
            return 1
        elif result == 'L' or  result == 'W':
            return 0
        else:
            return np.nan 
        
    # Get cumulative results during specific season
    def get_agg_results(season):
        matchres = get_matchres(season)
        cum_wins = matchres.applymap(get_wins).cumsum(axis=1)
        cum_losses = matchres.applymap(get_losses).cumsum(axis=1)
        cum_draws = matchres.applymap(get_draws).cumsum(axis=1)
        
        cum_wins[0] = 0
        cum_losses[0] = 0
        cum_draws[0] = 0
        
        HT_cum_wins = []
        AT_cum_wins = []
        
        HT_cum_losses = []
        AT_cum_losses = []
        
        HT_cum_draws = []
        AT_cum_draws = []
        
        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HT_cum_wins.append(cum_wins.loc[ht][ht_match])
            AT_cum_wins.append(cum_wins.loc[at][at_match])
            
            HT_cum_losses.append(cum_losses.loc[ht][ht_match])
            AT_cum_losses.append(cum_losses.loc[at][at_match])
            
            HT_cum_draws.append(cum_draws.loc[ht][ht_match])
            AT_cum_draws.append(cum_draws.loc[at][at_match])
                
        season['HT_cum_wins'] = HT_cum_wins
        season['AT_cum_wins'] = AT_cum_wins
        
        season['HT_cum_losses'] = HT_cum_losses
        season['AT_cum_losses'] = AT_cum_losses
        
        season['HT_cum_draws'] = HT_cum_draws
        season['AT_cum_draws'] = AT_cum_draws
        
        return season

    for season in seasons:
        season = get_agg_results(season)

    """## Offensive and Defensive strength

    The entire calculation process is presented in the link below:
    * [Calcuation process](https://help.smarkets.com/hc/en-gb/articles/115001457989-How-to-calculate-Poisson-distribution-for-football-betting)
    """

    def get_off_def_strength(playing_stat):
        
        List_off_strength_HT = []
        List_def_strength_HT = []
        List_off_strength_AT = []
        List_def_strength_AT = []

        for i in range(len(playing_stat)):
            df_specific_match= playing_stat.iloc[i]
            df_day_before_SpecMat= playing_stat[playing_stat.Date < df_specific_match.Date]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            Avg_score_Home = np.mean(df_day_before_SpecMat.FTHG) if np.mean(df_day_before_SpecMat.FTHG) !=0 else np.nan
            Avg_score_Away = np.mean(df_day_before_SpecMat.FTAG) if np.mean(df_day_before_SpecMat.FTAG) !=0 else np.nan

            Avg_conced_Home = Avg_score_Away
            Avg_conced_Away = Avg_score_Home

            # Home Team offensive strength
            Avg_home_goals_scored_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTHG.mean()
            Off_strength_HT = Avg_home_goals_scored_by_HT / Avg_score_Home

            # Away Team defensive strength
            Avg_away_goals_conceded_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTHG.mean()
            Def_strength_AT = Avg_away_goals_conceded_by_AT / Avg_conced_Away

            List_off_strength_HT.append(Off_strength_HT)
            List_def_strength_AT.append(Def_strength_AT)

            # Away Team offensive strength
            Avg_away_goals_scored_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTAG.mean()
            Off_strength_AT = Avg_away_goals_scored_by_AT / Avg_score_Away

            # Home Team defensive strength
            Avg_home_goals_conceded_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTAG.mean()
            Def_strength_HT = Avg_home_goals_conceded_by_HT / Avg_conced_Home

            List_off_strength_AT.append(Off_strength_AT)
            List_def_strength_HT.append(Def_strength_HT)

        playing_stat['Off_strength_HT'] = List_off_strength_HT
        playing_stat['Def_strength_HT'] = List_def_strength_HT
        playing_stat['Off_strength_AT'] = List_off_strength_AT
        playing_stat['Def_strength_AT'] = List_def_strength_AT

    for season in seasons:
        season = get_off_def_strength(season)

    """## League position from last season

    In European football leauges every season teams (usually 3 but it depends) can get promotion to the highest league and other teams can fall out. This way we get different teams every seasons. So to get league postion from last season I decided to assign to new teams a specific postion:
    * number of teams in the highest league + new team's league postion from the second highest league
    * for example in 2020/21 season Bielefeld was new team and they have 1st postion in 2019/20 season in the second highest league so I assign number 19
    """

    Standings = pd.read_csv('league_tables/Bundesliga_table_rank.csv', sep=';') # This table was created manually 
    Standings.set_index(['Team'], inplace=True)

    # Add variables with last league postion
    def get_last(playing_stat, Standings, year):
        HomeTeamLP = []
        AwayTeamLP = []
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam
            HomeTeamLP.append(Standings.loc[ht][year])
            AwayTeamLP.append(Standings.loc[at][year])
        playing_stat['HomeTeamLP'] = HomeTeamLP
        playing_stat['AwayTeamLP'] = AwayTeamLP
        return playing_stat

    Bundesliga_11_12_data = get_last(Bundesliga_11_12_data, Standings, '2010/11')
    Bundesliga_12_13_data = get_last(Bundesliga_12_13_data, Standings, '2011/12')
    Bundesliga_13_14_data = get_last(Bundesliga_13_14_data, Standings, '2012/13')
    Bundesliga_14_15_data = get_last(Bundesliga_14_15_data, Standings, '2013/14')
    Bundesliga_15_16_data = get_last(Bundesliga_15_16_data, Standings, '2014/15')
    Bundesliga_16_17_data = get_last(Bundesliga_16_17_data, Standings, '2015/16')
    Bundesliga_17_18_data = get_last(Bundesliga_17_18_data, Standings, '2016/17')
    Bundesliga_18_19_data = get_last(Bundesliga_18_19_data, Standings, '2017/18')
    Bundesliga_19_20_data = get_last(Bundesliga_19_20_data, Standings, '2018/19')
    Bundesliga_20_21_data = get_last(Bundesliga_20_21_data, Standings, '2019/20')

    # adding binary variable if specific team is a new team in specific season
    def get_newTeam(playing_stat):
        
        teams_number = len(playing_stat.HomeTeam.unique())
        
        playing_stat['Is_HT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.HomeTeamLP]
        playing_stat['Is_AT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.AwayTeamLP]
        
        return playing_stat

    # Data was only for the last 10 completed season so I cant use every historical dataset
    for season in [ Bundesliga_11_12_data,
                Bundesliga_12_13_data, 
                Bundesliga_13_14_data,
                Bundesliga_14_15_data, 
                Bundesliga_15_16_data, 
                Bundesliga_16_17_data, 
                Bundesliga_17_18_data, 
                Bundesliga_18_19_data, 
                Bundesliga_19_20_data, 
                Bundesliga_20_21_data]:
        season = get_newTeam(season)

    # Checking if everything is correct
    for season in [ Bundesliga_11_12_data,
                Bundesliga_12_13_data, 
                Bundesliga_13_14_data,
                Bundesliga_14_15_data, 
                Bundesliga_15_16_data, 
                Bundesliga_16_17_data, 
                Bundesliga_17_18_data, 
                Bundesliga_18_19_data, 
                Bundesliga_19_20_data, 
                Bundesliga_20_21_data]:
        print(season[season.HomeTeamLP > 18].HomeTeam.unique())

    """## Team data for every season from transfermarkt
    * Average age
    * Number of foreigners
    * Average market value in milions of Euro
    * Total market value in milions of Euro

    Example data from 2016/17 season:
    https://www.transfermarkt.com/bundesliga/startseite/wettbewerb/L1/plus/?saison_id=2016
    """

    Bundesliga_budget = pd.read_csv("budget_data/Bundesliga_transfermaket_data.csv", sep=';')
    Bundesliga_budget.set_index(['Club'], inplace=True)
    Bundesliga_budget

    """Changing string variables for float"""

    Bundesliga_budget['Avg_age'].replace(',', '.', regex=True, inplace=True)
    Bundesliga_budget['Avg_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)
    Bundesliga_budget['Total_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)

    Bundesliga_budget.info()

    Bundesliga_budget.Avg_age = Bundesliga_budget.Avg_age.astype(float)
    Bundesliga_budget.Avg_market_value_MLN_EUR = Bundesliga_budget.Avg_market_value_MLN_EUR.astype(float)
    Bundesliga_budget.Total_market_value_MLN_EUR = Bundesliga_budget.Total_market_value_MLN_EUR.astype(float)

    # function summarizing and adding variables
    def get_transfermarkt_data(season_data, Budget):
        
        HomeTeam_TM_avg_age = []
        HomeTeam_TM_foreigners_number = []
        HomeTeam_TM_avg_market_value_MLN_EUR = []
        HomeTeam_TM_total_market_value_MLN_EUR = []
        
        AwayTeam_TM_avg_age = []
        AwayTeam_TM_foreigners_number = []
        AwayTeam_TM_avg_market_value_MLN_EUR = []
        AwayTeam_TM_total_market_value_MLN_EUR = []
        
        for season in list(season_data.Season.unique()):
            
            Bundesliga_budget_season = Budget[Budget.Season == season]
            
            for i in range(len(season_data)):
                
                home_team = season_data[season_data.Season == season].iloc[i].HomeTeam
                away_team = season_data[season_data.Season == season].iloc[i].AwayTeam
                
                HomeTeam_TM_avg_age.append(Bundesliga_budget_season.loc[home_team]['Avg_age'])
                HomeTeam_TM_foreigners_number.append(Bundesliga_budget_season.loc[home_team]['Foreigners'])
                HomeTeam_TM_avg_market_value_MLN_EUR.append(Bundesliga_budget_season.loc[home_team]['Avg_market_value_MLN_EUR'])
                HomeTeam_TM_total_market_value_MLN_EUR.append(Bundesliga_budget_season.loc[home_team]['Total_market_value_MLN_EUR'])
                
                AwayTeam_TM_avg_age.append(Bundesliga_budget_season.loc[away_team]['Avg_age'])
                AwayTeam_TM_foreigners_number.append(Bundesliga_budget_season.loc[away_team]['Foreigners'])
                AwayTeam_TM_avg_market_value_MLN_EUR.append(Bundesliga_budget_season.loc[away_team]['Avg_market_value_MLN_EUR'])
                AwayTeam_TM_total_market_value_MLN_EUR.append(Bundesliga_budget_season.loc[away_team]['Total_market_value_MLN_EUR'])
                
                
        season_data['HomeTeam_TM_avg_age'] = HomeTeam_TM_avg_age
        season_data['HomeTeam_TM_foreigners_number'] = HomeTeam_TM_foreigners_number
        season_data['HomeTeam_TM_avg_market_value_MLN_EUR'] = HomeTeam_TM_avg_market_value_MLN_EUR
        season_data['HomeTeam_TM_total_market_value_MLN_EUR'] = HomeTeam_TM_total_market_value_MLN_EUR
        
        season_data['AwayTeam_TM_avg_age'] = AwayTeam_TM_avg_age
        season_data['AwayTeam_TM_foreigners_number'] = AwayTeam_TM_foreigners_number
        season_data['AwayTeam_TM_avg_market_value_MLN_EUR'] = AwayTeam_TM_avg_market_value_MLN_EUR
        season_data['AwayTeam_TM_total_market_value_MLN_EUR'] = AwayTeam_TM_total_market_value_MLN_EUR
        
        return season_data

    for season in [Bundesliga_11_12_data, Bundesliga_12_13_data, Bundesliga_13_14_data, Bundesliga_14_15_data, Bundesliga_15_16_data, 
                  Bundesliga_16_17_data, Bundesliga_17_18_data, Bundesliga_18_19_data, Bundesliga_19_20_data, Bundesliga_20_21_data]:
        
        get_transfermarkt_data(season, Bundesliga_budget)

    """## Data concatenating"""

    Bundesliga_data = pd.concat([Bundesliga_00_01_data, 
                          Bundesliga_01_02_data, 
                          Bundesliga_02_03_data, 
                          Bundesliga_03_04_data, 
                          Bundesliga_04_05_data, 
                          Bundesliga_05_06_data, 
                          Bundesliga_06_07_data, 
                          Bundesliga_07_08_data, 
                          Bundesliga_08_09_data, 
                          Bundesliga_09_10_data, 
                          Bundesliga_10_11_data, 
                          Bundesliga_11_12_data,
                          Bundesliga_12_13_data, 
                          Bundesliga_13_14_data,
                          Bundesliga_14_15_data, 
                          Bundesliga_15_16_data, 
                          Bundesliga_16_17_data, 
                          Bundesliga_17_18_data, 
                          Bundesliga_18_19_data, 
                          Bundesliga_19_20_data, 
                          Bundesliga_20_21_data], ignore_index=True)

    # deleting unnecessary data
    del Bundesliga_00_01_data, Bundesliga_01_02_data, Bundesliga_02_03_data, Bundesliga_03_04_data, Bundesliga_04_05_data, Bundesliga_05_06_data
    del Bundesliga_06_07_data, Bundesliga_07_08_data, Bundesliga_08_09_data, Bundesliga_09_10_data
    del season, seasons

    """## Calcuating betting odds
    * 1x (Home win or Draw) = 100 / (100/Home_win_odds + 100/Draw_odds)
    * x2 (Away win or Draw) = 100 / (100/Away_win_odds + 100/Draw_odds)
    """

    def get_more_odds(playing_stat):
        x1 = []
        x2 = []
        
        for i, row in playing_stat.iterrows():
            b365_H = playing_stat.iloc[i]['B365H']
            b365_D = playing_stat.iloc[i]['B365D']
            b365_A = playing_stat.iloc[i]['B365A']
            
            odd_H = 100/b365_H
            odd_D = 100/b365_D
            odd_A = 100/b365_A
            
            b365_H_D = round((100/(odd_H + odd_D)), 2)
            b365_A_D = round((100/(odd_A + odd_D)), 2)
            
            x1.append(b365_H_D)
            x2.append(b365_A_D)
            
        
        playing_stat['B365_HD'] = x1
        playing_stat['B365_AD'] = x2
        
        return playing_stat

    get_more_odds(Bundesliga_data)

    """## Add points collect by teams from last 3 and 5 games"""

    # Gets the form points.
    def get_form_points(string):
        sum = 0
        for letter in string:
            sum += get_points(letter) # using other function
        return sum

    # Concatenate variables with last team's results
    Bundesliga_data['HT_3match_FormPtsStr'] = Bundesliga_data['HM1'] + Bundesliga_data['HM2'] + Bundesliga_data['HM3']
    Bundesliga_data['AT_3match_FormPtsStr'] = Bundesliga_data['AM1'] + Bundesliga_data['AM2'] + Bundesliga_data['AM3']

    Bundesliga_data['HT_5match_FormPtsStr'] = Bundesliga_data['HM1'] + Bundesliga_data['HM2'] + Bundesliga_data['HM3'] + Bundesliga_data['HM4'] + Bundesliga_data['HM5']
    Bundesliga_data['AT_5match_FormPtsStr'] = Bundesliga_data['AM1'] + Bundesliga_data['AM2'] + Bundesliga_data['AM3'] + Bundesliga_data['AM4'] + Bundesliga_data['AM5']

    # Applying function to get points
    Bundesliga_data['HT_3match_FormPts'] = Bundesliga_data['HT_3match_FormPtsStr'].apply(get_form_points)
    Bundesliga_data['AT_3match_FormPts'] = Bundesliga_data['AT_3match_FormPtsStr'].apply(get_form_points)

    Bundesliga_data['HT_5match_FormPts'] = Bundesliga_data['HT_5match_FormPtsStr'].apply(get_form_points)
    Bundesliga_data['AT_5match_FormPts'] = Bundesliga_data['AT_5match_FormPtsStr'].apply(get_form_points)

    """## Win or loss streaks"""

    # Identify Win/Loss Streaks if any.
    def get_3game_ws(string):
        if string == 'WWW':
            return 1
        else:
            return 0
        
    def get_5game_ws(string):
        if string == 'WWWWW':
            return 1
        else:
            return 0
        
    def get_3game_ls(string):
        if string == 'LLL':
            return 1
        else:
            return 0
        
    def get_5game_ls(string):
        if string == 'LLLLL':
            return 1
        else:
            return 0

    Bundesliga_data['HTWinStreak3'] = Bundesliga_data['HT_3match_FormPtsStr'].apply(get_3game_ws)
    Bundesliga_data['HTWinStreak5'] = Bundesliga_data['HT_5match_FormPtsStr'].apply(get_5game_ws)
    Bundesliga_data['HTLossStreak3'] = Bundesliga_data['HT_3match_FormPtsStr'].apply(get_3game_ls)
    Bundesliga_data['HTLossStreak5'] = Bundesliga_data['HT_5match_FormPtsStr'].apply(get_5game_ls)

    Bundesliga_data['ATWinStreak3'] = Bundesliga_data['AT_3match_FormPtsStr'].apply(get_3game_ws)
    Bundesliga_data['ATWinStreak5'] = Bundesliga_data['AT_5match_FormPtsStr'].apply(get_5game_ws)
    Bundesliga_data['ATLossStreak3'] = Bundesliga_data['AT_3match_FormPtsStr'].apply(get_3game_ls)
    Bundesliga_data['ATLossStreak5'] = Bundesliga_data['AT_5match_FormPtsStr'].apply(get_5game_ls)

    """## Head to Head (H2H) variables
    * Getting last 3 head to head match results and change it for points
    """

    def h2h(playing_stat):
        
        playing_stat["H2H_Home"] = ""
        playing_stat["H2H_Away"] = ""

        for i in range(1,len(playing_stat)):
            # get teams from specifc match and all data before this match
            df = playing_stat.iloc[0:i-1,]
            ht = playing_stat.loc[i,"HomeTeam"]
            at = playing_stat.loc[i,"AwayTeam"]
            
            # getting last 3 head to head matches
            dff = df[(((df.HomeTeam == ht) &  (df.AwayTeam == at)) | ((df.HomeTeam == at) &  (df.AwayTeam == ht)))].tail(3)
            
            # using data with minimum 1 head to head match
            if (len(dff) >= 1):
                for j in range(0,len(dff)):
                    # Consider all possible scenarios
                    if ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                        
            # if there was no match beetwen the teams get 'ND'
            else: 
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] = "ND"
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] = "ND"
        return playing_stat


    # Calculating points for Home and Away teams in head to head matches and adding variables
    def get_points_h2h(playing_stat, col):
        
        playing_stat[col+"_pts"] = 0
        
        for i in range(1,len(playing_stat)):
            if (len(playing_stat.iloc[i][col]) == 1):
                if playing_stat.iloc[i][col][0] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][0] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                else:
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            elif (len(playing_stat.iloc[i][col]) == 2):
                for j in range(0,2):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
                    else:
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
            elif (len(playing_stat.iloc[i][col]) == 3):
                for j in range(0,3):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            else:
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] = 0
        
        return playing_stat

    h2h(Bundesliga_data)

    get_points_h2h(Bundesliga_data, "H2H_Home") 
    get_points_h2h(Bundesliga_data, "H2H_Away")

    """## Choose last 10 completed seasons"""

    Bundesliga_data = Bundesliga_data[Bundesliga_data.Season.isin(['2011/12', 
                                             '2012/13', 
                                             '2013/14', 
                                             '2014/15', 
                                             '2015/16', 
                                             '2016/17', 
                                             '2017/18', 
                                             '2018/19', 
                                             '2019/20', 
                                             '2020/21'])]

    Bundesliga_data.reset_index(drop=True, inplace=True)
    Bundesliga_data.shape

    """## Correct positioning of the columns and changing col names"""

    Bundesliga_data_final = Bundesliga_data[[
        'Div', 'Season', 'Date', 'HomeTeam', 'AwayTeam','FTHG', 'FTAG', 'FTR', 'HS','AS', 'HST', 'AST',
        'HT_last_matches_number', 'AT_last_matches_number', 'B365H', 'B365D', 'B365A', 'B365_HD', 'B365_AD',
        'HTGS', 'ATGS', 'HTGC', 'ATGC',
        'HTSOT_T','ATSOT_T', 'HTSOT_A', 'ATSOT_A', 
        'HT_Shot_Acc', 'AT_Shot_Acc','HT_Op_Shot_Acc', 'AT_Op_Shot_Acc',
        'HTP', 'ATP',
        'HT_cum_wins', 'AT_cum_wins', 'HT_cum_losses', 'AT_cum_losses', 'HT_cum_draws', 'AT_cum_draws',
        'Off_strength_HT', 'Def_strength_HT', 'Off_strength_AT', 'Def_strength_AT',
        'HM1', 'AM1', 'HM2', 'AM2', 'HM3', 'AM3', 'HM4', 'AM4', 'HM5', 'AM5',
        'HT_3match_FormPtsStr', 'AT_3match_FormPtsStr', 'HT_5match_FormPtsStr','AT_5match_FormPtsStr', 
        'HT_3match_FormPts', 'AT_3match_FormPts', 'HT_5match_FormPts', 'AT_5match_FormPts',
        'HTWinStreak3','HTWinStreak5', 'HTLossStreak3', 'HTLossStreak5', 
        'ATWinStreak3','ATWinStreak5', 'ATLossStreak3', 'ATLossStreak5',
        'HomeTeamLP', 'AwayTeamLP',
        'Is_HT_newTeam', 'Is_AT_newTeam',
        'H2H_Home', 'H2H_Away', 'H2H_Home_pts','H2H_Away_pts',
        'HomeTeam_TM_avg_age','HomeTeam_TM_foreigners_number', 
        'HomeTeam_TM_avg_market_value_MLN_EUR','HomeTeam_TM_total_market_value_MLN_EUR',
        'AwayTeam_TM_avg_age', 'AwayTeam_TM_foreigners_number', 
        'AwayTeam_TM_avg_market_value_MLN_EUR', 'AwayTeam_TM_total_market_value_MLN_EUR'
                              ]]

    Bundesliga_data_final.rename(columns={'HM1': 'HT_last_match1', 'AM1': 'AT_last_match1', 
                                   'HM2': 'HT_last_match2', 'AM2': 'AT_last_match2', 
                                   'HM3': 'HT_last_match3', 'AM3': 'AT_last_match3', 
                                   'HM4': 'HT_last_match4', 'AM4': 'AT_last_match4', 
                                   'HM5': 'HT_last_match5', 'AM5': 'AT_last_match5', 
                                   'HomeTeamLP': 'HT_last_LP', 'AwayTeamLP': 'AT_last_LP', 
                                   #'HTGD': 'HT_goal_diff', 'ATGD': 'AT_goal_diff', 
                                   #'DiffLP': 'Diff_last_LP',
                                   'HTP': 'HT_Pts', 'ATP': 'AT_Pts', 
                                   'HTSOT_T': 'HT_Shots_OT_took','ATSOT_T': 'AT_Shots_OT_took', 
                                   'HTSOT_A': 'HT_Shots_OT_allowed','ATSOT_A': 'AT_Shots_OT_allowed', 
                                   'HT_Shot_Acc': 'HT_Shots_took_Acc', 'AT_Shot_Acc': 'AT_Shots_took_Acc',
                                   'HT_Op_Shot_Acc': 'HT_Shots_allowed_Acc', 'AT_Op_Shot_Acc': 'AT_Shots_allowed_Acc'
                                  }, inplace=True)

    """## Save created dataset"""

    Bundesliga_data_final.to_csv(path_or_buf= 'Final_data/Bundesliga_data_final.csv', index=False)

```

```python
def LaLiga_dataset_creator():
    # -*- coding: utf-8 -*-
    """LaLiga_dataset_creator.ipynb

    Automatically generated by Colaboratory.

    Original file is located at
        https://colab.research.google.com/drive/1xlkX8vKU_ll5fJ9y0r2cuKEi9Hm-Ig47

    ## Uploading data and basic operations (add season variable and change the date format)
    """

    import numpy as np
    import pandas as pd

    league_code = 'SP1'
    seasons = ["00-01", "01-02", "02-03", "03-04", "04-05", 
              "05-06", "06-07", "07-08", "08-09", "09-10", 
              "10-11", "11-12", "12-13", "13-14", "14-15", 
              "15-16", "16-17", "17-18", "18-19", "19-20", "20-21"]

    urls = []
    for season in seasons:
        x = 'https://www.football-data.co.uk/mmz4281/' + season.replace("-", "") + '/'+league_code + '.csv'
        urls.append(x)

    columns = ['Div', 'Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'B365H', 'B365D', 'B365A']

    """I want to focus on the last 10 completed seasons (from 2011/12 to 2020/21) but i need more historical data to calculate H2H (Head To Head) variables

    In some data from historical season there was no columns with shots data so I decided to also cut shots data from 00/01 to 10/11 season
    """

    LaLiga_00_01_data = pd.read_csv(urls[0])[columns[0:-7]]
    LaLiga_01_02_data = pd.read_csv(urls[1])[columns[0:-7]]
    LaLiga_02_03_data = pd.read_csv(urls[2])[columns[0:-7]]
    LaLiga_03_04_data = pd.read_csv(urls[3])[columns[0:-7]]
    LaLiga_04_05_data = pd.read_csv("match_data/SP1_04_05.csv", sep=';', encoding= 'unicode_escape')[columns[0:-7]] # Error Tokenizing Data with urls upload
    LaLiga_05_06_data = pd.read_csv(urls[5])[columns[0:-7]]
    LaLiga_06_07_data = pd.read_csv(urls[6])[columns[0:-7]]
    LaLiga_07_08_data = pd.read_csv(urls[7])[columns[0:-7]]
    LaLiga_08_09_data = pd.read_csv(urls[8])[columns[0:-7]]
    LaLiga_09_10_data = pd.read_csv(urls[9])[columns[0:-7]]
    LaLiga_10_11_data = pd.read_csv(urls[10])[columns[0:-7]]
    LaLiga_11_12_data = pd.read_csv(urls[11])[columns]
    LaLiga_12_13_data = pd.read_csv(urls[12])[columns]
    LaLiga_13_14_data = pd.read_csv(urls[13])[columns]
    LaLiga_14_15_data = pd.read_csv(urls[14])[columns]
    LaLiga_15_16_data = pd.read_csv(urls[15])[columns]
    LaLiga_16_17_data = pd.read_csv(urls[16])[columns]
    LaLiga_17_18_data = pd.read_csv(urls[17])[columns]
    LaLiga_18_19_data = pd.read_csv(urls[18])[columns]
    LaLiga_19_20_data = pd.read_csv(urls[19])[columns]
    LaLiga_20_21_data = pd.read_csv(urls[20])[columns]

    LaLiga_20_21_data['Season'] = '2020/21'
    LaLiga_19_20_data['Season'] = '2019/20'
    LaLiga_18_19_data['Season'] = '2018/19'
    LaLiga_17_18_data['Season'] = '2017/18'
    LaLiga_16_17_data['Season'] = '2016/17'
    LaLiga_15_16_data['Season'] = '2015/16'
    LaLiga_14_15_data['Season'] = '2014/15'
    LaLiga_13_14_data['Season'] = '2013/14'
    LaLiga_12_13_data['Season'] = '2012/13'
    LaLiga_11_12_data['Season'] = '2011/12'
    LaLiga_10_11_data['Season'] = '2010/11'
    LaLiga_09_10_data['Season'] = '2009/10'
    LaLiga_08_09_data['Season'] = '2008/09'
    LaLiga_07_08_data['Season'] = '2007/08'
    LaLiga_06_07_data['Season'] = '2006/07'
    LaLiga_05_06_data['Season'] = '2005/06'
    LaLiga_04_05_data['Season'] = '2004/05'
    LaLiga_03_04_data['Season'] = '2003/04'
    LaLiga_02_03_data['Season'] = '2002/03'
    LaLiga_01_02_data['Season'] = '2001/02'
    LaLiga_00_01_data['Season'] = '2000/01'

    seasons = [LaLiga_00_01_data, LaLiga_01_02_data, LaLiga_02_03_data, LaLiga_03_04_data, LaLiga_04_05_data,
              LaLiga_05_06_data, LaLiga_06_07_data, LaLiga_07_08_data, LaLiga_08_09_data, LaLiga_09_10_data, 
              LaLiga_10_11_data, LaLiga_11_12_data, LaLiga_12_13_data, LaLiga_13_14_data, LaLiga_14_15_data, 
              LaLiga_15_16_data, LaLiga_16_17_data, LaLiga_17_18_data, LaLiga_18_19_data, LaLiga_19_20_data, 
              LaLiga_20_21_data]

    # checking if every season was uploaded correctly without NA
    for season in seasons:
        print(season.Season.unique())
        print(season.Date[0])
        print(len(season)) # every season should have 380 rows
        print('|----------------------------|')

    # changing Date format
    for season in seasons:
        season.Date = pd.to_datetime(season.Date, dayfirst=True)

    """## Adds the number of games played by the team before specific match"""

    def get_last_matches_number(playing_stat):
        
        # Diffrent lists for HomeTeam and AwayTeam because sometimes number of games played before 
        # can be diffrent becasue of postponing the match
        HT_Match_number = []
        AT_Match_number = []
        
        # dataset iteration over each row
        for i, row in playing_stat.iterrows():
            df_nad = playing_stat.iloc[0:i]
            df_specific_match = playing_stat.iloc[i]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            x = len(df_nad[(df_nad.HomeTeam == ht) | (df_nad.AwayTeam == ht)])
            y = len(df_nad[(df_nad.HomeTeam == at) | (df_nad.AwayTeam == at)])

            HT_Match_number.append(x)
            AT_Match_number.append(y)
            
        playing_stat['HT_last_matches_number'] = HT_Match_number
        playing_stat['AT_last_matches_number'] = AT_Match_number
        
        return playing_stat

    for season in seasons:
        season = get_last_matches_number(season)

    """## Goals scored and conceded before specific match"""

    # Gets the goals scored aggregation arranged by teams and matchweek
    def get_goals_scored(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTGS = playing_stat.iloc[i]['FTHG']
            ATGS = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGS)
            
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        GoalsScored[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,39): # we have 38 matchweeks (number of teams*2 - 2 becacue you cant play with yourself)
            GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
        return GoalsScored

    # Gets the goals conceded aggregation arranged by teams and matchweek
    def get_goals_conceded(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATGC = playing_stat.iloc[i]['FTHG']
            HTGC = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGC)
        
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        GoalsConceded[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,39):
            GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
        return GoalsConceded

    # function summarizing and adding variables
    def get_gss(playing_stat):
        GC = get_goals_conceded(playing_stat)
        GS = get_goals_scored(playing_stat)
       
        HTGS = []
        ATGS = []
        HTGC = []
        ATGC = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTGS.append(GS.loc[ht][ht_match])
            ATGS.append(GS.loc[at][at_match])
            
            HTGC.append(GC.loc[ht][ht_match])
            ATGC.append(GC.loc[at][at_match])
            
            
        playing_stat['HTGS'] = HTGS
        playing_stat['ATGS'] = ATGS
        playing_stat['HTGC'] = HTGC
        playing_stat['ATGC'] = ATGC
        
        return playing_stat

    for season in seasons:
        season = get_gss(season)

    """## Shots on target took and allowed before specific match"""

    # Gets the shots on target took aggregation arranged by teams and matchweek
    def get_shots_on_target_took(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTSOT = playing_stat.iloc[i]['HST']
            ATSOT = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT)
            
        # Create a dataframe for shots on target took where rows are teams and cols are matchweek.
        ShotsOT = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        ShotsOT[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,39):
            ShotsOT[i] = ShotsOT[i] + ShotsOT[i-1]
        return ShotsOT

    # Gets the shots on target allowed aggregation arranged by teams and matchweek
    def get_shots_on_target_allowed(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATSOT_A = playing_stat.iloc[i]['HST']
            HTSOT_A = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT_A)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT_A)
        
        # Create a dataframe for shots on target allowed where rows are teams and cols are matchweek.
        ShotsOT_Allowed = pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T
        ShotsOT_Allowed[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,39):
            ShotsOT_Allowed[i] = ShotsOT_Allowed[i] + ShotsOT_Allowed[i-1]
        return ShotsOT_Allowed

    # function summarizing and adding variables
    def get_sot(playing_stat):
        SOT_T = get_shots_on_target_took(playing_stat)
        SOT_A = get_shots_on_target_allowed(playing_stat)
       
        HTSOT_T = []
        ATSOT_T = []
        HTSOT_A = []
        ATSOT_A = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTSOT_T.append(SOT_T.loc[ht][ht_match])
            ATSOT_T.append(SOT_T.loc[at][at_match])
            
            HTSOT_A.append(SOT_A.loc[ht][ht_match])
            ATSOT_A.append(SOT_A.loc[at][at_match])
            
            
        playing_stat['HTSOT_T'] = HTSOT_T
        playing_stat['ATSOT_T'] = ATSOT_T
        playing_stat['HTSOT_A'] = HTSOT_A
        playing_stat['ATSOT_A'] = ATSOT_A
        
        return playing_stat

    for season in [ LaLiga_11_12_data,
                LaLiga_12_13_data, 
                LaLiga_13_14_data,
                LaLiga_14_15_data, 
                LaLiga_15_16_data, 
                LaLiga_16_17_data, 
                LaLiga_17_18_data, 
                LaLiga_18_19_data, 
                LaLiga_19_20_data, 
                LaLiga_20_21_data]:
        season = get_sot(season)

    # Get shots accuracy
    def get_shots_acc(playing_stat):
        
        HT_Shot_Acc = playing_stat.HTGS / playing_stat.HTSOT_T # HomeTeam shots took accuracy (Goals scoring accuracy)
        AT_Shot_Acc = playing_stat.ATGS / playing_stat.ATSOT_T # AwayTeam shots took accuracy (Goals scoring accuracy)
        
        HT_Op_Shot_Acc = playing_stat.HTGC / playing_stat.HTSOT_A # HomeTeam shots allowed accuracy (Goals conceded accuracy)
        AT_Op_Shot_Acc = playing_stat.ATGC / playing_stat.ATSOT_A # AwayTeam shots allowed accuracy (Goals conceded accuracy)
        
        playing_stat['HT_Shot_Acc'] = HT_Shot_Acc
        playing_stat['AT_Shot_Acc'] = AT_Shot_Acc
        playing_stat['HT_Op_Shot_Acc'] = HT_Op_Shot_Acc
        playing_stat['AT_Op_Shot_Acc'] = AT_Op_Shot_Acc
        
        return playing_stat

    for season in [ LaLiga_11_12_data,
                LaLiga_12_13_data, 
                LaLiga_13_14_data,
                LaLiga_14_15_data, 
                LaLiga_15_16_data, 
                LaLiga_16_17_data, 
                LaLiga_17_18_data, 
                LaLiga_18_19_data, 
                LaLiga_19_20_data, 
                LaLiga_20_21_data]:
        season = get_shots_acc(season)

    """## Cumulative points before specific match"""

    # Function for getting specific number of points depends on the match result
    def get_points(result):
        if result == 'W':
            return 3
        elif result == 'D':
            return 1
        else:
            return 0

    # Get match results for every team in every match week
    def get_matchres(season):
        #print("here")
        # Create a dictionary with team names as keys
        teams = {}
        for i in season.groupby('HomeTeam').mean().T.columns:
            teams[i] = []

        # the value corresponding to keys is a list containing the match result
        for i in range(len(season)):
            if season.iloc[i].FTR == 'H':
                teams[season.iloc[i].HomeTeam].append('W')
                teams[season.iloc[i].AwayTeam].append('L') # If HomeTeam wins the AwayTeam lose so wee need to add L
            elif season.iloc[i].FTR == 'A':
                teams[season.iloc[i].AwayTeam].append('W')
                teams[season.iloc[i].HomeTeam].append('L') # If AwayTeam wins the HomeTeam lose so wee need to add L
            else:
                teams[season.iloc[i].AwayTeam].append('D')
                teams[season.iloc[i].HomeTeam].append('D')

                
        return pd.DataFrame(data=teams, index = [i for i in range(1,39)]).T

    # Get cumulative points connected with match results
    def get_cuml_points(matchres):
        matchres_points = matchres.applymap(get_points)
        for i in range(2,38):
            matchres_points[i] = matchres_points[i] + matchres_points[i-1]
            
        matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(20)])
        return matchres_points

    # function summarizing and adding variables
    def get_agg_points(season):
        matchres = get_matchres(season)
        cum_pts = get_cuml_points(matchres)
        HTP = []
        ATP = []

        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HTP.append(cum_pts.loc[ht][ht_match])
            ATP.append(cum_pts.loc[at][at_match])
                
        season['HTP'] = HTP
        season['ATP'] = ATP
        return season

    for season in seasons:
        season = get_agg_points(season)

    """## Current teams form with match results for 5 last games"""

    # get and add to dataset last 5 match results
    def add_form(playing_stat,num):
        form = get_matchres(playing_stat)
        
        for number in range(5+1): # I am interesing to 5 last games
            form[-number] = 'ND'  # No data
        
        HT_form = []
        AT_form = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HT_form.append(form.loc[ht][ht_match - (num - 1)])
            AT_form.append(form.loc[at][at_match - (num - 1)])
            
        # every last match will have different variable  
        playing_stat['HM' + str(num)] = HT_form                 
        playing_stat['AM' + str(num)] = AT_form
        
        return playing_stat

    # function summarizing and adding variables
    def add_form_df(playing_statistics):
        playing_statistics = add_form(playing_statistics,1)
        playing_statistics = add_form(playing_statistics,2)
        playing_statistics = add_form(playing_statistics,3)
        playing_statistics = add_form(playing_statistics,4)
        playing_statistics = add_form(playing_statistics,5)
        return playing_statistics

    for season in seasons:
        season = add_form_df(season)

    """## Cumulative number of wins, losses nad draws before specific match"""

    # Binary variables connected with match results
    def get_losses(result):
        if result == 'L':
            return 1
        elif result == 'D' or  result == 'W':
            return 0
        else:
            return np.nan     

    def get_wins(result):
        if result == 'W':
            return 1
        elif result == 'L' or  result == 'D':
            return 0
        else:
            return np.nan 

    def get_draws(result):
        if result == 'D':
            return 1
        elif result == 'L' or  result == 'W':
            return 0
        else:
            return np.nan 
        
    # Get cumulative results during specific season
    def get_agg_results(season):
        matchres = get_matchres(season)
        cum_wins = matchres.applymap(get_wins).cumsum(axis=1)
        cum_losses = matchres.applymap(get_losses).cumsum(axis=1)
        cum_draws = matchres.applymap(get_draws).cumsum(axis=1)
        
        cum_wins[0] = 0
        cum_losses[0] = 0
        cum_draws[0] = 0
        
        HT_cum_wins = []
        AT_cum_wins = []
        
        HT_cum_losses = []
        AT_cum_losses = []
        
        HT_cum_draws = []
        AT_cum_draws = []
        
        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HT_cum_wins.append(cum_wins.loc[ht][ht_match])
            AT_cum_wins.append(cum_wins.loc[at][at_match])
            
            HT_cum_losses.append(cum_losses.loc[ht][ht_match])
            AT_cum_losses.append(cum_losses.loc[at][at_match])
            
            HT_cum_draws.append(cum_draws.loc[ht][ht_match])
            AT_cum_draws.append(cum_draws.loc[at][at_match])
                
        season['HT_cum_wins'] = HT_cum_wins
        season['AT_cum_wins'] = AT_cum_wins
        
        season['HT_cum_losses'] = HT_cum_losses
        season['AT_cum_losses'] = AT_cum_losses
        
        season['HT_cum_draws'] = HT_cum_draws
        season['AT_cum_draws'] = AT_cum_draws
        
        return season

    for season in seasons:
        season = get_agg_results(season)

    """## Offensive and Defensive strength

    The entire calculation process is presented in the link below:
    * [Calcuation process](https://help.smarkets.com/hc/en-gb/articles/115001457989-How-to-calculate-Poisson-distribution-for-football-betting)
    """

    def get_off_def_strength(playing_stat):
        
        List_off_strength_HT = []
        List_def_strength_HT = []
        List_off_strength_AT = []
        List_def_strength_AT = []

        for i in range(len(playing_stat)):
            df_specific_match = playing_stat.iloc[i]
            df_day_before_SpecMat = playing_stat[playing_stat.Date < df_specific_match.Date]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            Avg_score_Home = np.mean(df_day_before_SpecMat.FTHG) if np.mean(df_day_before_SpecMat.FTHG) !=0 else np.nan
            Avg_score_Away = np.mean(df_day_before_SpecMat.FTAG) if np.mean(df_day_before_SpecMat.FTAG) !=0 else np.nan

            Avg_conced_Home = Avg_score_Away
            Avg_conced_Away = Avg_score_Home

            # Home Team offensive strength
            Avg_home_goals_scored_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTHG.mean()
            Off_strength_HT = Avg_home_goals_scored_by_HT / Avg_score_Home

            # Away Team defensive strength
            Avg_away_goals_conceded_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTHG.mean()
            Def_strength_AT = Avg_away_goals_conceded_by_AT / Avg_conced_Away

            List_off_strength_HT.append(Off_strength_HT)
            List_def_strength_AT.append(Def_strength_AT)

            # Away Team offensive strength
            Avg_away_goals_scored_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTAG.mean()
            Off_strength_AT = Avg_away_goals_scored_by_AT / Avg_score_Away

            # Home Team defensive strength
            Avg_home_goals_conceded_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTAG.mean()
            Def_strength_HT = Avg_home_goals_conceded_by_HT / Avg_conced_Home

            List_off_strength_AT.append(Off_strength_AT)
            List_def_strength_HT.append(Def_strength_HT)
      
        playing_stat['Off_strength_HT'] = List_off_strength_HT
        playing_stat['Def_strength_HT'] = List_def_strength_HT
        playing_stat['Off_strength_AT'] = List_off_strength_AT
        playing_stat['Def_strength_AT'] = List_def_strength_AT

    for season in seasons:
        season = get_off_def_strength(season)

    """## League position from last season

    In European football leauges every season teams (usually 3 but it depends) can get promotion to the highest league and other teams can fall out. This way we get different teams every seasons. So to get league postion from last season I decided to assign to new teams a specific postion:
    * number of teams in the highest league + new team's league postion from the second highest league
    * for example in 2020/21 season Huesca was new team and they have 1st postion in 2019/20 season in the second highest league so I assign number 21
    """

    Standings = pd.read_csv('league_tables/LaLiga_table_rank.csv', sep=';')
    Standings.set_index(['Team'], inplace=True)

    # Add variables with last league postion
    def get_last(playing_stat, Standings, year):
        HomeTeamLP = []
        AwayTeamLP = []
        for i in range(380):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam
            HomeTeamLP.append(Standings.loc[ht][year])
            AwayTeamLP.append(Standings.loc[at][year])
        playing_stat['HomeTeamLP'] = HomeTeamLP
        playing_stat['AwayTeamLP'] = AwayTeamLP
        return playing_stat

    LaLiga_11_12_data = get_last(LaLiga_11_12_data, Standings, '2010/11')
    LaLiga_12_13_data = get_last(LaLiga_12_13_data, Standings, '2011/12')
    LaLiga_13_14_data = get_last(LaLiga_13_14_data, Standings, '2012/13')
    LaLiga_14_15_data = get_last(LaLiga_14_15_data, Standings, '2013/14')
    LaLiga_15_16_data = get_last(LaLiga_15_16_data, Standings, '2014/15')
    LaLiga_16_17_data = get_last(LaLiga_16_17_data, Standings, '2015/16')
    LaLiga_17_18_data = get_last(LaLiga_17_18_data, Standings, '2016/17')
    LaLiga_18_19_data = get_last(LaLiga_18_19_data, Standings, '2017/18')
    LaLiga_19_20_data = get_last(LaLiga_19_20_data, Standings, '2018/19')
    LaLiga_20_21_data = get_last(LaLiga_20_21_data, Standings, '2019/20')

    # adding binary variable if specific team is a new team in specific season
    def get_newTeam(playing_stat):
        
        teams_number = len(playing_stat.HomeTeam.unique())
        
        playing_stat['Is_HT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.HomeTeamLP]
        playing_stat['Is_AT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.AwayTeamLP]
        
        return playing_stat

    # Data was only for the last 10 completed season so I cant use every historical dataset
    for season in [ LaLiga_11_12_data,
                LaLiga_12_13_data, 
                LaLiga_13_14_data,
                LaLiga_14_15_data, 
                LaLiga_15_16_data, 
                LaLiga_16_17_data, 
                LaLiga_17_18_data, 
                LaLiga_18_19_data, 
                LaLiga_19_20_data, 
                LaLiga_20_21_data]:
        season = get_newTeam(season)

    # Checking if everything is correct
    for season in [ LaLiga_11_12_data,
                LaLiga_12_13_data, 
                LaLiga_13_14_data,
                LaLiga_14_15_data, 
                LaLiga_15_16_data, 
                LaLiga_16_17_data, 
                LaLiga_17_18_data, 
                LaLiga_18_19_data, 
                LaLiga_19_20_data, 
                LaLiga_20_21_data]:
        print(season[season.HomeTeamLP > 20].HomeTeam.unique())

    """## Team data for every season from transfermarkt
    * Average age
    * Number of foreigners
    * Average market value in milions of Euro
    * Total market value in milions of Euro

    Example data from 2016/17 season:
    https://www.transfermarkt.com/laliga/startseite/wettbewerb/ES1/plus/?saison_id=2016
    """

    LaLiga_budget = pd.read_csv("budget_data/LaLiga_transfermaket_data.csv", sep=';')
    LaLiga_budget.set_index(['Club'], inplace=True)
    LaLiga_budget

    """Changing string variables for float"""

    LaLiga_budget['Avg_age'].replace(',', '.', regex=True, inplace=True)
    LaLiga_budget['Avg_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)
    LaLiga_budget['Total_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)

    LaLiga_budget.info()

    LaLiga_budget.Avg_age = LaLiga_budget.Avg_age.astype(float)
    LaLiga_budget.Avg_market_value_MLN_EUR = LaLiga_budget.Avg_market_value_MLN_EUR.astype(float)
    LaLiga_budget.Total_market_value_MLN_EUR = LaLiga_budget.Total_market_value_MLN_EUR.astype(float)

    # function summarizing and adding variables
    def get_transfermarkt_data(season_data, Budget):
        
        HomeTeam_TM_avg_age = []
        HomeTeam_TM_foreigners_number = []
        HomeTeam_TM_avg_market_value_MLN_EUR = []
        HomeTeam_TM_total_market_value_MLN_EUR = []
        
        AwayTeam_TM_avg_age = []
        AwayTeam_TM_foreigners_number = []
        AwayTeam_TM_avg_market_value_MLN_EUR = []
        AwayTeam_TM_total_market_value_MLN_EUR = []
        
        for season in list(season_data.Season.unique()):
            
            LaLiga_budget_season = Budget[Budget.Season == season]
            
            for i in range(380):
                
                home_team = season_data[season_data.Season == season].iloc[i].HomeTeam
                away_team = season_data[season_data.Season == season].iloc[i].AwayTeam
                
                HomeTeam_TM_avg_age.append(LaLiga_budget_season.loc[home_team]['Avg_age'])
                HomeTeam_TM_foreigners_number.append(LaLiga_budget_season.loc[home_team]['Foreigners'])
                HomeTeam_TM_avg_market_value_MLN_EUR.append(LaLiga_budget_season.loc[home_team]['Avg_market_value_MLN_EUR'])
                HomeTeam_TM_total_market_value_MLN_EUR.append(LaLiga_budget_season.loc[home_team]['Total_market_value_MLN_EUR'])
                
                AwayTeam_TM_avg_age.append(LaLiga_budget_season.loc[away_team]['Avg_age'])
                AwayTeam_TM_foreigners_number.append(LaLiga_budget_season.loc[away_team]['Foreigners'])
                AwayTeam_TM_avg_market_value_MLN_EUR.append(LaLiga_budget_season.loc[away_team]['Avg_market_value_MLN_EUR'])
                AwayTeam_TM_total_market_value_MLN_EUR.append(LaLiga_budget_season.loc[away_team]['Total_market_value_MLN_EUR'])
                
                
        season_data['HomeTeam_TM_avg_age'] = HomeTeam_TM_avg_age
        season_data['HomeTeam_TM_foreigners_number'] = HomeTeam_TM_foreigners_number
        season_data['HomeTeam_TM_avg_market_value_MLN_EUR'] = HomeTeam_TM_avg_market_value_MLN_EUR
        season_data['HomeTeam_TM_total_market_value_MLN_EUR'] = HomeTeam_TM_total_market_value_MLN_EUR
        
        season_data['AwayTeam_TM_avg_age'] = AwayTeam_TM_avg_age
        season_data['AwayTeam_TM_foreigners_number'] = AwayTeam_TM_foreigners_number
        season_data['AwayTeam_TM_avg_market_value_MLN_EUR'] = AwayTeam_TM_avg_market_value_MLN_EUR
        season_data['AwayTeam_TM_total_market_value_MLN_EUR'] = AwayTeam_TM_total_market_value_MLN_EUR
        
        return season_data

    for season in [LaLiga_11_12_data, LaLiga_12_13_data, LaLiga_13_14_data, LaLiga_14_15_data, LaLiga_15_16_data, 
                  LaLiga_16_17_data, LaLiga_17_18_data, LaLiga_18_19_data, LaLiga_19_20_data, LaLiga_20_21_data]:
        
        get_transfermarkt_data(season, LaLiga_budget)

    """## Data concatenating"""

    LaLiga_data = pd.concat([LaLiga_00_01_data, 
                          LaLiga_01_02_data, 
                          LaLiga_02_03_data, 
                          LaLiga_03_04_data, 
                          LaLiga_04_05_data, 
                          LaLiga_05_06_data, 
                          LaLiga_06_07_data, 
                          LaLiga_07_08_data, 
                          LaLiga_08_09_data, 
                          LaLiga_09_10_data, 
                          LaLiga_10_11_data, 
                          LaLiga_11_12_data,
                          LaLiga_12_13_data, 
                          LaLiga_13_14_data,
                          LaLiga_14_15_data, 
                          LaLiga_15_16_data, 
                          LaLiga_16_17_data, 
                          LaLiga_17_18_data, 
                          LaLiga_18_19_data, 
                          LaLiga_19_20_data, 
                          LaLiga_20_21_data], ignore_index=True)

    # deleting unnecessary data
    del LaLiga_00_01_data, LaLiga_01_02_data, LaLiga_02_03_data, LaLiga_03_04_data, LaLiga_04_05_data, LaLiga_05_06_data
    del LaLiga_06_07_data, LaLiga_07_08_data, LaLiga_08_09_data, LaLiga_09_10_data
    del season, seasons

    """## Calcuating betting odds
    * 1x (Home win or Draw) = 100 / (100/Home_win_odds + 100/Draw_odds)
    * x2 (Away win or Draw) = 100 / (100/Away_win_odds + 100/Draw_odds)
    """

    def get_more_odds(playing_stat):
        x1 = []
        x2 = []
        
        for i, row in playing_stat.iterrows():
            b365_H = playing_stat.iloc[i]['B365H']
            b365_D = playing_stat.iloc[i]['B365D']
            b365_A = playing_stat.iloc[i]['B365A']
            
            odd_H = 100/b365_H
            odd_D = 100/b365_D
            odd_A = 100/b365_A
            
            b365_H_D = round((100/(odd_H + odd_D)), 2)
            b365_A_D = round((100/(odd_A + odd_D)), 2)
            
            x1.append(b365_H_D)
            x2.append(b365_A_D)
            
        
        playing_stat['B365_HD'] = x1
        playing_stat['B365_AD'] = x2
        
        return playing_stat

    get_more_odds(LaLiga_data)

    """## Add points collect by teams from last 3 and 5 games"""

    # Gets the form points.
    def get_form_points(string):
        sum = 0
        for letter in string:
            sum += get_points(letter) # using other function
        return sum

    # Concatenate variables with last team's results
    LaLiga_data['HT_3match_FormPtsStr'] = LaLiga_data['HM1'] + LaLiga_data['HM2'] + LaLiga_data['HM3']
    LaLiga_data['AT_3match_FormPtsStr'] = LaLiga_data['AM1'] + LaLiga_data['AM2'] + LaLiga_data['AM3']

    LaLiga_data['HT_5match_FormPtsStr'] = LaLiga_data['HM1'] + LaLiga_data['HM2'] + LaLiga_data['HM3'] + LaLiga_data['HM4'] + LaLiga_data['HM5']
    LaLiga_data['AT_5match_FormPtsStr'] = LaLiga_data['AM1'] + LaLiga_data['AM2'] + LaLiga_data['AM3'] + LaLiga_data['AM4'] + LaLiga_data['AM5']

    # Applying function to get points
    LaLiga_data['HT_3match_FormPts'] = LaLiga_data['HT_3match_FormPtsStr'].apply(get_form_points)
    LaLiga_data['AT_3match_FormPts'] = LaLiga_data['AT_3match_FormPtsStr'].apply(get_form_points)

    LaLiga_data['HT_5match_FormPts'] = LaLiga_data['HT_5match_FormPtsStr'].apply(get_form_points)
    LaLiga_data['AT_5match_FormPts'] = LaLiga_data['AT_5match_FormPtsStr'].apply(get_form_points)

    """## Win or loss streaks"""

    # Identify Win/Loss Streaks if any.
    def get_3game_ws(string):
        if string == 'WWW':
            return 1
        else:
            return 0
        
    def get_5game_ws(string):
        if string == 'WWWWW':
            return 1
        else:
            return 0
        
    def get_3game_ls(string):
        if string == 'LLL':
            return 1
        else:
            return 0
        
    def get_5game_ls(string):
        if string == 'LLLLL':
            return 1
        else:
            return 0

    LaLiga_data['HTWinStreak3'] = LaLiga_data['HT_3match_FormPtsStr'].apply(get_3game_ws)
    LaLiga_data['HTWinStreak5'] = LaLiga_data['HT_5match_FormPtsStr'].apply(get_5game_ws)
    LaLiga_data['HTLossStreak3'] = LaLiga_data['HT_3match_FormPtsStr'].apply(get_3game_ls)
    LaLiga_data['HTLossStreak5'] = LaLiga_data['HT_5match_FormPtsStr'].apply(get_5game_ls)

    LaLiga_data['ATWinStreak3'] = LaLiga_data['AT_3match_FormPtsStr'].apply(get_3game_ws)
    LaLiga_data['ATWinStreak5'] = LaLiga_data['AT_5match_FormPtsStr'].apply(get_5game_ws)
    LaLiga_data['ATLossStreak3'] = LaLiga_data['AT_3match_FormPtsStr'].apply(get_3game_ls)
    LaLiga_data['ATLossStreak5'] = LaLiga_data['AT_5match_FormPtsStr'].apply(get_5game_ls)

    """## Head to Head (H2H) variables
    * Getting last 3 head to head match results and change it for points
    """

    def h2h(playing_stat):
        
        playing_stat["H2H_Home"] = ""
        playing_stat["H2H_Away"] = ""

        for i in range(1,len(playing_stat)):
            # get teams from specifc match and all data before this match
            df = playing_stat.iloc[0:i-1,]
            ht = playing_stat.loc[i,"HomeTeam"]
            at = playing_stat.loc[i,"AwayTeam"]
            
            # getting last 3 head to head matches
            dff = df[(((df.HomeTeam == ht) &  (df.AwayTeam == at)) | ((df.HomeTeam == at) &  (df.AwayTeam == ht)))].tail(3)
            
            # using data with minimum 1 head to head match
            if (len(dff) >= 1):
                for j in range(0,len(dff)):
                    # Consider all possible scenarios
                    if ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                        
            # if there was no match beetwen the teams get 'ND'
            else: 
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] = "ND"
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] = "ND"
        return playing_stat


    # Calculating points for Home and Away teams in head to head matches and adding variables
    def get_points_h2h(playing_stat, col):
        
        playing_stat[col+"_pts"] = 0
        
        for i in range(1,len(playing_stat)):
            if (len(playing_stat.iloc[i][col]) == 1):
                if playing_stat.iloc[i][col][0] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][0] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                else:
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            elif (len(playing_stat.iloc[i][col]) == 2):
                for j in range(0,2):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
                    else:
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
            elif (len(playing_stat.iloc[i][col]) == 3):
                for j in range(0,3):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            else:
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] = 0
        
        return playing_stat

    h2h(LaLiga_data)

    get_points_h2h(LaLiga_data, "H2H_Home") 
    get_points_h2h(LaLiga_data, "H2H_Away")

    """## Choose last 10 completed seasons"""

    LaLiga_data = LaLiga_data[LaLiga_data.Season.isin(['2011/12', 
                                             '2012/13', 
                                             '2013/14', 
                                             '2014/15', 
                                             '2015/16', 
                                             '2016/17', 
                                             '2017/18', 
                                             '2018/19', 
                                             '2019/20', 
                                             '2020/21'])]

    LaLiga_data.reset_index(drop=True, inplace=True)
    LaLiga_data.shape

    LaLiga_data_final = LaLiga_data[[
        'Div', 'Season', 'Date', 'HomeTeam', 'AwayTeam','FTHG', 'FTAG', 'FTR', 'HS','AS', 'HST', 'AST',
        'HT_last_matches_number', 'AT_last_matches_number', 'B365H', 'B365D', 'B365A', 'B365_HD', 'B365_AD',
        'HTGS', 'ATGS', 'HTGC', 'ATGC',
        'HTSOT_T','ATSOT_T', 'HTSOT_A', 'ATSOT_A', 
        'HT_Shot_Acc', 'AT_Shot_Acc','HT_Op_Shot_Acc', 'AT_Op_Shot_Acc',
        'HTP', 'ATP',
        'HT_cum_wins', 'AT_cum_wins', 'HT_cum_losses', 'AT_cum_losses', 'HT_cum_draws', 'AT_cum_draws',
        'Off_strength_HT', 'Def_strength_HT', 'Off_strength_AT', 'Def_strength_AT',
        'HM1', 'AM1', 'HM2', 'AM2', 'HM3', 'AM3', 'HM4', 'AM4', 'HM5', 'AM5',
        'HT_3match_FormPtsStr', 'AT_3match_FormPtsStr', 'HT_5match_FormPtsStr','AT_5match_FormPtsStr', 
        'HT_3match_FormPts', 'AT_3match_FormPts', 'HT_5match_FormPts', 'AT_5match_FormPts',
        'HTWinStreak3','HTWinStreak5', 'HTLossStreak3', 'HTLossStreak5', 
        'ATWinStreak3','ATWinStreak5', 'ATLossStreak3', 'ATLossStreak5',
        'HomeTeamLP', 'AwayTeamLP',
        'Is_HT_newTeam', 'Is_AT_newTeam',
        'H2H_Home', 'H2H_Away', 'H2H_Home_pts','H2H_Away_pts',
        'HomeTeam_TM_avg_age','HomeTeam_TM_foreigners_number', 
        'HomeTeam_TM_avg_market_value_MLN_EUR','HomeTeam_TM_total_market_value_MLN_EUR',
        'AwayTeam_TM_avg_age', 'AwayTeam_TM_foreigners_number', 
        'AwayTeam_TM_avg_market_value_MLN_EUR', 'AwayTeam_TM_total_market_value_MLN_EUR'
                              ]]

    LaLiga_data_final.rename(columns={'HM1': 'HT_last_match1', 'AM1': 'AT_last_match1', 
                                   'HM2': 'HT_last_match2', 'AM2': 'AT_last_match2', 
                                   'HM3': 'HT_last_match3', 'AM3': 'AT_last_match3', 
                                   'HM4': 'HT_last_match4', 'AM4': 'AT_last_match4', 
                                   'HM5': 'HT_last_match5', 'AM5': 'AT_last_match5', 
                                   'HomeTeamLP': 'HT_last_LP', 'AwayTeamLP': 'AT_last_LP', 
                                   #'HTGD': 'HT_goal_diff', 'ATGD': 'AT_goal_diff', 
                                   #'DiffLP': 'Diff_last_LP',
                                   'HTP': 'HT_Pts', 'ATP': 'AT_Pts', 
                                   'HTSOT_T': 'HT_Shots_OT_took','ATSOT_T': 'AT_Shots_OT_took', 
                                   'HTSOT_A': 'HT_Shots_OT_allowed','ATSOT_A': 'AT_Shots_OT_allowed', 
                                   'HT_Shot_Acc': 'HT_Shots_took_Acc', 'AT_Shot_Acc': 'AT_Shots_took_Acc',
                                   'HT_Op_Shot_Acc': 'HT_Shots_allowed_Acc', 'AT_Op_Shot_Acc': 'AT_Shots_allowed_Acc'
                                  }, inplace=True)

    """## Save created dataset"""

    LaLiga_data_final.to_csv(path_or_buf= 'Final_data/LaLiga_data_final.csv', index=False)

```

```python
def Ligue1_dataset_creator():    
    # -*- coding: utf-8 -*-
    """Ligue1_dataset_creator.ipynb

    Automatically generated by Colaboratory.

    Original file is located at
        https://colab.research.google.com/drive/1Amp1bhGXe3Bj0Vg6PGvn5Cn620DQDwo8

    ## Uploading data and basic operations (add season variable and change the date format)
    """

    import numpy as np
    import pandas as pd

    league_code = 'F1'
    seasons = ["00-01", "01-02", "02-03", "03-04", "04-05", 
              "05-06", "06-07", "07-08", "08-09", "09-10", 
              "10-11", "11-12", "12-13", "13-14", "14-15", 
              "15-16", "16-17", "17-18", "18-19", "19-20", "20-21"]

    urls = []
    for season in seasons:
        x = 'https://www.football-data.co.uk/mmz4281/' + season.replace("-", "") + '/'+league_code + '.csv'
        urls.append(x)

    columns = ['Div', 'Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'B365H', 'B365D', 'B365A']

    """I want to focus on the last 10 completed seasons (from 2011/12 to 2020/21) but i need more historical data to calculate H2H (Head To Head) variables

    In some data from historical season there was no columns with shots data so I decided to also cut shots data from 00/01 to 10/11 season
    """

    Ligue1_00_01_data = pd.read_csv(urls[0])[columns[0:-7]]
    Ligue1_01_02_data = pd.read_csv(urls[1])[columns[0:-7]]
    Ligue1_02_03_data = pd.read_csv("match_data/F1_02_03.csv", sep=';')[columns[0:-7]] # Error Tokenizing Data with urls upload
    Ligue1_03_04_data = pd.read_csv("match_data/F1_03_04.csv", sep=';')[columns[0:-7]] # Error Tokenizing Data with urls upload
    Ligue1_04_05_data = pd.read_csv(urls[4])[columns[0:-7]]
    Ligue1_05_06_data = pd.read_csv(urls[5])[columns[0:-7]]
    Ligue1_06_07_data = pd.read_csv(urls[6])[columns[0:-7]]
    Ligue1_07_08_data = pd.read_csv(urls[7])[columns[0:-7]]
    Ligue1_08_09_data = pd.read_csv(urls[8])[columns[0:-7]]
    Ligue1_09_10_data = pd.read_csv(urls[9])[columns[0:-7]]
    Ligue1_10_11_data = pd.read_csv(urls[10])[columns[0:-7]]
    Ligue1_11_12_data = pd.read_csv(urls[11])[columns]
    Ligue1_12_13_data = pd.read_csv(urls[12])[columns]
    Ligue1_13_14_data = pd.read_csv(urls[13])[columns]
    Ligue1_14_15_data = pd.read_csv(urls[14])[columns]
    Ligue1_15_16_data = pd.read_csv(urls[15])[columns]
    Ligue1_16_17_data = pd.read_csv(urls[16])[columns]
    Ligue1_17_18_data = pd.read_csv(urls[17])[columns]
    Ligue1_18_19_data = pd.read_csv(urls[18])[columns]
    Ligue1_19_20_data = pd.read_csv(urls[19])[columns]
    Ligue1_20_21_data = pd.read_csv(urls[20])[columns]

    Ligue1_20_21_data['Season'] = '2020/21'
    Ligue1_19_20_data['Season'] = '2019/20'
    Ligue1_18_19_data['Season'] = '2018/19'
    Ligue1_17_18_data['Season'] = '2017/18'
    Ligue1_16_17_data['Season'] = '2016/17'
    Ligue1_15_16_data['Season'] = '2015/16'
    Ligue1_14_15_data['Season'] = '2014/15'
    Ligue1_13_14_data['Season'] = '2013/14'
    Ligue1_12_13_data['Season'] = '2012/13'
    Ligue1_11_12_data['Season'] = '2011/12'
    Ligue1_10_11_data['Season'] = '2010/11'
    Ligue1_09_10_data['Season'] = '2009/10'
    Ligue1_08_09_data['Season'] = '2008/09'
    Ligue1_07_08_data['Season'] = '2007/08'
    Ligue1_06_07_data['Season'] = '2006/07'
    Ligue1_05_06_data['Season'] = '2005/06'
    Ligue1_04_05_data['Season'] = '2004/05'
    Ligue1_03_04_data['Season'] = '2003/04'
    Ligue1_02_03_data['Season'] = '2002/03'
    Ligue1_01_02_data['Season'] = '2001/02'
    Ligue1_00_01_data['Season'] = '2000/01'

    seasons = [Ligue1_00_01_data, Ligue1_01_02_data, Ligue1_02_03_data, Ligue1_03_04_data, Ligue1_04_05_data,
              Ligue1_05_06_data, Ligue1_06_07_data, Ligue1_07_08_data, Ligue1_08_09_data, Ligue1_09_10_data, 
              Ligue1_10_11_data, Ligue1_11_12_data, Ligue1_12_13_data, Ligue1_13_14_data, Ligue1_14_15_data, 
              Ligue1_15_16_data, Ligue1_16_17_data, Ligue1_17_18_data, Ligue1_18_19_data, Ligue1_19_20_data, 
              Ligue1_20_21_data]

    # checking if every season was uploaded correctly without NA
    for season in seasons:
        print(season.Season.unique())
        print(season.Date[0])
        season.dropna(inplace=True)
        print(len(season)) # in 00/01 and 01/02 seasons there was 18 teams in the league
        print('|----------------------------|')

    """**No relevant data in 2015/16, 2016/17, 2019/20 and 2020/21 seasons!**

    Upload data one more time
    """

    Ligue1_00_01_data = pd.read_csv(urls[0])[columns[0:-7]]
    Ligue1_01_02_data = pd.read_csv(urls[1])[columns[0:-7]]
    Ligue1_02_03_data = pd.read_csv("match_data/F1_02_03.csv", sep=';')[columns[0:-7]]
    Ligue1_03_04_data = pd.read_csv("match_data/F1_03_04.csv", sep=';')[columns[0:-7]]
    Ligue1_04_05_data = pd.read_csv(urls[4])[columns[0:-7]]
    Ligue1_05_06_data = pd.read_csv(urls[5])[columns[0:-7]]
    Ligue1_06_07_data = pd.read_csv(urls[6])[columns[0:-7]]
    Ligue1_07_08_data = pd.read_csv(urls[7])[columns[0:-7]]
    Ligue1_08_09_data = pd.read_csv(urls[8])[columns[0:-7]]
    Ligue1_09_10_data = pd.read_csv(urls[9])[columns[0:-7]]
    Ligue1_10_11_data = pd.read_csv(urls[10])[columns[0:-7]]
    Ligue1_11_12_data = pd.read_csv(urls[11])[columns]
    Ligue1_12_13_data = pd.read_csv(urls[12])[columns]
    Ligue1_13_14_data = pd.read_csv(urls[13])[columns]
    Ligue1_14_15_data = pd.read_csv(urls[14])[columns]
    Ligue1_15_16_data = pd.read_csv(urls[15])[columns]
    Ligue1_16_17_data = pd.read_csv(urls[16])[columns]
    Ligue1_17_18_data = pd.read_csv(urls[17])[columns]
    Ligue1_18_19_data = pd.read_csv(urls[18])[columns]
    Ligue1_19_20_data = pd.read_csv(urls[19])[columns]
    Ligue1_20_21_data = pd.read_csv(urls[20])[columns]

    """#### Season 15/16"""

    Ligue1_15_16_data[Ligue1_15_16_data.isna().any(axis=1)] # No betting odds for Marsylia - Nantes  match

    Ligue1_15_16_data_all_col = pd.read_csv(urls[15]) # upload data for 15/16 season with all columns

    pd.set_option("display.max_columns", None)
    Ligue1_15_16_data_all_col[Ligue1_15_16_data_all_col.isna().any(axis=1)]

    # Use betting odds from other bets company
    odds_344 = Ligue1_15_16_data_all_col[(Ligue1_15_16_data_all_col.HomeTeam == 'Marseille') & 
                              (Ligue1_15_16_data_all_col.AwayTeam == 'Nantes')][['BWH', 'BWD', 'BWA']]

    Ligue1_15_16_data.loc[344, 'B365H'] = odds_344.loc[344, 'BWH']
    Ligue1_15_16_data.loc[344, 'B365D'] = odds_344.loc[344, 'BWD']
    Ligue1_15_16_data.loc[344, 'B365A'] = odds_344.loc[344, 'BWA']

    """#### Season 16/17"""

    Ligue1_16_17_data[Ligue1_16_17_data.isna().any(axis=1)] # Match Bastia - Lyon in 2017 was a walkover

    """Fill in the statistics of shots in such a way that they do not disturb the accuracy calculated later"""

    Ligue1_16_17_data.loc[325, 'HS'] = 0
    Ligue1_16_17_data.loc[325, 'AS'] = 3
    Ligue1_16_17_data.loc[325, 'HST'] = 0
    Ligue1_16_17_data.loc[325, 'AST'] = 3

    Ligue1_16_17_data.loc[325]

    # Check if everything is correct
    Ligue1_16_17_data.dropna()

    """#### Season 2019/20"""

    Ligue1_19_20_data[Ligue1_19_20_data.isna().any(axis=1)]

    Ligue1_19_20_data # The French league in 2020 was not finished because of the pandemic, therefore there are fewer matches

    """#### Season 2020/21"""

    Ligue1_20_21_data[Ligue1_20_21_data.isna().any(axis=1)] # No betting odds

    Ligue1_20_21_data_all_col = pd.read_csv(urls[20]) # upload data with all columns
    Ligue1_20_21_data_all_col[Ligue1_20_21_data_all_col.isna().any(axis=1)]

    # Use average betting odds
    odds_26 = Ligue1_20_21_data_all_col[(Ligue1_20_21_data_all_col.HomeTeam == 'Paris SG') & 
                              (Ligue1_20_21_data_all_col.AwayTeam == 'Marseille')][['AvgH', 'AvgD', 'AvgA']]

    odds_66 = Ligue1_20_21_data_all_col[(Ligue1_20_21_data_all_col.HomeTeam == 'Monaco') & 
                              (Ligue1_20_21_data_all_col.AwayTeam == 'Montpellier')][['AvgH', 'AvgD', 'AvgA']]

    Ligue1_20_21_data.loc[26, 'B365H'] = odds_26.loc[26, 'AvgH']
    Ligue1_20_21_data.loc[26, 'B365D'] = odds_26.loc[26, 'AvgD']
    Ligue1_20_21_data.loc[26, 'B365A'] = odds_26.loc[26, 'AvgA']

    Ligue1_20_21_data.loc[66, 'B365H'] = odds_66.loc[66, 'AvgH']
    Ligue1_20_21_data.loc[66, 'B365D'] = odds_66.loc[66, 'AvgD']
    Ligue1_20_21_data.loc[66, 'B365A'] = odds_66.loc[66, 'AvgA']

    seasons = [Ligue1_00_01_data, Ligue1_01_02_data, Ligue1_02_03_data, Ligue1_03_04_data, Ligue1_04_05_data,
              Ligue1_05_06_data, Ligue1_06_07_data, Ligue1_07_08_data, Ligue1_08_09_data, Ligue1_09_10_data, 
              Ligue1_10_11_data, Ligue1_11_12_data, Ligue1_12_13_data, Ligue1_13_14_data, Ligue1_14_15_data, 
              Ligue1_15_16_data, Ligue1_16_17_data, Ligue1_17_18_data, Ligue1_18_19_data, Ligue1_19_20_data, 
              Ligue1_20_21_data]

    # checking if every season was uploaded correctly without NA
    for season in seasons:
        print(season.Date[0])
        season.dropna(how='all' ,inplace=True)
        print(len(season))
        print('|----------------------------|')

    Ligue1_20_21_data['Season'] = '2020/21'
    Ligue1_19_20_data['Season'] = '2019/20'
    Ligue1_18_19_data['Season'] = '2018/19'
    Ligue1_17_18_data['Season'] = '2017/18'
    Ligue1_16_17_data['Season'] = '2016/17'
    Ligue1_15_16_data['Season'] = '2015/16'
    Ligue1_14_15_data['Season'] = '2014/15'
    Ligue1_13_14_data['Season'] = '2013/14'
    Ligue1_12_13_data['Season'] = '2012/13'
    Ligue1_11_12_data['Season'] = '2011/12'
    Ligue1_10_11_data['Season'] = '2010/11'
    Ligue1_09_10_data['Season'] = '2009/10'
    Ligue1_08_09_data['Season'] = '2008/09'
    Ligue1_07_08_data['Season'] = '2007/08'
    Ligue1_06_07_data['Season'] = '2006/07'
    Ligue1_05_06_data['Season'] = '2005/06'
    Ligue1_04_05_data['Season'] = '2004/05'
    Ligue1_03_04_data['Season'] = '2003/04'
    Ligue1_02_03_data['Season'] = '2002/03'
    Ligue1_01_02_data['Season'] = '2001/02'
    Ligue1_00_01_data['Season'] = '2000/01'

    """Simple checking corrected datasets"""

    Ligue1_20_21_data[Ligue1_20_21_data.isna().any(axis=1)]

    Ligue1_16_17_data[Ligue1_16_17_data.isna().any(axis=1)]

    # Change Date Format
    for season in seasons:
        season.Date = pd.to_datetime(season.Date, dayfirst=True)

    """## Adds the number of games played by the team before specific match"""

    def get_last_matches_number(playing_stat):
        
        # Diffrent lists for HomeTeam and AwayTeam because sometimes number of games played before 
        # can be diffrent becasue of postponing the match
        HT_Match_number = []
        AT_Match_number = []
        
        # dataset iteration over each row
        for i, row in playing_stat.iterrows():
            df_nad = playing_stat.iloc[0:i]
            df_specific_match = playing_stat.iloc[i]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            x = len(df_nad[(df_nad.HomeTeam == ht) | (df_nad.AwayTeam == ht)])
            y = len(df_nad[(df_nad.HomeTeam == at) | (df_nad.AwayTeam == at)])

            HT_Match_number.append(x)
            AT_Match_number.append(y)
            
        playing_stat['HT_last_matches_number'] = HT_Match_number
        playing_stat['AT_last_matches_number'] = AT_Match_number
        
        return playing_stat

    # apply function on every season data
    for season in seasons:
        season = get_last_matches_number(season)

    """## Goals scored and conceded before specific match"""

    # Gets the goals scored aggregation arranged by teams and matchweek
    def get_goals_scored(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTGS = playing_stat.iloc[i]['FTHG']
            ATGS = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGS)
        
        # Ligue 1 in 2019/20 was not finished because of the pandemic so I need to create special approach
        if playing_stat.Season[0] == '2019/20':
            teams['Paris SG'].append(0)
            teams['Strasbourg'].append(0)
            
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams['Lyon'])+1)]).T
        GoalsScored[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams['Lyon'])+1):
            GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
        return GoalsScored

    # Gets the goals conceded aggregation arranged by teams and matchweek
    def get_goals_conceded(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATGC = playing_stat.iloc[i]['FTHG']
            HTGC = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGC)
            
        if playing_stat.Season[0] == '2019/20':
            teams['Paris SG'].append(0)
            teams['Strasbourg'].append(0)
        
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams['Lyon'])+1)]).T
        GoalsConceded[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams['Lyon'])+1):
            GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
        return GoalsConceded

    # function summarizing and adding variables
    def get_gss(playing_stat):
        GC = get_goals_conceded(playing_stat)
        GS = get_goals_scored(playing_stat)
       
        HTGS = []
        ATGS = []
        HTGC = []
        ATGC = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTGS.append(GS.loc[ht][ht_match])
            ATGS.append(GS.loc[at][at_match])
            
            HTGC.append(GC.loc[ht][ht_match])
            ATGC.append(GC.loc[at][at_match])
            
            
        playing_stat['HTGS'] = HTGS
        playing_stat['ATGS'] = ATGS
        playing_stat['HTGC'] = HTGC
        playing_stat['ATGC'] = ATGC
        
        return playing_stat

    for season in seasons:
        season = get_gss(season)

    """## Shots on target took and allowed before specific match"""

    # Gets the shots on target took aggregation arranged by teams and matchweek
    def get_shots_on_target_took(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTSOT = playing_stat.iloc[i]['HST']
            ATSOT = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT)
        
        # Ligue 1 in 2019/20 was not finished because of the pandemic so I need to create special approach
        if playing_stat.Season[0] == '2019/20':
            teams['Paris SG'].append(0)
            teams['Strasbourg'].append(0)
            
        # Create a dataframe for shots on target took where rows are teams and cols are matchweek.
        ShotsOT = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams['Lyon'])+1)]).T
        ShotsOT[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams['Lyon'])+1):
            ShotsOT[i] = ShotsOT[i] + ShotsOT[i-1]
        return ShotsOT

    # Gets the shots on target allowed aggregation arranged by teams and matchweek
    def get_shots_on_target_allowed(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATSOT_A = playing_stat.iloc[i]['HST']
            HTSOT_A = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT_A)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT_A)
            
        if playing_stat.Season[0] == '2019/20':
            teams['Paris SG'].append(0)
            teams['Strasbourg'].append(0)
        
        # Create a dataframe for shots on target allowed where rows are teams and cols are matchweek.
        ShotsOT_Allowed = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams['Lyon'])+1)]).T
        ShotsOT_Allowed[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams['Lyon'])+1):
            ShotsOT_Allowed[i] = ShotsOT_Allowed[i] + ShotsOT_Allowed[i-1]
        return ShotsOT_Allowed

    # function summarizing and adding variables
    def get_sot(playing_stat):
        SOT_T = get_shots_on_target_took(playing_stat)
        SOT_A = get_shots_on_target_allowed(playing_stat)
       
        HTSOT_T = []
        ATSOT_T = []
        HTSOT_A = []
        ATSOT_A = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTSOT_T.append(SOT_T.loc[ht][ht_match])
            ATSOT_T.append(SOT_T.loc[at][at_match])
            
            HTSOT_A.append(SOT_A.loc[ht][ht_match])
            ATSOT_A.append(SOT_A.loc[at][at_match])
            
            
        playing_stat['HTSOT_T'] = HTSOT_T
        playing_stat['ATSOT_T'] = ATSOT_T
        playing_stat['HTSOT_A'] = HTSOT_A
        playing_stat['ATSOT_A'] = ATSOT_A
        
        return playing_stat

    # Datasets with shots data
    for season in [ Ligue1_11_12_data,
                Ligue1_12_13_data, 
                Ligue1_13_14_data,
                Ligue1_14_15_data, 
                Ligue1_15_16_data, 
                Ligue1_16_17_data, 
                Ligue1_17_18_data, 
                Ligue1_18_19_data, 
                Ligue1_19_20_data, 
                Ligue1_20_21_data]:
        season = get_sot(season)

    # Get shots accuracy
    def get_shots_acc(playing_stat):
        
        HT_Shot_Acc = playing_stat.HTGS / playing_stat.HTSOT_T # HomeTeam shots took accuracy (Goals scoring accuracy)
        AT_Shot_Acc = playing_stat.ATGS / playing_stat.ATSOT_T # AwayTeam shots took accuracy (Goals scoring accuracy)
        
        HT_Op_Shot_Acc = playing_stat.HTGC / playing_stat.HTSOT_A # HomeTeam shots allowed accuracy (Goals conceded accuracy)
        AT_Op_Shot_Acc = playing_stat.ATGC / playing_stat.ATSOT_A # AwayTeam shots allowed accuracy (Goals conceded accuracy)
        
        playing_stat['HT_Shot_Acc'] = HT_Shot_Acc
        playing_stat['AT_Shot_Acc'] = AT_Shot_Acc
        playing_stat['HT_Op_Shot_Acc'] = HT_Op_Shot_Acc
        playing_stat['AT_Op_Shot_Acc'] = AT_Op_Shot_Acc
        
        return playing_stat

    # Datasets with shots data
    for season in [ Ligue1_11_12_data,
                Ligue1_12_13_data, 
                Ligue1_13_14_data,
                Ligue1_14_15_data, 
                Ligue1_15_16_data, 
                Ligue1_16_17_data, 
                Ligue1_17_18_data, 
                Ligue1_18_19_data, 
                Ligue1_19_20_data, 
                Ligue1_20_21_data]:
        season = get_shots_acc(season)

    """## Cumulative points before specific match"""

    # Function for getting specific number of points depends on the match result
    def get_points(result):
        if result == 'W':
            return 3
        elif result == 'D':
            return 1
        else:
            return 0

    # Get match results for every team in every match week
    def get_matchres(season):
        # Create a dictionary with team names as keys
        teams = {}
        for i in season.groupby('HomeTeam').mean().T.columns:
            teams[i] = []

        # the value corresponding to keys is a list containing the match result
        for i in range(len(season)):
            if season.iloc[i].FTR == 'H':
                teams[season.iloc[i].HomeTeam].append('W')
                teams[season.iloc[i].AwayTeam].append('L')
            elif season.iloc[i].FTR == 'A':
                teams[season.iloc[i].AwayTeam].append('W')
                teams[season.iloc[i].HomeTeam].append('L')
            else:
                teams[season.iloc[i].AwayTeam].append('D')
                teams[season.iloc[i].HomeTeam].append('D')
                
        # Ligue 1 in 2019/20 was not finished because of the pandemic so I need to create special approach
        if season.Season[0] == '2019/20':
            teams['Paris SG'].append('NM') # NM = No Match
            teams['Strasbourg'].append('NM')

                
        return pd.DataFrame(data=teams, index = [i for i in range(1,len(teams['Lyon'])+1)]).T
        
    # Get cumulative points connected with match results
    def get_cuml_points(matchres):
        
        matchres_points = matchres.applymap(get_points)
        for i in range(2,matchres.shape[1]):
            matchres_points[i] = matchres_points[i] + matchres_points[i-1]
            
        matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(len(matchres))])
        return matchres_points

    # function summarizing and adding variables
    def get_agg_points(season):
        matchres = get_matchres(season)
        cum_pts = get_cuml_points(matchres)
        HTP = []
        ATP = []

        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HTP.append(cum_pts.loc[ht][ht_match])
            ATP.append(cum_pts.loc[at][at_match])
                
        season['HTP'] = HTP
        season['ATP'] = ATP
        return season

    for season in seasons:
        season = get_agg_points(season)

    """## Current teams form with match results for 5 last games"""

    # get and add to dataset last 5 match results
    def add_form(playing_stat,num):
        form = get_matchres(playing_stat)
        
        for number in range(5+1): # I am interesing to 5 last games
            form[-number] = 'ND' # No data
        
        HT_form = []
        AT_form = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HT_form.append(form.loc[ht][ht_match - (num - 1)])
            AT_form.append(form.loc[at][at_match - (num - 1)])
            
        # every last match will have different variable  
        playing_stat['HM' + str(num)] = HT_form                 
        playing_stat['AM' + str(num)] = AT_form
        
        return playing_stat

    # function summarizing and adding variables
    def add_form_df(playing_statistics):
        playing_statistics = add_form(playing_statistics,1)
        playing_statistics = add_form(playing_statistics,2)
        playing_statistics = add_form(playing_statistics,3)
        playing_statistics = add_form(playing_statistics,4)
        playing_statistics = add_form(playing_statistics,5)
        return playing_statistics

    for season in seasons:
        season = add_form_df(season)

    """## Cumulative number of wins, losses nad draws before specific match"""

    # Binary variables connected with match results
    def get_losses(result):
        if result == 'L':
            return 1
        elif result == 'D' or  result == 'W':
            return 0
        else:
            return np.nan     

    def get_wins(result):
        if result == 'W':
            return 1
        elif result == 'L' or  result == 'D':
            return 0
        else:
            return np.nan 

    def get_draws(result):
        if result == 'D':
            return 1
        elif result == 'L' or  result == 'W':
            return 0
        else:
            return np.nan 
        
    # Get cumulative results during specific season
    def get_agg_results(season):
        matchres = get_matchres(season)
        cum_wins = matchres.applymap(get_wins).cumsum(axis=1)
        cum_losses = matchres.applymap(get_losses).cumsum(axis=1)
        cum_draws = matchres.applymap(get_draws).cumsum(axis=1)
        
        cum_wins[0] = 0
        cum_losses[0] = 0
        cum_draws[0] = 0
        
        HT_cum_wins = []
        AT_cum_wins = []
        
        HT_cum_losses = []
        AT_cum_losses = []
        
        HT_cum_draws = []
        AT_cum_draws = []
        
        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HT_cum_wins.append(cum_wins.loc[ht][ht_match])
            AT_cum_wins.append(cum_wins.loc[at][at_match])
            
            HT_cum_losses.append(cum_losses.loc[ht][ht_match])
            AT_cum_losses.append(cum_losses.loc[at][at_match])
            
            HT_cum_draws.append(cum_draws.loc[ht][ht_match])
            AT_cum_draws.append(cum_draws.loc[at][at_match])
                
        season['HT_cum_wins'] = HT_cum_wins
        season['AT_cum_wins'] = AT_cum_wins
        
        season['HT_cum_losses'] = HT_cum_losses
        season['AT_cum_losses'] = AT_cum_losses
        
        season['HT_cum_draws'] = HT_cum_draws
        season['AT_cum_draws'] = AT_cum_draws
        
        return season

    for season in seasons:
        season = get_agg_results(season)

    """## Offensive and Defensive strength

    The entire calculation process is presented in the link below:
    * [Calcuation process](https://help.smarkets.com/hc/en-gb/articles/115001457989-How-to-calculate-Poisson-distribution-for-football-betting)
    """

    def get_off_def_strength(playing_stat):
        
        List_off_strength_HT = []
        List_def_strength_HT = []
        List_off_strength_AT = []
        List_def_strength_AT = []

        for i in range(len(playing_stat)):
            df_specific_match = playing_stat.iloc[i]
            df_day_before_SpecMat = playing_stat[playing_stat.Date < df_specific_match.Date]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            Avg_score_Home = np.mean(df_day_before_SpecMat.FTHG) if np.mean(df_day_before_SpecMat.FTHG) !=0 else np.nan
            Avg_score_Away = np.mean(df_day_before_SpecMat.FTAG) if np.mean(df_day_before_SpecMat.FTAG) !=0 else np.nan

            Avg_conced_Home = Avg_score_Away
            Avg_conced_Away = Avg_score_Home

            # Home Team offensive strength
            Avg_home_goals_scored_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTHG.mean()
            Off_strength_HT = Avg_home_goals_scored_by_HT / Avg_score_Home

            # Away Team defensive strength
            Avg_away_goals_conceded_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTHG.mean()
            Def_strength_AT = Avg_away_goals_conceded_by_AT / Avg_conced_Away

            List_off_strength_HT.append(Off_strength_HT)
            List_def_strength_AT.append(Def_strength_AT)

            # Away Team offensive strength
            Avg_away_goals_scored_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTAG.mean()
            Off_strength_AT = Avg_away_goals_scored_by_AT / Avg_score_Away

            # Home Team defensive strength
            Avg_home_goals_conceded_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTAG.mean()
            Def_strength_HT = Avg_home_goals_conceded_by_HT / Avg_conced_Home

            List_off_strength_AT.append(Off_strength_AT)
            List_def_strength_HT.append(Def_strength_HT)

        playing_stat['Off_strength_HT'] = List_off_strength_HT
        playing_stat['Def_strength_HT'] = List_def_strength_HT
        playing_stat['Off_strength_AT'] = List_off_strength_AT
        playing_stat['Def_strength_AT'] = List_def_strength_AT

    for season in seasons:
        season = get_off_def_strength(season)

    """## League position from last season

    In European football leauges every season teams (usually 3 but it depends) can get promotion to the highest league and other teams can fall out. This way we get different teams every seasons. So to get league postion from last season I decided to assign to new teams a specific postion:
    * number of teams in the highest league + new team's league postion from the second highest league
    * for example in 2020/21 season Lorient was new team and they have 1st postion in 2019/20 season in the second highest league so I assign number 21
    """

    Standings = pd.read_csv('league_tables/Ligue1_table_rank.csv', sep=';') # This table was created manually 
    Standings.set_index(['Team'], inplace=True)

    # Add variables with last league postion
    def get_last(playing_stat, Standings, year):
        HomeTeamLP = []
        AwayTeamLP = []
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam
            HomeTeamLP.append(Standings.loc[ht][year])
            AwayTeamLP.append(Standings.loc[at][year])
        playing_stat['HomeTeamLP'] = HomeTeamLP
        playing_stat['AwayTeamLP'] = AwayTeamLP
        return playing_stat

    Ligue1_11_12_data = get_last(Ligue1_11_12_data, Standings, '2010/11')
    Ligue1_12_13_data = get_last(Ligue1_12_13_data, Standings, '2011/12')
    Ligue1_13_14_data = get_last(Ligue1_13_14_data, Standings, '2012/13')
    Ligue1_14_15_data = get_last(Ligue1_14_15_data, Standings, '2013/14')
    Ligue1_15_16_data = get_last(Ligue1_15_16_data, Standings, '2014/15')
    Ligue1_16_17_data = get_last(Ligue1_16_17_data, Standings, '2015/16')
    Ligue1_17_18_data = get_last(Ligue1_17_18_data, Standings, '2016/17')
    Ligue1_18_19_data = get_last(Ligue1_18_19_data, Standings, '2017/18')
    Ligue1_19_20_data = get_last(Ligue1_19_20_data, Standings, '2018/19')
    Ligue1_20_21_data = get_last(Ligue1_20_21_data, Standings, '2019/20')

    # adding binary variable if specific team is a new team in specific season
    def get_newTeam(playing_stat):
        
        teams_number = len(playing_stat.HomeTeam.unique())
        
        playing_stat['Is_HT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.HomeTeamLP]
        playing_stat['Is_AT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.AwayTeamLP]
        
        return playing_stat

    # Data was only for the last 10 completed season so I cant use every historical dataset
    for season in [ Ligue1_11_12_data,
                Ligue1_12_13_data, 
                Ligue1_13_14_data,
                Ligue1_14_15_data, 
                Ligue1_15_16_data, 
                Ligue1_16_17_data, 
                Ligue1_17_18_data, 
                Ligue1_18_19_data, 
                Ligue1_19_20_data, 
                Ligue1_20_21_data]:
        season = get_newTeam(season)

    # Checking if everything is correct
    for season in [ Ligue1_11_12_data,
                Ligue1_12_13_data, 
                Ligue1_13_14_data,
                Ligue1_14_15_data, 
                Ligue1_15_16_data, 
                Ligue1_16_17_data, 
                Ligue1_17_18_data, 
                Ligue1_18_19_data, 
                Ligue1_19_20_data, 
                Ligue1_20_21_data]:
        print(season[season.HomeTeamLP > 20].HomeTeam.unique())

    """## Team data for every season from transfermarkt
    * Average age
    * Number of foreigners
    * Average market value in milions of Euro
    * Total market value in milions of Euro

    Example data from 2016/17 season:
    https://www.transfermarkt.com/ligue-1/startseite/wettbewerb/FR1/plus/?saison_id=2016
    """

    Ligue1_budget = pd.read_csv("budget_data/Ligue1_transfermaket_data.csv", sep=';')
    Ligue1_budget.set_index(['Club'], inplace=True)
    Ligue1_budget

    """Changing string variables for float"""

    Ligue1_budget['Avg_age'].replace(',', '.', regex=True, inplace=True)
    Ligue1_budget['Avg_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)
    Ligue1_budget['Total_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)

    Ligue1_budget.info()

    Ligue1_budget.Avg_age = Ligue1_budget.Avg_age.astype(float)
    Ligue1_budget.Avg_market_value_MLN_EUR = Ligue1_budget.Avg_market_value_MLN_EUR.astype(float)
    Ligue1_budget.Total_market_value_MLN_EUR = Ligue1_budget.Total_market_value_MLN_EUR.astype(float)

    # function summarizing and adding variables
    def get_transfermarkt_data(season_data, Budget):
        
        HomeTeam_TM_avg_age = []
        HomeTeam_TM_foreigners_number = []
        HomeTeam_TM_avg_market_value_MLN_EUR = []
        HomeTeam_TM_total_market_value_MLN_EUR = []
        
        AwayTeam_TM_avg_age = []
        AwayTeam_TM_foreigners_number = []
        AwayTeam_TM_avg_market_value_MLN_EUR = []
        AwayTeam_TM_total_market_value_MLN_EUR = []
        
        for season in list(season_data.Season.unique()):
            
            Ligue1_budget_season = Budget[Budget.Season == season]
            
            for i in range(len(season_data)):
                
                home_team = season_data[season_data.Season == season].iloc[i].HomeTeam
                away_team = season_data[season_data.Season == season].iloc[i].AwayTeam
                
                HomeTeam_TM_avg_age.append(Ligue1_budget_season.loc[home_team]['Avg_age'])
                HomeTeam_TM_foreigners_number.append(Ligue1_budget_season.loc[home_team]['Foreigners'])
                HomeTeam_TM_avg_market_value_MLN_EUR.append(Ligue1_budget_season.loc[home_team]['Avg_market_value_MLN_EUR'])
                HomeTeam_TM_total_market_value_MLN_EUR.append(Ligue1_budget_season.loc[home_team]['Total_market_value_MLN_EUR'])
                
                AwayTeam_TM_avg_age.append(Ligue1_budget_season.loc[away_team]['Avg_age'])
                AwayTeam_TM_foreigners_number.append(Ligue1_budget_season.loc[away_team]['Foreigners'])
                AwayTeam_TM_avg_market_value_MLN_EUR.append(Ligue1_budget_season.loc[away_team]['Avg_market_value_MLN_EUR'])
                AwayTeam_TM_total_market_value_MLN_EUR.append(Ligue1_budget_season.loc[away_team]['Total_market_value_MLN_EUR'])
                
                
        season_data['HomeTeam_TM_avg_age'] = HomeTeam_TM_avg_age
        season_data['HomeTeam_TM_foreigners_number'] = HomeTeam_TM_foreigners_number
        season_data['HomeTeam_TM_avg_market_value_MLN_EUR'] = HomeTeam_TM_avg_market_value_MLN_EUR
        season_data['HomeTeam_TM_total_market_value_MLN_EUR'] = HomeTeam_TM_total_market_value_MLN_EUR
        
        season_data['AwayTeam_TM_avg_age'] = AwayTeam_TM_avg_age
        season_data['AwayTeam_TM_foreigners_number'] = AwayTeam_TM_foreigners_number
        season_data['AwayTeam_TM_avg_market_value_MLN_EUR'] = AwayTeam_TM_avg_market_value_MLN_EUR
        season_data['AwayTeam_TM_total_market_value_MLN_EUR'] = AwayTeam_TM_total_market_value_MLN_EUR
        
        return season_data

    for season in [Ligue1_11_12_data, Ligue1_12_13_data, Ligue1_13_14_data, Ligue1_14_15_data, Ligue1_15_16_data, 
                  Ligue1_16_17_data, Ligue1_17_18_data, Ligue1_18_19_data, Ligue1_19_20_data, Ligue1_20_21_data]:
        
        get_transfermarkt_data(season, Ligue1_budget)

    """## Data concatenating"""

    Ligue1_data = pd.concat([Ligue1_00_01_data, 
                          Ligue1_01_02_data, 
                          Ligue1_02_03_data, 
                          Ligue1_03_04_data, 
                          Ligue1_04_05_data, 
                          Ligue1_05_06_data, 
                          Ligue1_06_07_data, 
                          Ligue1_07_08_data, 
                          Ligue1_08_09_data, 
                          Ligue1_09_10_data, 
                          Ligue1_10_11_data, 
                          Ligue1_11_12_data,
                          Ligue1_12_13_data, 
                          Ligue1_13_14_data,
                          Ligue1_14_15_data, 
                          Ligue1_15_16_data, 
                          Ligue1_16_17_data, 
                          Ligue1_17_18_data, 
                          Ligue1_18_19_data, 
                          Ligue1_19_20_data, 
                          Ligue1_20_21_data], ignore_index=True)

    # deleting unnecessary data
    del Ligue1_00_01_data, Ligue1_01_02_data, Ligue1_02_03_data, Ligue1_03_04_data, Ligue1_04_05_data, Ligue1_05_06_data
    del Ligue1_06_07_data, Ligue1_07_08_data, Ligue1_08_09_data, Ligue1_09_10_data
    del Ligue1_20_21_data_all_col
    del season, seasons

    """## Calcuating betting odds
    * 1x (Home win or Draw) = 100 / (100/Home_win_odds + 100/Draw_odds)
    * x2 (Away win or Draw) = 100 / (100/Away_win_odds + 100/Draw_odds)
    """

    def get_more_odds(playing_stat):
        x1 = []
        x2 = []
        
        for i, row in playing_stat.iterrows():
            b365_H = playing_stat.iloc[i]['B365H']
            b365_D = playing_stat.iloc[i]['B365D']
            b365_A = playing_stat.iloc[i]['B365A']
            
            odd_H = 100/b365_H
            odd_D = 100/b365_D
            odd_A = 100/b365_A
            
            b365_H_D = round((100/(odd_H + odd_D)), 2)
            b365_A_D = round((100/(odd_A + odd_D)), 2)
            
            x1.append(b365_H_D)
            x2.append(b365_A_D)
            
        
        playing_stat['B365_HD'] = x1
        playing_stat['B365_AD'] = x2
        
        return playing_stat

    get_more_odds(Ligue1_data)

    """## Add points collect by teams from last 3 and 5 games"""

    # Gets the form points.
    def get_form_points(string):
        sum = 0
        for letter in string:
            sum += get_points(letter) # using other function
        return sum

    # Concatenate variables with last team's results
    Ligue1_data['HT_3match_FormPtsStr'] = Ligue1_data['HM1'] + Ligue1_data['HM2'] + Ligue1_data['HM3']
    Ligue1_data['AT_3match_FormPtsStr'] = Ligue1_data['AM1'] + Ligue1_data['AM2'] + Ligue1_data['AM3']

    Ligue1_data['HT_5match_FormPtsStr'] = Ligue1_data['HM1'] + Ligue1_data['HM2'] + Ligue1_data['HM3'] + Ligue1_data['HM4'] + Ligue1_data['HM5']
    Ligue1_data['AT_5match_FormPtsStr'] = Ligue1_data['AM1'] + Ligue1_data['AM2'] + Ligue1_data['AM3'] + Ligue1_data['AM4'] + Ligue1_data['AM5']

    # Applying function to get points
    Ligue1_data['HT_3match_FormPts'] = Ligue1_data['HT_3match_FormPtsStr'].apply(get_form_points)
    Ligue1_data['AT_3match_FormPts'] = Ligue1_data['AT_3match_FormPtsStr'].apply(get_form_points)

    Ligue1_data['HT_5match_FormPts'] = Ligue1_data['HT_5match_FormPtsStr'].apply(get_form_points)
    Ligue1_data['AT_5match_FormPts'] = Ligue1_data['AT_5match_FormPtsStr'].apply(get_form_points)

    """## Win or loss streaks"""

    # Identify Win/Loss Streaks if any.
    def get_3game_ws(string):
        if string == 'WWW':
            return 1
        else:
            return 0
        
    def get_5game_ws(string):
        if string == 'WWWWW':
            return 1
        else:
            return 0
        
    def get_3game_ls(string):
        if string == 'LLL':
            return 1
        else:
            return 0
        
    def get_5game_ls(string):
        if string == 'LLLLL':
            return 1
        else:
            return 0

    Ligue1_data['HTWinStreak3'] = Ligue1_data['HT_3match_FormPtsStr'].apply(get_3game_ws)
    Ligue1_data['HTWinStreak5'] = Ligue1_data['HT_5match_FormPtsStr'].apply(get_5game_ws)
    Ligue1_data['HTLossStreak3'] = Ligue1_data['HT_3match_FormPtsStr'].apply(get_3game_ls)
    Ligue1_data['HTLossStreak5'] = Ligue1_data['HT_5match_FormPtsStr'].apply(get_5game_ls)

    Ligue1_data['ATWinStreak3'] = Ligue1_data['AT_3match_FormPtsStr'].apply(get_3game_ws)
    Ligue1_data['ATWinStreak5'] = Ligue1_data['AT_5match_FormPtsStr'].apply(get_5game_ws)
    Ligue1_data['ATLossStreak3'] = Ligue1_data['AT_3match_FormPtsStr'].apply(get_3game_ls)
    Ligue1_data['ATLossStreak5'] = Ligue1_data['AT_5match_FormPtsStr'].apply(get_5game_ls)

    """## Head to Head (H2H) variables
    * Getting last 3 head to head match results and change it for points
    """

    def h2h(playing_stat):
        
        playing_stat["H2H_Home"] = ""
        playing_stat["H2H_Away"] = ""

        for i in range(1,len(playing_stat)):
            # get teams from specifc match and all data before this match
            df = playing_stat.iloc[0:i-1,]
            ht = playing_stat.loc[i,"HomeTeam"]
            at = playing_stat.loc[i,"AwayTeam"]
            
            # getting last 3 head to head matches
            dff = df[(((df.HomeTeam == ht) &  (df.AwayTeam == at)) | ((df.HomeTeam == at) &  (df.AwayTeam == ht)))].tail(3)
            
            # using data with minimum 1 head to head match
            if (len(dff) >= 1):
                for j in range(0,len(dff)):
                    # Consider all possible scenarios
                    if ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                        
            # if there was no match beetwen the teams get 'ND'
            else: 
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] = "ND"
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] = "ND"
        return playing_stat


    # Calculating points for Home and Away teams in head to head matches and adding variables
    def get_points_h2h(playing_stat, col):
        
        playing_stat[col+"_pts"] = 0
        
        for i in range(1,len(playing_stat)):
            if (len(playing_stat.iloc[i][col]) == 1):
                if playing_stat.iloc[i][col][0] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][0] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                else:
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            elif (len(playing_stat.iloc[i][col]) == 2):
                for j in range(0,2):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
                    else:
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
            elif (len(playing_stat.iloc[i][col]) == 3):
                for j in range(0,3):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            else:
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] = 0
        
        return playing_stat

    h2h(Ligue1_data)

    get_points_h2h(Ligue1_data, "H2H_Home") 
    get_points_h2h(Ligue1_data, "H2H_Away")

    """## Choose last 10 completed seasons"""

    Ligue1_data = Ligue1_data[Ligue1_data.Season.isin(['2011/12', 
                                             '2012/13', 
                                             '2013/14', 
                                             '2014/15', 
                                             '2015/16', 
                                             '2016/17', 
                                             '2017/18', 
                                             '2018/19', 
                                             '2019/20', 
                                             '2020/21'])]

    Ligue1_data.reset_index(drop=True, inplace=True)
    Ligue1_data.shape

    """## Correct positioning of the columns and changing col names"""

    Ligue1_data_final = Ligue1_data[[
        'Div', 'Season', 'Date', 'HomeTeam', 'AwayTeam','FTHG', 'FTAG', 'FTR', 'HS','AS', 'HST', 'AST',
        'HT_last_matches_number', 'AT_last_matches_number', 'B365H', 'B365D', 'B365A', 'B365_HD', 'B365_AD',
        'HTGS', 'ATGS', 'HTGC', 'ATGC',
        'HTSOT_T','ATSOT_T', 'HTSOT_A', 'ATSOT_A', 
        'HT_Shot_Acc', 'AT_Shot_Acc','HT_Op_Shot_Acc', 'AT_Op_Shot_Acc',
        'HTP', 'ATP', 
        'HT_cum_wins', 'AT_cum_wins', 'HT_cum_losses', 'AT_cum_losses', 'HT_cum_draws', 'AT_cum_draws',
        'Off_strength_HT', 'Def_strength_HT', 'Off_strength_AT', 'Def_strength_AT',
        'HM1', 'AM1', 'HM2', 'AM2', 'HM3', 'AM3', 'HM4', 'AM4', 'HM5', 'AM5',
        'HT_3match_FormPtsStr', 'AT_3match_FormPtsStr', 'HT_5match_FormPtsStr','AT_5match_FormPtsStr', 
        'HT_3match_FormPts', 'AT_3match_FormPts', 'HT_5match_FormPts', 'AT_5match_FormPts',
        'HTWinStreak3','HTWinStreak5', 'HTLossStreak3', 'HTLossStreak5', 
        'ATWinStreak3','ATWinStreak5', 'ATLossStreak3', 'ATLossStreak5',
        'HomeTeamLP', 'AwayTeamLP',
        'Is_HT_newTeam', 'Is_AT_newTeam',
        'H2H_Home', 'H2H_Away', 'H2H_Home_pts','H2H_Away_pts',
        'HomeTeam_TM_avg_age','HomeTeam_TM_foreigners_number', 
        'HomeTeam_TM_avg_market_value_MLN_EUR','HomeTeam_TM_total_market_value_MLN_EUR',
        'AwayTeam_TM_avg_age', 'AwayTeam_TM_foreigners_number', 
        'AwayTeam_TM_avg_market_value_MLN_EUR', 'AwayTeam_TM_total_market_value_MLN_EUR'
                              ]]

    Ligue1_data_final.rename(columns={'HM1': 'HT_last_match1', 'AM1': 'AT_last_match1', 
                                   'HM2': 'HT_last_match2', 'AM2': 'AT_last_match2', 
                                   'HM3': 'HT_last_match3', 'AM3': 'AT_last_match3', 
                                   'HM4': 'HT_last_match4', 'AM4': 'AT_last_match4', 
                                   'HM5': 'HT_last_match5', 'AM5': 'AT_last_match5', 
                                   'HomeTeamLP': 'HT_last_LP', 'AwayTeamLP': 'AT_last_LP', 
                                   #'HTGD': 'HT_goal_diff', 'ATGD': 'AT_goal_diff', 
                                   #'DiffLP': 'Diff_last_LP',
                                   'HTP': 'HT_Pts', 'ATP': 'AT_Pts', 
                                   'HTSOT_T': 'HT_Shots_OT_took','ATSOT_T': 'AT_Shots_OT_took', 
                                   'HTSOT_A': 'HT_Shots_OT_allowed','ATSOT_A': 'AT_Shots_OT_allowed', 
                                   'HT_Shot_Acc': 'HT_Shots_took_Acc', 'AT_Shot_Acc': 'AT_Shots_took_Acc',
                                   'HT_Op_Shot_Acc': 'HT_Shots_allowed_Acc', 'AT_Op_Shot_Acc': 'AT_Shots_allowed_Acc'
                                  }, inplace=True)

    """## Save created dataset"""

    Ligue1_data_final.to_csv(path_or_buf= 'Final_data/Ligue1_data_final.csv', index=False)
```

```python
def SerieA_dataset_creator():    
    # -*- coding: utf-8 -*-
    """SerieA_dataset_creator.ipynb

    Automatically generated by Colaboratory.

    Original file is located at
        https://colab.research.google.com/drive/1pzBRRXgujQQVZhByzZznntn5jZPavfvF

    ## Uploading data and basic operations (add season variable and change the date format)
    """

    import numpy as np
    import pandas as pd

    league_code = 'I1'
    seasons = ["00-01", "01-02", "02-03", "03-04", "04-05", 
              "05-06", "06-07", "07-08", "08-09", "09-10", 
              "10-11", "11-12", "12-13", "13-14", "14-15", 
              "15-16", "16-17", "17-18", "18-19", "19-20", "20-21"]

    urls = []
    for season in seasons:
        x = 'https://www.football-data.co.uk/mmz4281/' + season.replace("-", "") + '/'+league_code + '.csv'
        urls.append(x)

    columns = ['Div', 'Date','HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HS', 'AS', 'HST', 'AST', 'B365H', 'B365D', 'B365A']

    """I want to focus on the last 10 completed seasons (from 2011/12 to 2020/21) but i need more historical data to calculate H2H (Head To Head) variables

    In some data from historical season there was no columns with shots data so I decided to also cut shots data from 00/01 to 10/11 season
    """

    SerieA_00_01_data = pd.read_csv(urls[0])[columns[0:-7]]
    SerieA_01_02_data = pd.read_csv(urls[1])[columns[0:-7]]
    SerieA_02_03_data = pd.read_csv(urls[2])[columns[0:-7]]
    SerieA_03_04_data = pd.read_csv("match_data/I1_03_04.csv", sep=';')[columns[0:-7]] # Error Tokenizing Data with urls upload
    SerieA_04_05_data = pd.read_csv("match_data/I1_04_05.csv", sep=';')[columns[0:-7]] # Error Tokenizing Data with urls upload
    SerieA_05_06_data = pd.read_csv(urls[5])[columns[0:-7]]
    SerieA_06_07_data = pd.read_csv(urls[6])[columns[0:-7]]
    SerieA_07_08_data = pd.read_csv(urls[7])[columns[0:-7]]
    SerieA_08_09_data = pd.read_csv(urls[8])[columns[0:-7]]
    SerieA_09_10_data = pd.read_csv(urls[9])[columns[0:-7]]
    SerieA_10_11_data = pd.read_csv(urls[10])[columns[0:-7]]
    SerieA_11_12_data = pd.read_csv(urls[11])[columns]
    SerieA_12_13_data = pd.read_csv(urls[12])[columns]
    SerieA_13_14_data = pd.read_csv(urls[13])[columns]
    SerieA_14_15_data = pd.read_csv(urls[14])[columns]
    SerieA_15_16_data = pd.read_csv(urls[15])[columns]
    SerieA_16_17_data = pd.read_csv(urls[16])[columns]
    SerieA_17_18_data = pd.read_csv(urls[17])[columns]
    SerieA_18_19_data = pd.read_csv(urls[18])[columns]
    SerieA_19_20_data = pd.read_csv(urls[19])[columns]
    SerieA_20_21_data = pd.read_csv(urls[20])[columns]

    SerieA_20_21_data['Season'] = '2020/21'
    SerieA_19_20_data['Season'] = '2019/20'
    SerieA_18_19_data['Season'] = '2018/19'
    SerieA_17_18_data['Season'] = '2017/18'
    SerieA_16_17_data['Season'] = '2016/17'
    SerieA_15_16_data['Season'] = '2015/16'
    SerieA_14_15_data['Season'] = '2014/15'
    SerieA_13_14_data['Season'] = '2013/14'
    SerieA_12_13_data['Season'] = '2012/13'
    SerieA_11_12_data['Season'] = '2011/12'
    SerieA_10_11_data['Season'] = '2010/11'
    SerieA_09_10_data['Season'] = '2009/10'
    SerieA_08_09_data['Season'] = '2008/09'
    SerieA_07_08_data['Season'] = '2007/08'
    SerieA_06_07_data['Season'] = '2006/07'
    SerieA_05_06_data['Season'] = '2005/06'
    SerieA_04_05_data['Season'] = '2004/05'
    SerieA_03_04_data['Season'] = '2003/04'
    SerieA_02_03_data['Season'] = '2002/03'
    SerieA_01_02_data['Season'] = '2001/02'
    SerieA_00_01_data['Season'] = '2000/01'

    seasons = [SerieA_00_01_data, SerieA_01_02_data, SerieA_02_03_data, SerieA_03_04_data, SerieA_04_05_data,
              SerieA_05_06_data, SerieA_06_07_data, SerieA_07_08_data, SerieA_08_09_data, SerieA_09_10_data, 
              SerieA_10_11_data, SerieA_11_12_data, SerieA_12_13_data, SerieA_13_14_data, SerieA_14_15_data, 
              SerieA_15_16_data, SerieA_16_17_data, SerieA_17_18_data, SerieA_18_19_data, SerieA_19_20_data, 
              SerieA_20_21_data]

    # checking if every season was uploaded correctly without NA
    for season in seasons:
        print(season.Season.unique())
        print(season.Date[0])
        season.dropna(inplace=True)
        print(len(season)) # in 00/01, 01/02 and 02/03 seasons there was 18 teams in the league
        print('|----------------------------|')

    """**No relevant data in 2012/13 and 2020/21 seasons!**

    Upload data one more time
    """

    SerieA_00_01_data = pd.read_csv(urls[0])[columns[0:-7]]
    SerieA_01_02_data = pd.read_csv(urls[1])[columns[0:-7]]
    SerieA_02_03_data = pd.read_csv(urls[2])[columns[0:-7]]
    SerieA_03_04_data = pd.read_csv("match_data/I1_03_04.csv", sep=';')[columns[0:-7]]
    SerieA_04_05_data = pd.read_csv("match_data/I1_04_05.csv", sep=';')[columns[0:-7]]
    SerieA_05_06_data = pd.read_csv(urls[5])[columns[0:-7]]
    SerieA_06_07_data = pd.read_csv(urls[6])[columns[0:-7]]
    SerieA_07_08_data = pd.read_csv(urls[7])[columns[0:-7]]
    SerieA_08_09_data = pd.read_csv(urls[8])[columns[0:-7]]
    SerieA_09_10_data = pd.read_csv(urls[9])[columns[0:-7]]
    SerieA_10_11_data = pd.read_csv(urls[10])[columns[0:-7]]
    SerieA_11_12_data = pd.read_csv(urls[11])[columns]
    SerieA_12_13_data = pd.read_csv(urls[12])[columns]
    SerieA_13_14_data = pd.read_csv(urls[13])[columns]
    SerieA_14_15_data = pd.read_csv(urls[14])[columns]
    SerieA_15_16_data = pd.read_csv(urls[15])[columns]
    SerieA_16_17_data = pd.read_csv(urls[16])[columns]
    SerieA_17_18_data = pd.read_csv(urls[17])[columns]
    SerieA_18_19_data = pd.read_csv(urls[18])[columns]
    SerieA_19_20_data = pd.read_csv(urls[19])[columns]
    SerieA_20_21_data = pd.read_csv(urls[20])[columns]

    """#### Season 12/13"""

    SerieA_12_13_data[SerieA_12_13_data.isna().any(axis=1)] # Match Cagliari-Roma was a walkover

    """Fill in the statistics of shots in such a way that they do not disturb the accuracy calculated later"""

    SerieA_12_13_data.loc[34, 'HS'] = 0
    SerieA_12_13_data.loc[34, 'AS'] = 3
    SerieA_12_13_data.loc[34, 'HST'] = 0
    SerieA_12_13_data.loc[34, 'AST'] = 3

    SerieA_12_13_data.loc[34]

    SerieA_12_13_data.dropna()

    """#### Season 2020/21"""

    SerieA_20_21_data[SerieA_20_21_data.isna().any(axis=1)] # No betting odds

    SerieA_20_21_data_all_col = pd.read_csv(urls[20]) # upload data with all columns

    pd.set_option("display.max_columns", None)
    SerieA_20_21_data_all_col[SerieA_20_21_data_all_col.isna().any(axis=1)]

    # Use average betting odds
    odds_35 = SerieA_20_21_data_all_col[(SerieA_20_21_data_all_col.HomeTeam == 'Udinese') & 
                              (SerieA_20_21_data_all_col.AwayTeam == 'Parma')][['AvgH', 'AvgD', 'AvgA']]

    odds_37 = SerieA_20_21_data_all_col[(SerieA_20_21_data_all_col.HomeTeam == 'Verona') & 
                              (SerieA_20_21_data_all_col.AwayTeam == 'Genoa')][['AvgH', 'AvgD', 'AvgA']]

    SerieA_20_21_data.loc[37]

    SerieA_20_21_data.loc[35, 'B365H'] = odds_35.loc[35, 'AvgH']
    SerieA_20_21_data.loc[35, 'B365D'] = odds_35.loc[35, 'AvgD']
    SerieA_20_21_data.loc[35, 'B365A'] = odds_35.loc[35, 'AvgA']

    SerieA_20_21_data.loc[37, 'B365H'] = odds_37.loc[37, 'AvgH']
    SerieA_20_21_data.loc[37, 'B365D'] = odds_37.loc[37, 'AvgD']
    SerieA_20_21_data.loc[37, 'B365A'] = odds_37.loc[37, 'AvgA']

    SerieA_20_21_data[SerieA_20_21_data.isna().any(axis=1)]

    seasons = [SerieA_00_01_data, SerieA_01_02_data, SerieA_02_03_data, SerieA_03_04_data, SerieA_04_05_data,
              SerieA_05_06_data, SerieA_06_07_data, SerieA_07_08_data, SerieA_08_09_data, SerieA_09_10_data, 
              SerieA_10_11_data, SerieA_11_12_data, SerieA_12_13_data, SerieA_13_14_data, SerieA_14_15_data, 
              SerieA_15_16_data, SerieA_16_17_data, SerieA_17_18_data, SerieA_18_19_data, SerieA_19_20_data, 
              SerieA_20_21_data]
    seasons

    # checking if every season was uploaded correctly without NA
    for season in seasons:
        #print(season.Season.unique())
        print(season.Date[0])
        season.dropna(how='all' ,inplace=True)
        print(len(season))
        print('|----------------------------|')

    SerieA_20_21_data['Season'] = '2020/21'
    SerieA_19_20_data['Season'] = '2019/20'
    SerieA_18_19_data['Season'] = '2018/19'
    SerieA_17_18_data['Season'] = '2017/18'
    SerieA_16_17_data['Season'] = '2016/17'
    SerieA_15_16_data['Season'] = '2015/16'
    SerieA_14_15_data['Season'] = '2014/15'
    SerieA_13_14_data['Season'] = '2013/14'
    SerieA_12_13_data['Season'] = '2012/13'
    SerieA_11_12_data['Season'] = '2011/12'
    SerieA_10_11_data['Season'] = '2010/11'
    SerieA_09_10_data['Season'] = '2009/10'
    SerieA_08_09_data['Season'] = '2008/09'
    SerieA_07_08_data['Season'] = '2007/08'
    SerieA_06_07_data['Season'] = '2006/07'
    SerieA_05_06_data['Season'] = '2005/06'
    SerieA_04_05_data['Season'] = '2004/05'
    SerieA_03_04_data['Season'] = '2003/04'
    SerieA_02_03_data['Season'] = '2002/03'
    SerieA_01_02_data['Season'] = '2001/02'
    SerieA_00_01_data['Season'] = '2000/01'

    """Simple checking corrected datasets"""

    SerieA_20_21_data[SerieA_20_21_data.isna().any(axis=1)]

    SerieA_12_13_data[SerieA_12_13_data.isna().any(axis=1)]

    # Change Date Format
    for season in seasons:
        season.Date = pd.to_datetime(season.Date, dayfirst=True)

    """## Adds the number of games played by the team before specific match"""

    def get_last_matches_number(playing_stat):
        
        # Diffrent lists for HomeTeam and AwayTeam because sometimes number of games played before 
        # can be diffrent becasue of postponing the match
        HT_Match_number = []
        AT_Match_number = []
        
        # dataset iteration over each row
        for i, row in playing_stat.iterrows():
            df_nad = playing_stat.iloc[0:i]
            df_specific_match = playing_stat.iloc[i]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            x = len(df_nad[(df_nad.HomeTeam == ht) | (df_nad.AwayTeam == ht)])
            y = len(df_nad[(df_nad.HomeTeam == at) | (df_nad.AwayTeam == at)])

            HT_Match_number.append(x)
            AT_Match_number.append(y)
            
        playing_stat['HT_last_matches_number'] = HT_Match_number
        playing_stat['AT_last_matches_number'] = AT_Match_number
        
        return playing_stat

    # apply function on every season data
    for season in seasons:
        season = get_last_matches_number(season)

    """## Goals scored and conceded before specific match"""

    # Gets the goals scored aggregation arranged by teams and matchweek
    def get_goals_scored(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTGS = playing_stat.iloc[i]['FTHG']
            ATGS = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGS)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGS)
            
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        GoalsScored[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
        return GoalsScored

    # Gets the goals conceded aggregation arranged by teams and matchweek
    def get_goals_conceded(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
    #         print("check {} \n".format(i))
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATGC = playing_stat.iloc[i]['FTHG']
            HTGC = playing_stat.iloc[i]['FTAG']
            teams[playing_stat.iloc[i].HomeTeam].append(HTGC)
            teams[playing_stat.iloc[i].AwayTeam].append(ATGC)
        
        # Create a dataframe for goals scored where rows are teams and cols are matchweek.
        GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        GoalsConceded[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
        return GoalsConceded

    # function summarizing and adding variables
    def get_gss(playing_stat):
        GC = get_goals_conceded(playing_stat)
        GS = get_goals_scored(playing_stat)
       
        HTGS = []
        ATGS = []
        HTGC = []
        ATGC = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTGS.append(GS.loc[ht][ht_match])
            ATGS.append(GS.loc[at][at_match])
            
            HTGC.append(GC.loc[ht][ht_match])
            ATGC.append(GC.loc[at][at_match])
            
            
        playing_stat['HTGS'] = HTGS
        playing_stat['ATGS'] = ATGS
        playing_stat['HTGC'] = HTGC
        playing_stat['ATGC'] = ATGC
        
        return playing_stat

    for season in seasons:
        season = get_gss(season)

    """## Shots on target took and allowed before specific match"""

    # Gets the shots on target took aggregation arranged by teams and matchweek
    def get_shots_on_target_took(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            HTSOT = playing_stat.iloc[i]['HST']
            ATSOT = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT)
            
        # Create a dataframe for shots on target took where rows are teams and cols are matchweek.
        ShotsOT = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        ShotsOT[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            ShotsOT[i] = ShotsOT[i] + ShotsOT[i-1]
        return ShotsOT

    # Gets the shots on target allowed aggregation arranged by teams and matchweek
    def get_shots_on_target_allowed(playing_stat):
        # Create a dictionary with team names as keys
        teams = {}
        for i in playing_stat.groupby('HomeTeam').mean().T.columns:
            teams[i] = []
        
        # the value corresponding to keys is a list containing the match location.
        for i in range(len(playing_stat)):
            ATSOT_A = playing_stat.iloc[i]['HST']
            HTSOT_A = playing_stat.iloc[i]['AST']
            teams[playing_stat.iloc[i].HomeTeam].append(HTSOT_A)
            teams[playing_stat.iloc[i].AwayTeam].append(ATSOT_A)
        
        # Create a dataframe for shots on target allowed where rows are teams and cols are matchweek.
        ShotsOT_Allowed = pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        ShotsOT_Allowed[0] = 0
        # Aggregate to get uptil that point
        for i in range(2,len(teams.keys())*2-1):
            ShotsOT_Allowed[i] = ShotsOT_Allowed[i] + ShotsOT_Allowed[i-1]
        return ShotsOT_Allowed

    # function summarizing and adding variables
    def get_sot(playing_stat):
        SOT_T = get_shots_on_target_took(playing_stat)
        SOT_A = get_shots_on_target_allowed(playing_stat)
       
        HTSOT_T = []
        ATSOT_T = []
        HTSOT_A = []
        ATSOT_A = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HTSOT_T.append(SOT_T.loc[ht][ht_match])
            ATSOT_T.append(SOT_T.loc[at][at_match])
            
            HTSOT_A.append(SOT_A.loc[ht][ht_match])
            ATSOT_A.append(SOT_A.loc[at][at_match])
            
            
        playing_stat['HTSOT_T'] = HTSOT_T
        playing_stat['ATSOT_T'] = ATSOT_T
        playing_stat['HTSOT_A'] = HTSOT_A
        playing_stat['ATSOT_A'] = ATSOT_A
        
        return playing_stat

    # Datasets with shots data
    for season in [ SerieA_11_12_data,
                SerieA_12_13_data, 
                SerieA_13_14_data,
                SerieA_14_15_data, 
                SerieA_15_16_data, 
                SerieA_16_17_data, 
                SerieA_17_18_data, 
                SerieA_18_19_data, 
                SerieA_19_20_data, 
                SerieA_20_21_data]:
        season = get_sot(season)

    # Get shots accuracy
    def get_shots_acc(playing_stat):
        
        HT_Shot_Acc = playing_stat.HTGS / playing_stat.HTSOT_T # HomeTeam shots took accuracy (Goals scoring accuracy)
        AT_Shot_Acc = playing_stat.ATGS / playing_stat.ATSOT_T # AwayTeam shots took accuracy (Goals scoring accuracy)
        
        HT_Op_Shot_Acc = playing_stat.HTGC / playing_stat.HTSOT_A # HomeTeam shots allowed accuracy (Goals conceded accuracy)
        AT_Op_Shot_Acc = playing_stat.ATGC / playing_stat.ATSOT_A # AwayTeam shots allowed accuracy (Goals conceded accuracy)
        
        playing_stat['HT_Shot_Acc'] = HT_Shot_Acc
        playing_stat['AT_Shot_Acc'] = AT_Shot_Acc
        playing_stat['HT_Op_Shot_Acc'] = HT_Op_Shot_Acc
        playing_stat['AT_Op_Shot_Acc'] = AT_Op_Shot_Acc
        
        return playing_stat

    # Datasets with shots data
    for season in [ SerieA_11_12_data,
                SerieA_12_13_data, 
                SerieA_13_14_data,
                SerieA_14_15_data, 
                SerieA_15_16_data, 
                SerieA_16_17_data, 
                SerieA_17_18_data, 
                SerieA_18_19_data, 
                SerieA_19_20_data, 
                SerieA_20_21_data]:
        season = get_shots_acc(season)

    """## Cumulative points before specific match"""

    # Function for getting specific number of points depends on the match result
    def get_points(result):
        if result == 'W':
            return 3
        elif result == 'D':
            return 1
        else:
            return 0

    # Get match results for every team in every match week
    def get_matchres(season):
        # Create a dictionary with team names as keys
        teams = {}
        for i in season.groupby('HomeTeam').mean().T.columns:
            teams[i] = []

        # the value corresponding to keys is a list containing the match result
        for i in range(len(season)):
            if season.iloc[i].FTR == 'H':
                teams[season.iloc[i].HomeTeam].append('W')
                teams[season.iloc[i].AwayTeam].append('L')
            elif season.iloc[i].FTR == 'A':
                teams[season.iloc[i].AwayTeam].append('W')
                teams[season.iloc[i].HomeTeam].append('L')
            else:
                teams[season.iloc[i].AwayTeam].append('D')
                teams[season.iloc[i].HomeTeam].append('D')

                
        return pd.DataFrame(data=teams, index = [i for i in range(1,len(teams.keys())*2-1)]).T
        
    # Get cumulative points connected with match results
    def get_cuml_points(matchres):
        
        matchres_points = matchres.applymap(get_points)
        for i in range(2,len(matchres)*2-2):
            matchres_points[i] = matchres_points[i] + matchres_points[i-1]
            
        matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(len(matchres))])
        return matchres_points

    # function summarizing and adding variables
    def get_agg_points(season):
        matchres = get_matchres(season)
        cum_pts = get_cuml_points(matchres)
        HTP = []
        ATP = []

        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HTP.append(cum_pts.loc[ht][ht_match])
            ATP.append(cum_pts.loc[at][at_match])
                
        season['HTP'] = HTP
        season['ATP'] = ATP
        return season

    for season in seasons:
        season = get_agg_points(season)

    """## Current teams form with match results for 5 last games"""

    # get and add to dataset last 5 match results
    def add_form(playing_stat,num):
        form = get_matchres(playing_stat)
        
        for number in range(5+1): # I am interesing to 5 last games
            form[-number] = 'ND'  # No data
        
        HT_form = []
        AT_form = []
        
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam

            ht_match = playing_stat.iloc[i].HT_last_matches_number
            at_match = playing_stat.iloc[i].AT_last_matches_number

            HT_form.append(form.loc[ht][ht_match - (num - 1)])
            AT_form.append(form.loc[at][at_match - (num - 1)])

        # every last match will have different variable 
        playing_stat['HM' + str(num)] = HT_form                 
        playing_stat['AM' + str(num)] = AT_form
        
        return playing_stat

    # function summarizing and adding variables
    def add_form_df(playing_statistics):
        playing_statistics = add_form(playing_statistics,1)
        playing_statistics = add_form(playing_statistics,2)
        playing_statistics = add_form(playing_statistics,3)
        playing_statistics = add_form(playing_statistics,4)
        playing_statistics = add_form(playing_statistics,5)
        return playing_statistics

    for season in seasons:
        season = add_form_df(season)

    """## Cumulative number of wins, losses nad draws before specific match"""

    # Binary variables connected with match results
    def get_losses(result):
        if result == 'L':
            return 1
        elif result == 'D' or  result == 'W':
            return 0
        else:
            return np.nan     

    def get_wins(result):
        if result == 'W':
            return 1
        elif result == 'L' or  result == 'D':
            return 0
        else:
            return np.nan 

    def get_draws(result):
        if result == 'D':
            return 1
        elif result == 'L' or  result == 'W':
            return 0
        else:
            return np.nan 
        
    # Get cumulative results during specific season
    def get_agg_results(season):
        matchres = get_matchres(season)
        cum_wins = matchres.applymap(get_wins).cumsum(axis=1)
        cum_losses = matchres.applymap(get_losses).cumsum(axis=1)
        cum_draws = matchres.applymap(get_draws).cumsum(axis=1)
        
        cum_wins[0] = 0
        cum_losses[0] = 0
        cum_draws[0] = 0
        
        HT_cum_wins = []
        AT_cum_wins = []
        
        HT_cum_losses = []
        AT_cum_losses = []
        
        HT_cum_draws = []
        AT_cum_draws = []
        
        for i in range(len(season)):
            ht = season.iloc[i].HomeTeam
            at = season.iloc[i].AwayTeam
            
            ht_match = season.iloc[i].HT_last_matches_number
            at_match = season.iloc[i].AT_last_matches_number
            
            HT_cum_wins.append(cum_wins.loc[ht][ht_match])
            AT_cum_wins.append(cum_wins.loc[at][at_match])
            
            HT_cum_losses.append(cum_losses.loc[ht][ht_match])
            AT_cum_losses.append(cum_losses.loc[at][at_match])
            
            HT_cum_draws.append(cum_draws.loc[ht][ht_match])
            AT_cum_draws.append(cum_draws.loc[at][at_match])
                
        season['HT_cum_wins'] = HT_cum_wins
        season['AT_cum_wins'] = AT_cum_wins
        
        season['HT_cum_losses'] = HT_cum_losses
        season['AT_cum_losses'] = AT_cum_losses
        
        season['HT_cum_draws'] = HT_cum_draws
        season['AT_cum_draws'] = AT_cum_draws
        
        return season

    for season in seasons:
        season = get_agg_results(season)

    """## Offensive and Defensive strength

    The entire calculation process is presented in the link below:
    * [Calcuation process](https://help.smarkets.com/hc/en-gb/articles/115001457989-How-to-calculate-Poisson-distribution-for-football-betting)
    """

    def get_off_def_strength(playing_stat):

        List_off_strength_HT = []
        List_def_strength_HT = []
        List_off_strength_AT = []
        List_def_strength_AT = []

        for i in range(len(playing_stat)):
            df_specific_match = playing_stat.iloc[i]
            df_day_before_SpecMat = playing_stat[playing_stat.Date < df_specific_match.Date]

            ht = df_specific_match.HomeTeam
            at = df_specific_match.AwayTeam

            Avg_score_Home = np.mean(df_day_before_SpecMat.FTHG) if np.mean(df_day_before_SpecMat.FTHG) !=0 else np.nan
            Avg_score_Away = np.mean(df_day_before_SpecMat.FTAG) if np.mean(df_day_before_SpecMat.FTAG) !=0 else np.nan

            Avg_conced_Home = Avg_score_Away
            Avg_conced_Away = Avg_score_Home

            # Home Team offensive strength
            Avg_home_goals_scored_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTHG.mean()
            Off_strength_HT = Avg_home_goals_scored_by_HT / Avg_score_Home

            # Away Team defensive strength
            Avg_away_goals_conceded_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTHG.mean()
            Def_strength_AT = Avg_away_goals_conceded_by_AT / Avg_conced_Away

            List_off_strength_HT.append(Off_strength_HT)
            List_def_strength_AT.append(Def_strength_AT)

            # Away Team offensive strength
            Avg_away_goals_scored_by_AT = df_day_before_SpecMat[(df_day_before_SpecMat.AwayTeam == at)].FTAG.mean()
            Off_strength_AT = Avg_away_goals_scored_by_AT / Avg_score_Away

            # Home Team defensive strength
            Avg_home_goals_conceded_by_HT = df_day_before_SpecMat[(df_day_before_SpecMat.HomeTeam == ht)].FTAG.mean()
            Def_strength_HT = Avg_home_goals_conceded_by_HT / Avg_conced_Home

            List_off_strength_AT.append(Off_strength_AT)
            List_def_strength_HT.append(Def_strength_HT)

        playing_stat['Off_strength_HT'] = List_off_strength_HT
        playing_stat['Def_strength_HT'] = List_def_strength_HT
        playing_stat['Off_strength_AT'] = List_off_strength_AT
        playing_stat['Def_strength_AT'] = List_def_strength_AT

    for season in seasons:
        season = get_off_def_strength(season)

    """## League position from last season

    In European football leauges every season teams (usually 3 but it depends) can get promotion to the highest league and other teams can fall out. This way we get different teams every seasons. So to get league postion from last season I decided to assign to new teams a specific postion:
    * number of teams in the highest league + new team's league postion from the second highest league
    * for example in 2020/21 season Benevento was new team and they have 1st postion in 2019/20 season in the second highest league so I assign number 21
    """

    Standings = pd.read_csv('league_tables/SerieA_table_rank.csv', sep=';') # This table was created manually 
    Standings.set_index(['Team'], inplace=True)

    # Add variables with last league postion
    def get_last(playing_stat, Standings, year):
        HomeTeamLP = []
        AwayTeamLP = []
        for i in range(len(playing_stat)):
            ht = playing_stat.iloc[i].HomeTeam
            at = playing_stat.iloc[i].AwayTeam
            HomeTeamLP.append(Standings.loc[ht][year])
            AwayTeamLP.append(Standings.loc[at][year])
        playing_stat['HomeTeamLP'] = HomeTeamLP
        playing_stat['AwayTeamLP'] = AwayTeamLP
        return playing_stat

    SerieA_11_12_data = get_last(SerieA_11_12_data, Standings, '2010/11')
    SerieA_12_13_data = get_last(SerieA_12_13_data, Standings, '2011/12')
    SerieA_13_14_data = get_last(SerieA_13_14_data, Standings, '2012/13')
    SerieA_14_15_data = get_last(SerieA_14_15_data, Standings, '2013/14')
    SerieA_15_16_data = get_last(SerieA_15_16_data, Standings, '2014/15')
    SerieA_16_17_data = get_last(SerieA_16_17_data, Standings, '2015/16')
    SerieA_17_18_data = get_last(SerieA_17_18_data, Standings, '2016/17')
    SerieA_18_19_data = get_last(SerieA_18_19_data, Standings, '2017/18')
    SerieA_19_20_data = get_last(SerieA_19_20_data, Standings, '2018/19')
    SerieA_20_21_data = get_last(SerieA_20_21_data, Standings, '2019/20')

    # adding binary variable if specific team is a new team in specific season
    def get_newTeam(playing_stat):
        
        teams_number = len(playing_stat.HomeTeam.unique())
        
        playing_stat['Is_HT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.HomeTeamLP]
        playing_stat['Is_AT_newTeam'] = [1 if x > teams_number else 0 for x in playing_stat.AwayTeamLP]
        
        return playing_stat

    # Data was only for the last 10 completed season so I cant use every historical dataset
    for season in [ SerieA_11_12_data,
                SerieA_12_13_data, 
                SerieA_13_14_data,
                SerieA_14_15_data, 
                SerieA_15_16_data, 
                SerieA_16_17_data, 
                SerieA_17_18_data, 
                SerieA_18_19_data, 
                SerieA_19_20_data, 
                SerieA_20_21_data]:
        season = get_newTeam(season)

    # Checking if everything is correct
    for season in [ SerieA_11_12_data,
                SerieA_12_13_data, 
                SerieA_13_14_data,
                SerieA_14_15_data, 
                SerieA_15_16_data, 
                SerieA_16_17_data, 
                SerieA_17_18_data, 
                SerieA_18_19_data, 
                SerieA_19_20_data, 
                SerieA_20_21_data]:
        print(season[season.HomeTeamLP > 20].HomeTeam.unique())

    """## Team data for every season from transfermarkt
    * Average age
    * Number of foreigners
    * Average market value in milions of Euro
    * Total market value in milions of Euro

    Example data from 2016/17 season:
    https://www.transfermarkt.com/serie-a/startseite/wettbewerb/IT1/plus/?saison_id=2016
    """

    SerieA_budget = pd.read_csv("budget_data/SerieA_transfermaket_data.csv", sep=';')
    SerieA_budget.set_index(['Club'], inplace=True)
    SerieA_budget

    """Changing string variables for float"""

    SerieA_budget['Avg_age'].replace(',', '.', regex=True, inplace=True)
    SerieA_budget['Avg_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)
    SerieA_budget['Total_market_value_MLN_EUR'].replace(',', '.', regex=True, inplace=True)

    SerieA_budget.info()

    SerieA_budget.Avg_age = SerieA_budget.Avg_age.astype(float)
    SerieA_budget.Avg_market_value_MLN_EUR = SerieA_budget.Avg_market_value_MLN_EUR.astype(float)
    SerieA_budget.Total_market_value_MLN_EUR = SerieA_budget.Total_market_value_MLN_EUR.astype(float)

    # function summarizing and adding variables
    def get_transfermarkt_data(season_data, Budget):
        
        HomeTeam_TM_avg_age = []
        HomeTeam_TM_foreigners_number = []
        HomeTeam_TM_avg_market_value_MLN_EUR = []
        HomeTeam_TM_total_market_value_MLN_EUR = []
        
        AwayTeam_TM_avg_age = []
        AwayTeam_TM_foreigners_number = []
        AwayTeam_TM_avg_market_value_MLN_EUR = []
        AwayTeam_TM_total_market_value_MLN_EUR = []
        
        for season in list(season_data.Season.unique()):
            
            SerieA_budget_season = Budget[Budget.Season == season]
            
            for i in range(len(season_data)):
                
                home_team = season_data[season_data.Season == season].iloc[i].HomeTeam
                away_team = season_data[season_data.Season == season].iloc[i].AwayTeam
                
                HomeTeam_TM_avg_age.append(SerieA_budget_season.loc[home_team]['Avg_age'])
                HomeTeam_TM_foreigners_number.append(SerieA_budget_season.loc[home_team]['Foreigners'])
                HomeTeam_TM_avg_market_value_MLN_EUR.append(SerieA_budget_season.loc[home_team]['Avg_market_value_MLN_EUR'])
                HomeTeam_TM_total_market_value_MLN_EUR.append(SerieA_budget_season.loc[home_team]['Total_market_value_MLN_EUR'])
                
                AwayTeam_TM_avg_age.append(SerieA_budget_season.loc[away_team]['Avg_age'])
                AwayTeam_TM_foreigners_number.append(SerieA_budget_season.loc[away_team]['Foreigners'])
                AwayTeam_TM_avg_market_value_MLN_EUR.append(SerieA_budget_season.loc[away_team]['Avg_market_value_MLN_EUR'])
                AwayTeam_TM_total_market_value_MLN_EUR.append(SerieA_budget_season.loc[away_team]['Total_market_value_MLN_EUR'])
                
                
        season_data['HomeTeam_TM_avg_age'] = HomeTeam_TM_avg_age
        season_data['HomeTeam_TM_foreigners_number'] = HomeTeam_TM_foreigners_number
        season_data['HomeTeam_TM_avg_market_value_MLN_EUR'] = HomeTeam_TM_avg_market_value_MLN_EUR
        season_data['HomeTeam_TM_total_market_value_MLN_EUR'] = HomeTeam_TM_total_market_value_MLN_EUR
        
        season_data['AwayTeam_TM_avg_age'] = AwayTeam_TM_avg_age
        season_data['AwayTeam_TM_foreigners_number'] = AwayTeam_TM_foreigners_number
        season_data['AwayTeam_TM_avg_market_value_MLN_EUR'] = AwayTeam_TM_avg_market_value_MLN_EUR
        season_data['AwayTeam_TM_total_market_value_MLN_EUR'] = AwayTeam_TM_total_market_value_MLN_EUR
        
        return season_data

    for season in [SerieA_11_12_data, SerieA_12_13_data, SerieA_13_14_data, SerieA_14_15_data, SerieA_15_16_data, 
                  SerieA_16_17_data, SerieA_17_18_data, SerieA_18_19_data, SerieA_19_20_data, SerieA_20_21_data]:
        
        get_transfermarkt_data(season, SerieA_budget)

    """## Data concatenating"""

    SerieA_data = pd.concat([SerieA_00_01_data, 
                          SerieA_01_02_data, 
                          SerieA_02_03_data, 
                          SerieA_03_04_data, 
                          SerieA_04_05_data, 
                          SerieA_05_06_data, 
                          SerieA_06_07_data, 
                          SerieA_07_08_data, 
                          SerieA_08_09_data, 
                          SerieA_09_10_data, 
                          SerieA_10_11_data, 
                          SerieA_11_12_data,
                          SerieA_12_13_data, 
                          SerieA_13_14_data,
                          SerieA_14_15_data, 
                          SerieA_15_16_data, 
                          SerieA_16_17_data, 
                          SerieA_17_18_data, 
                          SerieA_18_19_data, 
                          SerieA_19_20_data, 
                          SerieA_20_21_data], ignore_index=True)

    # deleting unnecessary data
    del SerieA_00_01_data, SerieA_01_02_data, SerieA_02_03_data, SerieA_03_04_data, SerieA_04_05_data, SerieA_05_06_data
    del SerieA_06_07_data, SerieA_07_08_data, SerieA_08_09_data, SerieA_09_10_data
    del SerieA_20_21_data_all_col
    del season, seasons

    """## Calcuating betting odds
    * 1x (Home win or Draw) = 100 / (100/Home_win_odds + 100/Draw_odds)
    * x2 (Away win or Draw) = 100 / (100/Away_win_odds + 100/Draw_odds)
    """

    def get_more_odds(playing_stat):
        x1 = []
        x2 = []
        
        for i, row in playing_stat.iterrows():
            b365_H = playing_stat.iloc[i]['B365H']
            b365_D = playing_stat.iloc[i]['B365D']
            b365_A = playing_stat.iloc[i]['B365A']
            
            odd_H = 100/b365_H
            odd_D = 100/b365_D
            odd_A = 100/b365_A
            
            b365_H_D = round((100/(odd_H + odd_D)), 2)
            b365_A_D = round((100/(odd_A + odd_D)), 2)
            
            x1.append(b365_H_D)
            x2.append(b365_A_D)
            
        
        playing_stat['B365_HD'] = x1
        playing_stat['B365_AD'] = x2
        
        return playing_stat

    get_more_odds(SerieA_data)

    """## Add points collect by teams from last 3 and 5 games"""

    # Gets the form points.
    def get_form_points(string):
        sum = 0
        for letter in string:
            sum += get_points(letter) # using other function
        return sum

    # Concatenate variables with last team's results
    SerieA_data['HT_3match_FormPtsStr'] = SerieA_data['HM1'] + SerieA_data['HM2'] + SerieA_data['HM3']
    SerieA_data['AT_3match_FormPtsStr'] = SerieA_data['AM1'] + SerieA_data['AM2'] + SerieA_data['AM3']

    SerieA_data['HT_5match_FormPtsStr'] = SerieA_data['HM1'] + SerieA_data['HM2'] + SerieA_data['HM3'] + SerieA_data['HM4'] + SerieA_data['HM5']
    SerieA_data['AT_5match_FormPtsStr'] = SerieA_data['AM1'] + SerieA_data['AM2'] + SerieA_data['AM3'] + SerieA_data['AM4'] + SerieA_data['AM5']

    # Applying function to get points
    SerieA_data['HT_3match_FormPts'] = SerieA_data['HT_3match_FormPtsStr'].apply(get_form_points)
    SerieA_data['AT_3match_FormPts'] = SerieA_data['AT_3match_FormPtsStr'].apply(get_form_points)

    SerieA_data['HT_5match_FormPts'] = SerieA_data['HT_5match_FormPtsStr'].apply(get_form_points)
    SerieA_data['AT_5match_FormPts'] = SerieA_data['AT_5match_FormPtsStr'].apply(get_form_points)

    """## Win or loss streaks"""

    # Identify Win/Loss Streaks if any.
    def get_3game_ws(string):
        if string == 'WWW':
            return 1
        else:
            return 0
        
    def get_5game_ws(string):
        if string == 'WWWWW':
            return 1
        else:
            return 0
        
    def get_3game_ls(string):
        if string == 'LLL':
            return 1
        else:
            return 0
        
    def get_5game_ls(string):
        if string == 'LLLLL':
            return 1
        else:
            return 0

    SerieA_data['HTWinStreak3'] = SerieA_data['HT_3match_FormPtsStr'].apply(get_3game_ws)
    SerieA_data['HTWinStreak5'] = SerieA_data['HT_5match_FormPtsStr'].apply(get_5game_ws)
    SerieA_data['HTLossStreak3'] = SerieA_data['HT_3match_FormPtsStr'].apply(get_3game_ls)
    SerieA_data['HTLossStreak5'] = SerieA_data['HT_5match_FormPtsStr'].apply(get_5game_ls)

    SerieA_data['ATWinStreak3'] = SerieA_data['AT_3match_FormPtsStr'].apply(get_3game_ws)
    SerieA_data['ATWinStreak5'] = SerieA_data['AT_5match_FormPtsStr'].apply(get_5game_ws)
    SerieA_data['ATLossStreak3'] = SerieA_data['AT_3match_FormPtsStr'].apply(get_3game_ls)
    SerieA_data['ATLossStreak5'] = SerieA_data['AT_5match_FormPtsStr'].apply(get_5game_ls)

    """## Head to Head (H2H) variables
    * Getting last 3 head to head match results and change it for points
    """

    def h2h(playing_stat):
        
        playing_stat["H2H_Home"] = ""
        playing_stat["H2H_Away"] = ""

        for i in range(1,len(playing_stat)):
            # get teams from specifc match and all data before this match
            df = playing_stat.iloc[0:i-1,]
            ht = playing_stat.loc[i,"HomeTeam"]
            at = playing_stat.loc[i,"AwayTeam"]
            
            # getting last 3 head to head matches
            dff = df[(((df.HomeTeam == ht) &  (df.AwayTeam == at)) | ((df.HomeTeam == at) &  (df.AwayTeam == ht)))].tail(3)
            
            # using data with minimum 1 head to head match
            if (len(dff) >= 1):
                for j in range(0,len(dff)):
                    # Consider all possible scenarios
                    if ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "H")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "L"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "W"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "A")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "W"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "L"
                    elif ((dff.iloc[j].HomeTeam == ht)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                    elif ((dff.iloc[j].HomeTeam == at)  and (dff.iloc[j]["FTR"] == "D")):
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] += "D"
                        playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] += "D"
                        
            # if there was no match beetwen the teams get 'ND'
            else: 
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Home')] = "ND"
                playing_stat.iloc[i, playing_stat.columns.get_loc('H2H_Away')] = "ND"
        return playing_stat


    # Calculating points for Home and Away teams in head to head matches and adding variables
    def get_points_h2h(playing_stat, col):
        
        playing_stat[col+"_pts"] = 0
        
        for i in range(1,len(playing_stat)):
            if (len(playing_stat.iloc[i][col]) == 1):
                if playing_stat.iloc[i][col][0] =='L':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                elif playing_stat.iloc[i][col][0] =='W':
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                else:
                    playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            elif (len(playing_stat.iloc[i][col]) == 2):
                for j in range(0,2):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
                    else:
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
            elif (len(playing_stat.iloc[i][col]) == 3):
                for j in range(0,3):
                    if playing_stat.iloc[i][col][j] =='L':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] += 0
                    elif playing_stat.iloc[i][col][j] =='W':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  3
                    elif playing_stat.iloc[i][col][j] =='D':
                        playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] +=  1
            else:
                playing_stat.iloc[i, playing_stat.columns.get_loc(col+"_pts")] = 0
        
        return playing_stat

    h2h(SerieA_data)

    get_points_h2h(SerieA_data, "H2H_Home") 
    get_points_h2h(SerieA_data, "H2H_Away")

    """## Choose last 10 completed seasons"""

    SerieA_data = SerieA_data[SerieA_data.Season.isin(['2011/12', 
                                             '2012/13', 
                                             '2013/14', 
                                             '2014/15', 
                                             '2015/16', 
                                             '2016/17', 
                                             '2017/18', 
                                             '2018/19', 
                                             '2019/20', 
                                             '2020/21'])]

    SerieA_data.reset_index(drop=True, inplace=True)
    SerieA_data.shape

    """## Correct positioning of the columns and changing col names"""

    SerieA_data_final = SerieA_data[[
        'Div', 'Season', 'Date', 'HomeTeam', 'AwayTeam','FTHG', 'FTAG', 'FTR', 'HS','AS', 'HST', 'AST',
        'HT_last_matches_number', 'AT_last_matches_number', 'B365H', 'B365D', 'B365A', 'B365_HD', 'B365_AD',
        'HTGS', 'ATGS', 'HTGC', 'ATGC',
        'HTSOT_T','ATSOT_T', 'HTSOT_A', 'ATSOT_A', 
        'HT_Shot_Acc', 'AT_Shot_Acc','HT_Op_Shot_Acc', 'AT_Op_Shot_Acc',
        'HTP', 'ATP',
        'HT_cum_wins', 'AT_cum_wins', 'HT_cum_losses', 'AT_cum_losses', 'HT_cum_draws', 'AT_cum_draws',
        'Off_strength_HT', 'Def_strength_HT', 'Off_strength_AT', 'Def_strength_AT',
        'HM1', 'AM1', 'HM2', 'AM2', 'HM3', 'AM3', 'HM4', 'AM4', 'HM5', 'AM5',
        'HT_3match_FormPtsStr', 'AT_3match_FormPtsStr', 'HT_5match_FormPtsStr','AT_5match_FormPtsStr', 
        'HT_3match_FormPts', 'AT_3match_FormPts', 'HT_5match_FormPts', 'AT_5match_FormPts',
        'HTWinStreak3','HTWinStreak5', 'HTLossStreak3', 'HTLossStreak5', 
        'ATWinStreak3','ATWinStreak5', 'ATLossStreak3', 'ATLossStreak5',
        'HomeTeamLP', 'AwayTeamLP', 
        'Is_HT_newTeam', 'Is_AT_newTeam',
        'H2H_Home', 'H2H_Away', 'H2H_Home_pts','H2H_Away_pts',
        'HomeTeam_TM_avg_age','HomeTeam_TM_foreigners_number', 
        'HomeTeam_TM_avg_market_value_MLN_EUR','HomeTeam_TM_total_market_value_MLN_EUR',
        'AwayTeam_TM_avg_age', 'AwayTeam_TM_foreigners_number', 
        'AwayTeam_TM_avg_market_value_MLN_EUR', 'AwayTeam_TM_total_market_value_MLN_EUR'
                              ]]

    SerieA_data_final.rename(columns={'HM1': 'HT_last_match1', 'AM1': 'AT_last_match1', 
                                   'HM2': 'HT_last_match2', 'AM2': 'AT_last_match2', 
                                   'HM3': 'HT_last_match3', 'AM3': 'AT_last_match3', 
                                   'HM4': 'HT_last_match4', 'AM4': 'AT_last_match4', 
                                   'HM5': 'HT_last_match5', 'AM5': 'AT_last_match5', 
                                   'HomeTeamLP': 'HT_last_LP', 'AwayTeamLP': 'AT_last_LP', 
                                   #'HTGD': 'HT_goal_diff', 'ATGD': 'AT_goal_diff', 
                                   #'DiffLP': 'Diff_last_LP',
                                   'HTP': 'HT_Pts', 'ATP': 'AT_Pts', 
                                   'HTSOT_T': 'HT_Shots_OT_took','ATSOT_T': 'AT_Shots_OT_took', 
                                   'HTSOT_A': 'HT_Shots_OT_allowed','ATSOT_A': 'AT_Shots_OT_allowed', 
                                   'HT_Shot_Acc': 'HT_Shots_took_Acc', 'AT_Shot_Acc': 'AT_Shots_took_Acc',
                                   'HT_Op_Shot_Acc': 'HT_Shots_allowed_Acc', 'AT_Op_Shot_Acc': 'AT_Shots_allowed_Acc'
                                  }, inplace=True)

    """## Save created dataset"""

    SerieA_data_final.to_csv(path_or_buf= 'Final_data/SerieA_data_final.csv', index=False)

    SerieA_data_final
```
